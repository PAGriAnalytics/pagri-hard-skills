## Дизайн исследования

Дизайн исследования подходит ко всем областям, например, если нам нужно проанализировать рынок труда,  
чтобы найти работу, то хорошим решением будет сначала задизайнить проект

дизайн - это каркас для сбора и анализа данных с целью согласования и логичного решения задач проекта 
- формулировка проблемы
- цель
- ключевой исследовательский вопрос  
- задачи 
    - описательная статистика (база знаний)
    это данные, которые мы можем получить из разных источников для анализа
    - сегментация
    выделяем из всех данных тот сегмент, который нам интересен  
    - сравнительный анализ
    сегментация и сравнительный анализ выполняется для конкретного сегмента (конечно бывает и для всего вобщем, но чаще для конкретного сектора)  
    это значит, что нужно определиться с сегментом, который нам интересен и сравнивать уже будем внутри этого сегмента  
    например, мы изучаем рынок труда, мы выбрали сегмент аналитики, далее мы сравниваем работадателей уже в этом сегмента
    - поиск инсайтов
    это значит, что для каждой задачи нам нужно найти инсайты, чтобы были аргументы,  
    например, мы идем на собеседование и инсайтом будет узнать о компании, что она из себя представляет, тогда эта информация нам поможет
    - предсказания
- набор методов и решений

- описываем шаги и задачи, которые выполнены в ходе проекта  
это план, который мы сначала до действий пишем и    
после выполнения проекта план превращаем в этот пункт

- Добавьте оглавление проекта с гиперссылками (опционально)
Так будет легче понять суть вашего проекта и перейти в нужный раздел.
4. Шаг 1
    * Когда получили любой результат (график, выборку, аггрегацию, сравнение переменных и дру), то мы не только пишем, что получили,  
    но что намного важнее, пишем почему это так. Думаем а как должно быть. Например, мы построили график зависимоти пола от количества тренировок и получили,  
    что чем старше, тем меньше тренировок. Вот тут очень важно подумать, а так и должно быть. В этом примере это очевидно, но  могут примеры,  
    когда это не очевидно. Именно отвечая на вопросы почему это так, а так и должно было получиться, мы можем получить правильные и важные выводы.  
    * описание данных. Предоставьте другую важную информацию: например, расскажите о файлах, с которыми работаете;
    * Подход такой - мы выполняем кусок кода в ячейке и видим результат, анализируем и если видим что-то неочевидно и важное, то это записываем,  
    не нужно писать воду и очевидные вещи, каждое предложение несет пользу, прежде чем написать что-то подумай, а какая от этого польза, если ответа нет,  
    то лучше не писать. Тогда все выводы и комментарии будут по существу. Так как много воды это плохо, так как она отвлекает от важных моментов и самое

    * Например, выполнили ячейку с кодом `df.info` мы увидели результаты и увидели что-то важное, записали, очевидные вещи писать не нужно, но нужно

    * И всегда помним, что мы сегодня это увидели, а завтра не заметим. Поэтому если что-то увидел в выводе ячейки, таблице, графике, то запиши,  
    самому же потом будет проще. И к тому же отчет потом будешь перечитывать и благодаря тому, что ты ранее записал мысли, они тебя сегодняшнего  
    натолкнут на новые мысли.  
    * Также, если мы видим, что отсутствуют значения, то нужно не просто это отметить, а лучше написать гипотезы, почему это могло произойти.  
    Особенно это важно на шаге изучения и предобработки данных, так как тут нужно много думать и могут быть хорошие выводы и важно помнить,  
    что читает человек, который не может заметить то, что ты можешь заметить, поэтому это и есть твоя работа, не только это заметить, а донести до  
    читающего

        - С чем коррелирует?

    Изучаем соседние колонки, что общего у этих записей. Возможно общие устройства, регионы и прочее. Вот у нас уже есть гипотезы,

## Оформление

Когда закончил проект, то представь, что ты посторонний человек и первый раз открыл его  
Какие моменты неудобны, что раздражает, что бы ты улучшил и исправь это  

- Хороший заголовок и подписи осей и легенд это уже 50 процентов эффекта. Так как если мы четко и точно назвали оси и заголовок,  
то читающий сразу поймет о чем речь. Так как сначал люди смотрят на график, а потом читают выводы. То первое впечатление будет от графика.  
И если на графике непонятный заголовок или подписи, то это уже будет большой минус, даже если под графиком все расписано хорошо.  
- Отельно нужно сказать про названия графиков. Это очень важно. Хороший заголовок сразу сообщает много инфы и человку нужно меньше думать.  
Например, название графика типа 'В декабре произошел рост продаж', намного лучше, чем 'Зависимость продаж от месяца'. Итак видно, что это зависимсоть.  
Нужно использовать заголовок как вывод. Это хороший ход. Так как человек смотрит на график, читает заголовок и понимает что происходит.   
Конечно заголовок должен быть краткий и нужно сжать смысл, но принцип хороший. Не нужно писать банальные заголовки, что это зависимость того то  
с тем то.
- Для графиков выбираем цвета сразу. То есть в начале выбираем цветовую тему, ставим ее по дефолту,  
чтобы все графики были в одной цветовой гамме. Можно выбирать любую, которая улучшает восприятие, но важно одну на все графики.  
- Всегда представляем, что мы работаем в паре с читающим. Тон должен быть 'мы получили', 'как видим', 'у нас было ранее' и прочее,  
то есть пишем, так как будто работаешь не один, а с кем-то  
- Делаем заголовки разного уровня, в зависимости от уровня вложенности и иерархии  
- Ведем нумерацию блоков и шагов  
- Используем списки и булиты  
- В начале делаем оглавление с сылками на все шаги, чтобы можно было перейти к нужному шагу  
- Используем форматирование  
- Выводы можно помещать в рамку, чтобы выделялось, но цвет лучше черный
- Добавляй пустые строки после строк с комментариями и разделяй логические блоки внутри ячейке также пустой строкой  
Это добавляет читабельности коду, чтобы не было все в куче. Это как абзацы в тексте, без них сложно читать.  
- Для таблиц лучше добавлять заговлоки по возможности, так как многие не будут смотреть код и так им будет сразу понятно о чем эта таблица



## Советы по составлению отчета

### Делать функции для действий, которые повторяются из проекта в проект

Например, преобработка содержит много проверок на нули, пустые значения, количество уникальных записей и другое, которое  
мы делаем в каждом проекте.  Поэтому нужно сделать функцию, которая будет получать на вход колонку или датафрейм и выдавать результат,  
который потом уже будешь анализировать.  
И вообе любые однотипные действия нужно формировать в функции, сохранять в отдельный модуль и потом просто использовать из проекта в проект.  

### **Разделите обзор и предобработку данных**

- В разделе «Обзор данных» вы применяете `head()`, `info()` и другие методы, позволяющие произвести первичное исследование данных;
- В разделе «Обработка данных» вы переходите к изменению данных: поиску дубликатов, работе с пропусками.

### Выделите общий вывод

Не совмещайте его с выводами по последнему разделу. Лучше сделайте общий вывод по проекту отдельным разделом.

### Оставьте **код, который помогал вам принимать решения**

Допустим, вы отправили тимлиду проект без пункта «в каких столбцах пропущены значения». Вы провели этот анализ, но потом удалили его и сразу перешли к обработке пропущенных значений. Тогда тимлиду придётся восстановить удалённый код — иначе нельзя будет оценить, все ли пропуски вы отыскали и обработали. В результате вы потратите время на удаление важной части кода, а ваш тимлид — на её восстановление.

Не следует удалять и проверки вроде «корректно ли прошло объединение таблиц», «достигнут ли нужный результат после применения функции» и т.п. Если вы регулярно проверяете работу кода, то проявляете себя как надёжный сотрудник. Это ценится.

### **Удалите неуспешный код**

Неудачный или неработающий код удаляйте, а не превращайте в комментарии.
### **Больше поясняйте**

В проекте важно описать процесс, а не только результат. Не скупитесь на пояснения: пишите свои мысли, интересные наблюдения и выводы. Из кода не всегда понятно, какими соображениями вы руководствовались и какие полезные выводы получили.
### **Разбивайте код**

Ячейки помогают сделать алгоритм решения более понятным. Не пытайтесь уместить всё в одной, но и на каждый «чих» создавать ячейку тоже не нужно. Одна ячейка — один важный шаг. Полезно сопровождать каждую ячейку коротким пояснением её назначения.

### Добавляйте ячейки с установкой библиотек
Если это нестандартная библиотека, типа пандас и нампай, то сделай ячейку перед использованием этой библиотеки,   
в которой будет `!pip install module_name`   
Чтобы, если у читающего нет этой библиотеки, он не искал как ее поставить.  

### **Оформите графики**

- **Дайте им названия.** Например: «Итоговые продаж игр по годам (шт)», «Зависимость продаж игр от оценок критиков», «Гистограмма распределения среднего количества сообщений в месяц по двум тарифам».
- **Подпишите оси.** Например: «год выпуска», «оценки критиков», «количество сообщений».

### Общий вывод можно построить так:

- **Краткий обзор проведённой работы.** Что выполнили, сделали, рассчитали;
- **Главные выводы.** Дайте ответы на поставленные вопросы;
- **Рекомендации**. Например, вы можете посоветовать отобрать определённые признаки для прогнозирования в будущем, или дать рекомендации о том как избежать появления дубликатов или пропущенных значений.

### Как использовать выводы

Любые важные моменты, которые мы увидели во время изучения данных, должны приводить к выдвижению гипотез. Это не должно быть фактом сразу.  
Вот это нужно понимать. Работа аналитика приводит к возникновению гипотез, которые ещё нужно проверять. Поэтому наша задача проанализировать и 
выдвинуть гипотезы, которые должны быть основаны на данных. Это не просто догадки и предположения, это именно гипотезы выдвинутые после изученя данных и  
они основываются на данных. Поэтому очень важно в выводах следить, чтобы не добавлять свои фактазии и соображения, которые не основываются на полученных  
результатах. Все выводы только на данных.  

### Итоговые выводы привязывай к данным
То есть когда делаешь выводы, то пиши типа, ваша система лояльности могла бы работать лучше, основываясь на полученных данных.  
Или исходя из результатов анализа данных делаем вывод. То есть data driven подход должен пронизывать твои выводы. Не должно быть ощущения, что ты  
делаешь выводы основываясь на своих догатках и интуиции. Тебе это может быть не заметно, но со стороны будет намного лучше, если ты будешь всегда  
подчеркивать, что выводы основаны на данных.  

### **Немного психологии**

Пишите проект от первого лица множественного числа: ****сделали, рассчитали, заменим. Так читающий станет сопричастен и не будет чувствовать, что им командуют.



### Делайте комментарии в коде
Когда идет вычисление чего-то и нет результата под ячейкой,  
либо когда не очевидно, что делается в строке  
например `2 * np.pi / 3   # 120 `  
вот такие комментраии очень полезны, чтобы в голове не нужно было вычислять  

### Составление плана проекта 
Начинамем с основных этапов, потом каждый этап разбиваем на подэтапы, и так далее,  
наша задача, чтобы каждая задача была атомарной еденицей, которую можно решать и не думать как это делать.  
Хорошо составленный план, детализированные шаги экономит очень много времени, так как не тратится время перед каждым шагом.  
Ты просто идешь по плану и все. Конечно в процессе работы план может корректироваться.  
И конечно нужно составить шаблон отчета и туда постоянно довносить важные пункты, но это не значит, что нужно брать все из шаблона.  
Для каждой задачи берем свое и подгоняем под конкретный случай.
