Таблицам исходникам (те, которые загрузили) даем префикс `s_`, чтобы отделить от остальных созданных

Если есть даты, то создаем календарь (новую таблицу)  
Это нужно, чтобы связывать разные таблицы через дату    
Переходим на вкладку таблиц, нажимаем новая таблица  
В верхней строчке (возле xv) пишем `d_calendar = CALENDAR(DATE(2015, 1, 1), DATE(2020, 12, 31`  
Также можно указать, что созадть автоматом диапазон, тогда диапазон будет как у даты в нашей таблице.  
Далее идем во вкладку соединений и соединяем дату в нашей таблице с датой в календаре                                                                                   


На графиах с датами можно провалиться на уровень ниже, для этого в верхней части окне графика нажимаем соответствующую иконку    
И в таблицах также можно проваливаться также  

Если мы хотим к графику добавить таблицу или дргой графический объект, то можно скопировать имеющийся график ниже, выбрать его и нажать иконку соответствующего  
графического объкта в меню справа. Не обязательно ещё раз создавать пустой объект и добавлять в него колонки по осям. Можно использовать уже готовый график.  


Если у нас есть пропуски в колонке, но также есть другой столбец, для которого в этих местах нет пропусков, и мы видим, что в предыдущие периоды у нас есть   
значения без пропусков, то есть мы можем их идентифицировать по второй колонке 
например, у нас есть описание с пропусками и вторая колонка коды, мы берем эти две колонки сортируем по коду и видим, что для одного кода есть описание и есть пустое  
значение.  
Чтобы восстановить данные создаем новую таблицу на вкладке таблиц  
и в верхней строчке пишем  
`d_group = ALL(название колонки с кодами без пропусков, название колонки с описанием с пропусками`  
`ALL` создаст таблицу с уникальными строками  
Теперь нам нужно ввести данные из этой таблицы (то есть мы создали таблицу, и теперь делаем ее испочником, добавляем в наш дата сет), чтобы получился справочник,  
где будет код и описание, и потом мы соеденим эту таблицу с нашей основной  
Для этого нажимаем `Enter data`, вводим название, например, `d_code_group`, нажимаем `load`  
И теперь на вкладке соединений таблиц соеднияем эту таблицу справочник кодов с нашей основной таблицей по полю кода (поле без пропусков)  
И теперь мы в нашу таблицу или график добавляем поле описание (поле код тоже можно из справочника, но главное описание брать из справочника)  
не из основной таблицы, а из таблицы справочника.  

Если мы вывели на бар плот с наколпрелением легенды (доабвили разбивку по категориями для оси икс) и у нас легенд много (значений третьей категории), то  
можно фильтровать. В фильре выбираем эту категорию для легенд и в ней выбираем например топ n и ставим значение сколько нужно и в `by` перетаскиваем  
этот же столбец, и выбираем подсчет значений count (по автомату будет просто браться первые) так как мы хотим его отфильтровать по нему же тогда у нас в легендах останутся только топ n категорий по количеству

То есть удобно формировать простенькие отчеты.  

Важно  
После загрузки данных через power query или напряму обязательно проверяем целостность связи с источником.  
Это делается просто - мы открываем источник и считаем какую-нибудь аггрегацию по всей таблице, например, сумму продаж.  
И в power bi добавляем элемент на дашборд с табличкой и смотрим также сумму продаж и если сходится, то все отлично.  
Этот этап простой, но позволяет сразу заметить проблемы в загрузке данных.  

Если у нас в Power query пришла сводная таблица, то чтобы превратить ее в обычную (сделать melt как в пандасе), нужно  
- выделить колонку, которую мы хотим не трогать (это колонка которая верно расположена, например у нас была группировка по полю и это столбец как нам и нужно)
- идем на вкладку преобразование и находим пиктограммку отменить свертывение столбцов
- выбираем отменить свертывание других столбцов (одни мы выбрали и он не тронется)

Хорошая практика, когда мы соединяем две таблицы, то скрываем поле в таблице фактов,  
чтобы у нас на панели визуализации не было двух полей с одним названием из разных таблиц, если мы соединяем,   
то использовать будем только одно поле. Поэтому оставляем только поле в справочнике.  
И так постоянно, приучаемся делать 2 денйствия одновременно  
- соеденили таблицы
- скрыли поле, по которому соединаяли из таблицы фактов

Добавлять высичляемые поля мы можем 2 способами
- создать меру  
Это вычисляемое поле на лету (то есть данные не хранятся в памяти)  
Важно, что при создании меры, мы должны указывать способ агрегации данных.  
Поэтому если мы хотим получить разность двух столбцов, то мы должны выбрать как   
они будут аггрегировать перед разностью.  То есть всегда когда мы будем  
строить какие-то группировки, то мера будет агрегироваться определенным способом.  
И в отлии от столбца мы не сможем выбрать способ агрегации на панели моедлирования.  
- создать столбец  

В dax название талбиц идет в ковычках, а названия столбцов в квардратных скобках.  
Поэтому когда мы хотим чтобы у нас появислся выподающий списков при наборе формулы, то мы просто  
ставим ковычку и выбираем что нам нужно.  

Если мы добавляем скрытую фильтрацию на странице, то нужно обязательно указать это либо в названии листа (либо в названии кнопки листа, либо в названии  
самого листа вверху). Это очень важно, так как пользователь не знает о скрытой фильтрации и может быть введен в заблуждение.  

Если у нас по оси X дата, то power bi автоматически скрывает даты без данных, таким образом мы вводим в заблуждение.  
Поэтому не забываем для оси даты переходить в блок настроек виджета с датой и в поле где мы вставили дату нажать на дату правой кнопкой мыши  
и выбрать покаазть элементы без данных.  
Это может и в других местах встречаться не только с датой, поэтому не забываем, что по умолчанию power bi не показывает элементы без данных.  

Очень удобно делать drop down для созданных категорий.  
Например, мы сделали новую категорию из возраста, и у нас получились группы возрстов.  
Теперь мы на график помещаем на верхний уровень эту категориальную переменную, а на уровень ниже нашу числовую переменную,  
таким образом мы можем проваливаться на уровень ниже.  

Есил мы хотим сделать на одном листе слои, которые будут показываться отдельно, например, таблица и график для этой таблицы,  то 
используем закладки. Создаем закладку (это по сути слой) с таблице и закладку с графиком.  
И настраиваем видимость элементов на закладке.  Далее добавляем кнопки,  которые будут переключать видимость закладок.
Для этого создаем кнопки и привязываем  их к событиям на закладках.  В событии на закладке мы указываем ,  
что при нажатии на кнопку,  должна быть показана эта закладка и  должны быть скрыты все остальные.  В результате у нас получится, что на одном листе
будет две закладки,  которые будут показываться отдельно,  и две кнопки, которые также будут показываться с каждой закладкой.  
Вместо кнопки можно использовать любое изображение, у которого в меню включаем действие и указываем какая закладка активируется при ее нажатии.  
Мы можем взять несклько картинок переключателя, добавить текст сбоку от переключателя и у нас будет переключатель.  



Когда мы создаем закладки, то фильтры будут сбрасываться на каждой закладке,  
чтобы этого не было, мы в настройках закладки снимаем галочку с данные, то есть данные не будут обвновляться при выборе закладки.  
Таким образом закладки будут менять только отображение эелементов.  

Если у нас в дашборде больше 3-4 листов, то лучше делать титульный лист с оглавлением, который будет упрощать навигацию.  
Создаем лист (первый) на него добавляем кнопки для каждой страницы дашборда.  
Можно добавить логотип компании.  Также удобно добавлять дату последней операции в отчете, чтобы было понятно, смотрел он эти данные или нет ещё.  
Если дашборд обновляется, то это очень удобно, сразу понять нужно ли смотреть.  
Также на каждой странице создаем кнопку к содеражнию или хотя бы добавляем кнопку назад, чтобы можно было быстро вернуться к содержанию.  
Не забываем для кнопок добавлять смену фона на более темный при наведении на нее, так сразу понятно, что это кнопка а не фон. 

Когда делаем карточку товара или другой детализированный отчет, то есть создали отдельный лист, в его настройках вклчюиил детализированный отчет,  
добавили туда же поля по которому будет появляться детализированный отчет, включаем сохранение фильтрации там же в настрйоках листа.  
Тогда когда на нужном листе в нужном элементе, где есть поле, которое мы добавили в детализированный отчет (то есть по нему будет детализация),  
будет выбран пункт меню детализированный отчет, то при переходе в детализированный отчет, фильтры страницы сохраняться.
