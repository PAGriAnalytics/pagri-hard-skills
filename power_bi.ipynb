Аккаунт в Power BI можно зарегистрировать только с корпоративной почты. Если у  
вас нет рабочей почты, можно использовать сервисы временной почты, например,  
10 Minute Mail: временная почта понадобится вам для создания учётной записи и  
получения письма с подтверждением для завершения регистрации.
