Очень полезно создавать свою базу данных о рынке, отрасли, в которой ты работаешь.  
То есть мы собираем данные, чтобы изучить среду и отрасль. Информацию о конкурента (какие у них конверсии, мау, дау и прочее), информацию о рынке.  
И все это обязательно понадобиться в будущем. Нужно будет сравнить наш мау с мау рынка и конкурентов. Или кто-то спросит что-то о сравнени.  
И ты берешь делаешь запрос в базу и получаешь ответ. Поэтому нужно все данные, которые мы собираем в процессе работы,  
отправлять в базу данных. Нормализовывать, чтобы потом можно было быстро получить ответы.  
Это может быть база данных настоящая на сервер, моет быть даже гугл таблицы, к которым можно быстро сделать запрос из питона.  
Главное аналитик должен всю информацию сохранять, если она не сохранятся в компании автоматически.   
Запонмнить все нереально, а искать снова это долго.  

При получении задачи очень важно подумать на какие части она делиться. Берем предложение с задачей и делим  
его на логические части. Это также нам даст поянть какие данные нам нужны (какждая логическая еденица задачи детализирует данные)
Например, у нас задача 'Какие категории и подкатегории в поисковых системах приводят больше пользователей'  
Разбиваем на части
- Какие категории и подкатегории (сразу понимаем, что нам нужны данные по категориям и подкатегориям)
- в поисковых системах (нам нужны данные где для каждой категории будет метрика по поисковой системе, например количество входов по категории)
- приводят больше пользователей (нам нужен механизм, чтобы определить какие лучше, например, можем взять суммарное количество входов по категориям)
Вот мы разбили задачу на логические составляющие. Не путать с планом. Это другое. Тут нам нужно поянть саму задачу и понять какие данные нам нужны.  

## Генерация идей и гипотез

### Быстрый анализ среды

1. Отрасль (конкуренты, новые игроки на рынке, заменители, цеполчка участников)
Посмотреть про место которое занимает на рынке, посмотреть конкурентов, посмотреть похожих игрово 
Для этого хороший сайт similarweb, чтобы посмотреть про сайт компании, которую изучаем
2. Рынок (сегменты, спрос, переход к конкуренту, влияение на цену, движущие силы)
Ищем отчеты о рынке. Достаточно бесплатных отчетов. Вбиваем в поиск нашу отрасль и пишем аналитический отчет  
Читаем, чтобы понять какие есть сегменты, какой спрос, кто лидер рынка, кто его конкуренты, что влияет на цену и прочее
3. Тенденции (технологические, законодательные, общественно-культурные, социоэкономические, сырьевые)
Идет в поиск и забиваем тенденции и отрасль нашей компании.  
Наша задача понять что нас может ожидать и что происходит прямо сейчас.  
Какие тенденции сейчас на рынке, в каком направлении будет развиваться отрасль, что является драйвером и прочее.  



### Карта эмпатии

Представляем какого-нибудь воображаемого конкретного нашего покупателя

И отвечаем себе на следующие вопросы
- Что он думает и чувствует?  
что на самом деле важно, основные занятия, проблемы и радости
- Что он видит?  
среда, друзья, что предлагает рынок, где живет и работает, 
- Что он говорит и делает?  
как держит себя в обществе, как выглядит, отношение к окружающим
- Что он слышит?  
что говорят друзья, что говорит начальник, что говорят влиятельные лица

Также представляем какая у него боль, страхи, разочаровния и препятствия

И думаем какие у него достижения, потребности, желания, мерило успеха, препятствия

Теперь когда мы много  знаем о нашем покупателе, мы можем пойти от частно к общему.  
Мы можем понять
- что он хочет
- в какой момент он этого хочет
- в каком канале он этого хочет
- как это должно выглядеть
- как мы его можем завлечь

В итоге мы получили возможные точки роста.  
Мы знаем как сделать так, чтобы она принесла нам больше денег.  

Дополнительно можно придумать целую историю о придуманном пользователе.  
Например, у нас аналитик Миша, который живет в новом жк, увлекается спортом, но времени мало.  
Переехал Миша в новый жк, а там велодорожка или спорт плошадка классная. И захотел Миша купить себе то-то и то-то и так далее приудмываем.  
В итоге можно так набрести на отличные точки роста.  

Далее представляем других наших покупателй, отличающихся друг от дурга.  
В итоге мы получаем сегментацию наших покупателей.  

### Идеи основанные на данных

Очень часто в выбросах могут лежать идеи.  
Например, у нас в продажах есть аномально большие покупки у нескольких покупателей.  
Мы смотрим и это начало февраля. Возможно это покупки для подарков. Следовательно появляется точка роста - активизировать рекламу в это время и в другие  
подобные праздники.  

Много идей можно выловить в пабликах соцсетей.  
Нужно проанализировать комментарии сообществ связанные с нашей отраслью / сферой.  
Приходит понимание чего хотя люди, что ими движет в этой отрасли.  

Берем парсим комменты к пабликам. Ищем сколько сообщений с ключевым словом, которое нас интересуюет.  
Например, мы продаем велосипеды, тогда мы ищем сколько таких сообщений. Сколько лайкают, репостят.  

Также берем и смотрим какие слова чаще встречаются вообще. Если например, часще встречаются в паблике про туризм  
слова мотоцикл, квадрик, то мы уже получаем точки роста, то есть мы можем начать больше продавать все что с этим связано, аксесуары и прочее.  
Ещё можно посмотреть не просто, что чаще встречается, а взять топ 1000 или более и уже из них посмотреть какие слова чаще встречаются.  
Так мы поймем на что люди реагируют более активно.  

Важно посмотреть результаты запросов в яндексе и гугле о том, какие слова из нашей сферы больше всего  
спрашивают.   
https://wordstat.yandex.ru/  
Словоеб.рф позволяет без токенов быстро получить данные с wordstat  
Там же посмотреть сезонность.    
Таким образом мы на больших данных можем посмотреть что сейчас происходит.  
Схема тут следующая
- мы берем все категории наших товаров
- смотрим сколько люди ищут в поисковиках слова связанные с каждой категорией
- в итоге получаем какие категории лучше развивать, а какие не пользуются спросом

В итоге мы видим какие наши категории пользуеются спросом какие нет, можем скорректировать наши действия.  
Плюс мы увидели какие вообще слова ищут люди в нашей сфере.  
И нам было бы неплохо поучаствовать в тех, которые ищут много людей. Это  потенциальные точки роста.

Сервис Blueprint

Service Blueprint — это визуализация взаимодействия между клиентами и различными элементами сервиса компании.

Она представляет собой определённую схему или диаграмму,  
которая воспроизводит процесс предоставления услуг и коммуникации с потенциальными потребителями на всех этапах их сотрудничества с брендом.

Service Blueprint позволяет:

наглядно подтвердить, какие процессы в компании требуют оптимизации;

оценить, насколько эффективно работает тот или иной сервис;

выявить, что можно улучшить или упразднить.
Если мы работаем в вебе или мобайле, то скорее всего у нас есть возможность сразу посмотреть что просиходит.  
То есть, например, мы находимся на нашем сайте. Кликаем куда-то и смотрим в яндекс метрике как меняются метрики.  
Тут суть в том, что если у нас появилась какая-то гипотеза, то мы можем пойти и проверить ее.  
Также удобно это использовать, когда хотим проверить как отображается в данных что-то особенное.  
То есть у нас появлиась гипотеза выбросов например, и мы идем и создаем сами эту ситуацию на сайте и  
сразу смотрим на результат метрики. И можно понять верная гипотеза или нет.  


### Источники вдохновения

Схема WWWWWH (Who What Why Where When How)
Задаем себе всевозможные вопросы с этими словами о нашем сервисе
- Кто использует наш товар?
- Для чего использует?
- Где испльзуют?
- Когда использует?
- Почему использует?
- Как использует?

Важно, что ответы односложные.  
В итоге по ответам можно найти новые точки роста, о которых мы не думали.  
На основе ответов можно придумывать какие-нибудь идеи для развития сервиса.  

Тут ещё полезно собрать все ответы на вопросы и закидываем в рандомайзер  
Например в https://ciox.ru/select-a-random-word-from-the-list  
или модуль random в питоне  
Выбираем 2 случайных слова и пытаемся состыковать, найти ассоциацию. То есть мы придумываем как эти 2 слова быть вместе.  

Таким способом можно искать возможные точки роста.  
Также так можно создавать банеры, так как некоторые ассоциации не придут в голову просто так,  
а тут могут быть очень интересные идеи.  

Brainwriting

Например, у нас 6 человек сидят по кругу  
Это может быть онлйн и любое количество идей
6 (человек) - 3 (идеи) - 5 (минут)
- Озвучиваем задачу, например, как нам увеличить количество посетителей
- За 5 минут каждый пишет на бумажке 3 идеи как достигнуть цели или что можнет помочь
- Далее каждый передает свою бумажку рядом сидящему и каждый уже на основе 3 написанных идей,  
их расширяет, углубляет, разворачивает мысль, может совмещает.  
То есть они пишит свои идеи на основе тех, которые написаны.  
Это могут быть, либо 3 идеи на основе имеющихся трех, либо он может выбрать 1-2 и на основе их расширить
И важно, когда написал, то перед передачей рядом сидящему, он закрвает предыдущие идеи,  
чтобы остались только его идеи
И каждый раз дается 5 минут
- И так по кругу, пока бумажка не вернется к тому, кто ее первый написал (пройдет один круг)

В итоге за 30 минут мы получим очень много идей, причем размвернутых
