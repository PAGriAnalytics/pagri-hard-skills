Задача любого бизнеса -  увеличить прибыль.  
Но чтобы бизнес начал приносить прибыль, нужно найти пользователей, которые будут его покупать.  
Чтобы найти пользователей, нужно понять их боль, что им нужно. И далее , нужно создать продукт, который будет решать эту проблему.  
Нужно определить где могут находиться пользователи, которым нужен наш продукт, чтобы решить их проблему.  
Далее нужно провести опрос и убедиться, что наш продукт действительно решает проблему.  
Далее нужно создать mvp (минимально жизнеспособный продукт), который будет решать проблему.  
Это этап проверки соответсвия нашей идеи проблемам пользователей. Может ли она его решить.
Если наши предположения о продукте верно (что он позволит решить проблемы пользователей), то пользователи будут покупать его.  
Далее нужно развивать продукт и масшатбировать его.  
Это этап достижения соответствия продукта рынку. Когда мы уже не для пары пользователей сделали продукт и убедились, что он решает их проблемы, а уже для большого количества пользователей. И тут возникает множество проблем, даже если продукт решает проблемы пользователей. Это и конкруенция, привлечение клиентов, масштабирование и другое.  
На данном этапе компания может полностью изменить свою идею, это называется pivot.  


Основные вопросы создателей продукта и продукт менеджеров
- Нужно ли людям то, что я хочу сделать?
- Кто эти люди и где их найти?
- Каким должен быть продукт, чтобы его захотели купить?

Customer development - это подход к созданию стартапов и новых продуктов, когда через взаимодействие с потенциальными пользователями валидируются гипотезы о проблемах потребителей, ее решении, рынке, каналах привлечения. 

Задача продукт менеджера, чтобы продукт развивался не на предположениях команды, а на реальных данных, которые получены при взаимодействии с пользователями.  
Иначе высока вероятность сделать то, что не нужно пользователями.  

Очень важно сегментирование пользователей.  
Сегментирование позволяет выявить именно тот сегмент, которому будет интересен наш продукт.  
Далее сегментация позволяет уже внутри имеющегося сегмента (потенциальных пользователей) выделить группы, с которыми нужно взаимодействовать по разному.  
У пользователей разные запросы, потребности, мотивации. Разным пользователям нужен разный функционал продукта. Эти пользователи несут для компании разную ценность.  

Сегментация нужна для привлечения клиентов, так как для разных групп нужны разные рекламы. У всех групп свои мотивации и боли.  
И сегментация помогает создавать фичи, так как для разных сегментов нужны разные фичи в продукте.  
И в business model canvas в разделе пользователи нужно указывть именно сегменты пользователей.  

Задача маркетинга - донести ценность продукта до потенциальных клиентов, чтобы они купили его.

### Фреймворки для взаимодействия с пользователяим

#### Метод персон  

Метод персон  
Собирается образный персонаж потенциального пользователя, он максимаьно реалистичен. Он описывает поведение потенциальной группы пользователй.  
Суть в том, что мы уходим от многих людей к одному, так намного проще продумывать что нужно конкретному пользователю. А затем масштабируем на всех.
- имя
- фотография
- демографические данные
- предыстория
- цели
- мотивация
- барьеры

После того как мы сотавили портрет потенциально пользователя, то нужно проникнуться его проблемами.  
Нужна эмпатия.  
Эмпатия - осознанное сопереживание эмоциональному состоянию другого человека.
Виды 
- эмоциональная (когда мы на уровне эмоций чувствуем человека)
- когнитивная (когда мы глазами другого человека смотрим на мир и за счет этого понимаем его мысли, чувства, но без своих эмоция, рационально)

Для развития продукта важна когнитивная эмпатия, так как привлечение эмоций может мешать и искажать действия.  

Карта эмпатии

Представляем какого-нибудь воображаемого конкретного нашего покупателя

Чтобы лучше запомнить, можно представить говолоку и сверху идти.  
Вверху мозг (что думает, чувствует), далее глаза (что видит), далее рот (что говорит), далее уши (что слышить) и внизу тело (что делает).  
Тут важно, чтобы предолжения, которые мы пишем в каждый раздел перекликались с другими.  
То есть мы пишем, что он что-то видит или слышит, то скорее всего он и говорить будет на вещи, связанные с этим.  

И отвечаем себе на следующие вопросы
- Что он думает и чувствует?  
что на самом деле важно, основные занятия, проблемы и радости
- Что он видит?  
среда, друзья, что предлагает рынок, где живет и работает, 
- Что он говорит и делает?  
как держит себя в обществе, как выглядит, отношение к окружающим
- Что он слышит?  
что говорят друзья, что говорит начальник, что говорят влиятельные лица

Также представляем какая у него боль, страхи, разочаровния и препятствия

И думаем какие у него достижения, потребности, желания, мерило успеха, препятствия

Теперь когда мы много  знаем о нашем покупателе, мы можем пойти от частно к общему.  
Мы можем понять
- что он хочет
- в какой момент он этого хочет
- в каком канале он этого хочет
- как это должно выглядеть
- как мы его можем завлечь

В итоге мы получили возможные точки роста.  
Мы знаем как сделать так, чтобы она принесла нам больше денег.  

Дополнительно можно придумать целую историю о придуманном пользователе.  
Например, у нас аналитик Миша, который живет в новом жк, увлекается спортом, но времени мало.  
Переехал Миша в новый жк, а там велодорожка или спорт плошадка классная. И захотел Миша купить себе то-то и то-то и так далее приудмываем.  
В итоге можно так набрести на отличные точки роста.  

Далее представляем других наших покупателй, отличающихся друг от дурга.  
В итоге мы получаем сегментацию наших покупателей.  

#### Jobs to be done (jtbd)

Работа, которую необходимо выполнить.  
Это реализация неудовлетворительной потребности или желания (в ответ на тригер)

То есть тут нужно фокусироваться на работе, которую нужно сделать пользователю, чтобы решить его проблему.  
Например, сбросить лишний вес.

В отличии от метода персон, когда у нас много лишней информации, которая напрямую не ведет к понимаю того, что нужно пользователю, в методе jtbd цель именно определить работу, которую нужно сделать пользователю, чтобы решить проблему.  

У всех людей постоянно стоят задачи, которые нужно решить.  
И это не какие-то сложные задачи, а такие как 
- заполнить чем-то время пока стоишь в очереди, едешь в метро
- поделиться чем-то интересным с другими
- сходить в магазин

И тут смысл в том, что мы как бы нанимаем конкретный продукт, чтобы он выполнил эту работу за нас.  

То есть продукт сделает жизнь человека лучше или легче. 

Формула 
- контекст
- мотивация
- результат

When (situation) I want to (motivation) So I can (expected outcome)

Когда я еду в метро, я хочу посмотреть сериал, чтобы время быыстро прошло.  


Тут важно, чтобы все 3 части были конкретны и связаны с продуктом.  
Например, для службы доставки: Когда я пришел домой, я хочу чтобы продукты были  доставлены, чтобы сэкономить время.  
Не совсем конкретно, jtbd подразумевает конкретику
Нужно так - Когда я пришел домой уставший и не хочу идти в магазин, я хочу, чтобы продукты были доставлены в течении максимум 1 часа, чтобы мне не нужно было ехать или идти за продуктами, сэкономить время для более приятных вещей.  

по сути три вопроса 
- когда 
- я хочу
- чтобы

Если мы так начинаем думать, то мы можем разработать новую фичу, которая поможет пользователям.  
То есть мы думаем, когда пользователь заходит на сайт, он хочет быстрее выбрать товар, чтобы не тратить время.  
И если найти более детальные потребности, то мы можем сделать так, чтобы решить эту проблему. 

Вопросы, которые помгут
- Для какой работы пользователь нанимает потенциальный продукт?
- В каком контексте или ситуации?
- Что нужно пользователю на каждом этапе работы?
- Что пользователь пытается улучшить?
- О чем ещё пытается думать пользователь?

#### Value proposition canvas

Шаблон ценностного предложения

Профиль клиента
- Задачи (jbtd)
- Проблемы и сложности
- Преимущества и удобства (что польователь ожидет получить после выполнения работы)

Тут задача - это то, что нужно сделать, чтобы решить проблему.  
Например, человек едет в метро, проблема, что ему скучно, работа - это то, что нужно сделать, чтобы решить проблему, чтобы человек не скучал. Выгода - это хорошее настроение.  

Профиль ценностного предложения (продукта)
- товары и услуги (то что мы пытаемся предложить пользователю)
- факторы выгоды (описание того, как именно продукт поможет достичь выгоды пользователю)
- факторы помощи (описание того, как именно продукт решает проблему пользователя)

Если профиль пользователя согласуется с профилем продукта, то решение согласовано.  
Problem solution fit

Очень важно ценностное предложжение продумывать для каждого сегмента пользователей. Для этого нужно сначала сегментирвоать пользователей.

И чтобы сделать продукт лучше нужно формировать гипотезы, основанные на взаимодействие с пользователями. 
И важно эти гипотезы проверять с помощью экспериментов, аб-тестов, фокус групп, интервью и других методов.

### Основы управления продуктом

Видение и стратегия - основополагающие элементы в продукт менеджменте.  
Порядок вопросов
- Почему и для чего (идея, цель)  
Цель должна быть по SMART (specific, measurable, achievable, relevant, time-bound)  
Неправильно - увеличить популярность фичи Видеозвонки  
Правильно - Увеличить дневную аудиторию юзеров, пользующихся фичей Видеозвонки, на 10% за 3 месяца (до 2млн пользователей в сутки)
- Как (стратегия, продуктовый roadmap или дорожная карта продукта)
- Что (продукт)

vision - strategy - product

Перспектива 3 3 3  
3 Года  
Продуктовые идеи и видение нужно строить на горизонт 3 года  
3 Месяца  
Стратегию нуно строить на 3 месяца. Так как в мире все быстро меняется и если больше времени тратить, то может оказаться, что стратегия не работает и время было зря
3 Недели  
Так называемые спринты занимают 3 недели. То есть улучшения в продукте должны быть достигнуты в течении 3 недель.

Компания должны отвечть на три вопроса
- Какова ваша основная цель на 3 года и какие критерии успеха?
- Какая у вас стратегия для достижения этой цели?
- Что вы собираетесь делать, чтобы проверить свои стратегии на валидность? Как будете экспериментировать и проводить спринты?

Основные бизнес-модели
- Freemium (бесплатный, но с ограничениями, а полная версия платная, компания рассчитвает, что пользователю понравится продукт и он перейдет на платную версию)
- E-comerce (продажа товаров через интернет-магазин)
- Маркетплейс (продажа товаров через интернет-магазин, но товары не принадлежат компании, а продаются другими продавцами) прибыль получают за счет комиссии
- Subscription (подписка на продукт, например, Netflix)
- IaaS (Infrastructure as a Service), PaaS (Platform as a Service), SaaS (Software as a Service)

### Управление проектами

Ключевые фазы проекта
- инициализация (выдвижение идеи и подготовка проектных документов)
- планирование (определение ресурсов, распределение задач, расчет бюджета)
- выполнение (выполнение работ)
- контроль (контроль выполнения работ)
- завершение (закрытие проекта)

Waterfall
- ключевые особенности - последовательнсоть и ориентация на требования
- основные этапы - сбор и анализ требований, разработка и процесс внедрения решения и исправление проблем
- каждый этап этого процесса автономен. Чтобы перейти к следующему этапу, необходимо завершить предыдущий

Agile (проворный)  
Характеризуется небольшими циклическими изменениями, которые внедряют в ответ на изменения требований.  
Всегда ориентируется на обратную связь от пользователя или от стэкхолдера.  
Цель каждой итерации - создать рабочий продукт или хотя бы фичу, то есть создать ценность.  

Ценности agile
- люди и взаимодействие превыше процессов и инструментов
- рабочий продукт важнее исчерпивающей документации
- сотрудничество с клиентом важнее согласования условий контракта
- готовность к изменению важнее следования изначальному плану

Agile это больше философия

Scrum  
Подход при котором в жестко фиксированные и небольшие по времени итерации (спринты), команда работает над продуктом, и в конце каждой итерации предоставляет конечному пользователю дополнительную ценность (инкремент).  

- каждый спринт планируется (ставится задача на спринт)
- каждый день проводится стэндап (скрам митинг) для мелких коректировок цикла
- в конце спринта проводится ревью спринта. Оценивается процесс и ценность результата для продукта.  
- ретроспектива (команда оценивает прошедший спринт и планирует усовершенствования для следующео спринта)

Бэклог - это приоретизированный список требований и задач.

Крупные задачи назваются эпиками.  

Груминг - причесывание задач. Когда после каждого спринта корректируется список задач после обсуждения с заказчиком. 

Kanbab

Kanban - это подход баланса.   
Тут команда не делится, и эта методология хорошо подходит цельным командам, например, команда аналитиков.  
Основная задача - сбалансировать разных специалистов внутри команды и избежать ситуации неоптимального расходования времени.  
Принципы
- вся команда едина
- процесс делится не на спринты, а на стадии выполения конкретных задач
    - планируется
    - разрабатывается
    - тестируется
    - завершено
- главный показатель эффективности - время прохождения задачи по доске.

### Юнит экономика

Юнит экономика смотрит на бизнес через призму одной еденицы (пользователь, покупатель, транзакция).

Задача юнит экономики понять зарабатываем мы на одной еденице или нет. 

Для этого нужно определить сколько мы потратили на пользьвателя и сколько заработали.  

Юнит экономика дает следующие возможности
- оценивать рентабильность бизнеса
- оценивать и приоретизировать гипотезы
- определять эффективность основных каналов привлечения пользователей
- находить перспективы и проблемы в продукте и определять вектор развития продукта.  

Если с ростом количества товаров нам нужно больше определенных людей, то их зарплаты нужно отправлять в cogs.  
А если рабочая сила не зависит от объема производства, то их зарплаты нужно отправлять в opex (operating expenses).

### Приоретизация гипотез

Гиптезы бывают
- гипотезы ценности (предположение о том в чем видят ценность нашего продукта пользователи)
- гипотезы роста (предположение связанные с ростом метрик)

То есть гипотезы ценности фомируются как что-то что может увеличить ценность продукта для пользователя.  
А гипоетзы роста - как что-то что может увеличить метрики продукта.  

При генерации гипотез важно проверять себя вопросами
- как знания полученные в результате проверки этой гипотезы помогут продукту, то есть что я хочу узнать по этой гипотезе
- как предложенные изменения повлияют на продукт
- является ли гипотеза приоритетной

Фреймворки для приоретизации

- impact / effort  
Гипотеза оценивается по двум критериям - влияние на продукт и трудозатраты на проверку гипотезы  
Делим плоскость  на две оси (impact и effort) и получаем четыре квадранта:  
в каждом квадрате идет - quick wins (проверяем в первую очередь), big bets (несут много ценности, но нужно больше времени на реализацию), maybes (ценности не дадут, но легко реализовать), time sink (не нужно даже проверять)  
- rice (reach impact confidence effort)  
важно каждую оценку считать на одном периоде для всех гипотез, то есть если мы охват считаем за месяц для одной гипотезы, то для других также нужно за месяц   
confidence зависит от того, насколько наши оценки для impact и rice основаны на данных, то есть если это просто наши убеждения, то тут оценка низкая, а если например охват у нас получен из crm, то оценка будет выше.  
rice score = r * i * c / e

В иделае impact оценивается по выручке, то есть насколько вырастет выручка от реализации этой гипотезы

Шаблон дизайна эксперимента
- Что проверяем (гипотеза)?
- Что делаем в продукте?
- На каких пользователях тестируем?
- На какие метрики смотрим?
- Какой эффект ожидаем? Какой размер выборки нужен?
- Какой план действий в зависимости от результатов эксперимента?

Схема следующая
- мы формируем гипотезу, важно чтобы они были конкретными и измеримыми и были связаны с метриками, которые связаны с выручкой формулой (например через юнит экономику)
- используем rice или ice фреймворки для приоретизации гипотез
- impact считаем как прирост выручки при удачной проверке гипотезы, то есть мы берем калькулятор юнит экономики и для каждой гипотезы считаем прирост выручки и на основе этого прироста ставим оценки гипотезами.  
например у нас одна гипотеза - измение кнопки увеличит конверсию на 10% и это нам даст прирост в 5% выручки  
вторая гиотеза, что если мы изменим картинку рекламного банера, то у нас появится на 5% больше пользователей и по юнит экономике это нам даст прирост в 10% выручки  
И так для каждой гипотезы. В итоге по приросту выручки ставим оценки
- далее используем формлу для расчета rice score (можно просто в impact оставить значение прироста, а в других оценках значение от 1 до 10, но в идеале прирост тоже перевести в оценку от 1 до 10)
- упорядочиваем гипотезы и выбираем самые приоритетные
