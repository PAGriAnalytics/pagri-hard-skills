# Составление резюме

Резюме формирует первое впечатление о человеке. Поэтому оно  очень важно.  
Рекрутеров учат читать резюме не больше 10 секунд.  
Поэтому нужно правильно составить резюме, чтобы не отлететь на этом этапе.  
Очень важно использовать элементы графического дизайна, чтобы привлечь внимание к важным моментам.  
И важно структурировать, чтобы можно было за 10 секунд понять, что человек может.  
Можно сказать, что резюме это пропуск на интервью.  
И тут важно оставить интригу в резюме, чтобы работадатель захотел познакомитсья поближе.  
Это значит, в резюме не нужно много писать, все должно быть кратко, но максимально интересно для работадателя.  
То есть нужно указать, что его заинтересует, и не врать конечно. Но при этом нужно понимать. что если резюме читают и  
оно длинное, или есть вода, то может сразу появиться негативное ощущение и это может сыграть против вас.  
Поэтому ценим время работадателя и при этом сообщаем все что он хочет узнать из резюме.  

Хороший совет, что добавить в резюме - это чтобы резюме максимально было похоже на вакансию.   
Нужно изучить вакансии, требования, технологии, основные ключевые слова.  
Можно собрать все чето просят в разных вакансиях и добавить к себе топ из всех.   
Тут дело в том, что рекруторы часто это скрипты и они просто мониторят то, что есть в вакансии и чем больше резюме похоже на вакансию, тем больше   
внимания к ней будет.

Резюме представляет список квалификаций и достижений специалиста, а сопроводительное письмо подчеркивает их важность для новой работы. Готовое резюме должно выглядеть формально в виде тезисов, а сопроводительное письмо лично адресовано тому, кто его прочитает.

Из каких пунктов состоит резюме  
Очень важное правило, что чем больше информации, то это не значит лучше, есть грань, за которой будет только хуже.  
Поэтому в каждый пункт помещаем только самое главное и релевантное. Но при этом количество и наполнение пунктов должно быть не малым.  
То есть нужно как бы самое главное, но и достаточно.  
1. Контактная информация  
ФИО, адрес, email для связи, номер телефона, город проживания, ссылку на гит хаб
2. Цель
В паре строчек описать почему именно вы подходите на данную должность,  
что вы можете дать компании. И эта строка может сказать больше чем все резюме.  
Эта строка может очень заинтересовать работадателя.  
3. Образование  
информация о высшем образовании, пройденных курсах (тут важно не указывать много, только самое релевантное)  
до 5-7 курсов релевантных стоит указывать, особенно когда нет опыта, это будет плюсом, и в иделае курсы на навыки, которые нужны для работы.  
Можно курс по питону на степике, курс по sql, курс по мат статистике и тер веру, лучше хорошие курсы, которые известны многим.  
4. Опыт (важно в обратно хронологическом порядке, то есть последняя должность идет первой и не нужно более 3 компаний или должностей)  
Можно так и писать - Финальный проект по курсу аналитик данных там то  
Очень важно, чтобы опыт был релевантен именно той вакансии, которую мы подаем резюме (то есть для маркетиноговой аналитике будут цнеиться свои навыки аналитики, чем для продуктовой и поэтому проекты должны быть в идеале со спецификой вакансии)  
И инструменты в проектах долны использоваться те, что указанаы в вакансии.  
Лучше всего указать релевантный опыт, если его нет, то пишем проекты, которые делал.  
В идеале ссылку на сайт с проектами.
Краткое описание проекта, какие навыки использовал. И ссылки на гитхаб и подобное.  
Лучше немного проектов, но обширных и важно, чтобы они отличались от других людей.  
Полезно вместе с проектом написать какие инструменты и для чего испольовались в этом проекте)
То есть не нужно туда помещать проекты из обучения, так как многие это делали и рекрутеры это замечают.  
информация о всех компаниях в обратном хронологическом порядке  
лучше указать не больше 3-х, чтобы не растягивать резюме, тем более мало интересно, что было до этих трех компаний,  
тем более если это было давно, лучше указать 3 последнии и указать хорошие достижения на них  
название компании, занимаемая должность, дату начала и окончания работы  
будет большим плюсом, если указать 3 наиболее значимых достижения на этой должности  
5. Навыки   
информация о всех ваших навыках, умениях, программах, которыми владеете, владение иностранными языками  
На hh.ru пишем все возможные теги с навыками, которыми владеем и они релевантны вакансии.  
Мат. статистика, теория вероятности, работа по Agile (Scrum, Kanban), знание метрик, понимание А/Б тестирования.  
Можно написать не просто SQL, а SQL (свободное владение оконными функциями и табличными выржениями)  
Приложить ссылку на репозиторий со своим кодом, если есть.  
6. Пара слов о себе и своей мотивации.  
Очень важно, когда мы меняем профессию, то написать мотивацию почему сменили сферу деятельности, почему сейчас и так далее.  
Тут сусть в том, что рекрутер, когда выдит, что челвек не после универа, работал занимался чем-то и решил стать аналитиком, то сразу куча вопросов.  
Нужно их удовлетворить. Пишем почему, зачем и пишем о своей большой мотивации. Что чем больше изучаю анализ данных, тем больше увлеаюсь и охота развиваться в этом.  И в этом блоке важно более персонализированно писать про себя.  
Важно в софты не нужно писать клише - комуникабельность, стрессоустойчивость, аналитический склад ума, умение работать в команде.  
Лучше так написать про софт скилл:   
Имею более 5 лет опыта в разметке данных и решил перейти в сферу анализа данных, так как это направление меня действительно увлекает. В процессе изучения анализа данных я убедился в своем желании развиваться в этой области. У меня есть сильная мотивация изучать все аспекты аналитики. Обладаю навыками автоматизации процессов и умею извлекать ценную информацию из данных, а также эффективно доносить её до других. Хочу заниматься тем, от чего драйвит, поэтому хочу работать аналитиком данных.

Частые ошибки, при составлении резюме
1. грамматические и пунктуационные ошибки
важно внимательно и несколько раз проверить в разные дни и проверить в онлайн сервисах правописания  
2. небрежное форматирование и использование разнообразных шрифтов  
максимально простой шрифт, контрастные цвета, лучше черный и белый, достаточно крупный шрифт
также важно использовать абзацы, отступы, чтобы легко читалось все
Вообще любое напряжение, которое создасть резюме, может привести к тому, что просто пойдут дальше,  
поэтому самое главное, чтобы ничего не раздражало в резюме.  
3. Слишком короткое или слишком длинное  
В идеале 1 страница, максимум 2, но лучше уложиться в 1 страницу.  
4. Ложь и недоговорки. 
тут важно точно указывать особенно даты начала и окончания работы,  
чтобы работадатель четко понимал сколько вы работали, если потом вскроется, что привравли, то  
будет большой минус, так как о опыте будут доп вопросы и можно спалиться  
5. устаревшие контактные данные или ошибки в этих данных  
email должен быть максимально официальным, в идеале содержать фио, без дат и других слов.  

Когда разместил резюме на сайтах типа hh.ru и других, то нужно убедиться, что ты есть на радарах,  
для этого можно поделать запросы с ключевыми словами и посмотреть есть ли ты в выдаче  


У многих компаний оценку резюме делают роботы.  
Поэтому резюме нужно писать не только для человека, но и для робота,  
так как очень много резюме сначала ищуться роботами, а человек уже из этой выборки смотрит,  
поэтому очень важно сделать так, чтобы робот тебя взял, самая частая ошибка  
это не писать ничего в опыт, если его нет, обязательно нужно что-то написать, чтобы робот тебя не отфильтровал.  


Автоподсказки на hh.ru нужно заполнять, так как это работает в поиске, так быстрее будет замечено твое резюме.  
Это также влияет на резюме, которые будут показываться hr, когда он будет смотреть резюме.  
Чтобы среди них было и твое, нужно не игнорировать подсказки.  

Для составления резюме хорошо просмотреть резюме на карьерных сайтах и выбрать из них самые удачные на ваш взгляд.  
Далее выберите из них наиболее удачные части и кастомизируется по себя.  
Но не нужно копировать, обязательно переделываем под свой стиль и опыт.

Если очень нравится вакансия, но не откликаются, то можно поискать контакты тим лида в соц сетях (то же лингедин) и попроситься в команду.  
Написать, что очень хочу у вас работать, есть такие-то проекты, хочу развиваться, кинуть ссылку на резюме и свои проекты.  
Если сейчас не готовы, то я готов в будущем к вам прийти. Также можно спросить что подтянуть в знаниях и прочее, чтобы конткретно к вам прийти.

Ещё часто есть ситуации, что сразу не берут, но говорят, что нужно прокачать и тогда возьмут.  
И если даже сами это не предложили, то всегда нужно при отказе спрашивать что нужно прокачать.  
Всегда говорить. что я сейчас прохожу продвинутые курсы, развиваюсь, и просить что нужно подтянуть, чтобы они взяли например через пару месяцев.  
И когда прокачаешь навыки, то часто берут.  
Нужно занимать максимально проактивную позицию.  

Нужно понимать, что когда висит вакансия, то откликов много и часто ситуация, что просто не просмотрели ваше резюме или просмотрели но мельком.  
Поэтому если нравится вакансия, то нужно проявить немного наголости и написать сопроводительное письмо, напомнить о себе.  

# Интервью 

ОЧЕНЬ ВАЖНО  
На собеседование нужно идит со своим мини блокнотом и ручкой.  
Чтобы можно было записать задачку и решить ее.  
Так как иначе придется просить, а это уже не очень.

ОЧЕНЬ ВАЖНО  
Перед собеседованием (не с рекрутером, а уже для проверки навыков и теории) нужно изучить специфику компании.  
То есть если это e-comerce то почитать что важно и как они работают. Прочитать про конкрерную компанию.  
Изучить какие метрики важны именно для этой компании и области в целом, как устроен бизнес, на основе чего принимается решение, что влияет на успех и прчоее.  
Это очень важно, так как ситуационные, кейсовые и бизнес вопросы скорее всего будут связаны с самой компанией и будет отлично заранее понимать контекст.

ОЧЕНЬ ВАЖНО  
Когда ты не знаешь ответ на вопрос, то рассуждай и задавай уточняющие вопросы.  
Это часто помогает сдвинуться с места и потом на удивление ты уже знаешь ответ.  
Ни в коем случае не уходить в себя и не паниковать.  
Вообще задавание уточняющих вопросов для аналитика является очень важным навыком.  

ОЧЕНЬ ВАЖНО  
Нужно готовиться к интервью с hr, так как от этого интервью зависит посоветует ли он вас или нет.  
Бывает ситуация, когда 2 человека проходили интерьвю и hr в итоге скажет нанимающему менеделжеру, что первый сильнее по хардам, но второй настолько мотивирован, что я бы советовала взять второго.  
Поэтому нужно готовиться, описать свою мотивацию, изучить компанию, сказать что хочешь заниматься в этой сфере и назвать конкретику.  

Что важно сделать до собеседования

- Поискать информацию о компании  
Даже если ваша цель — просто найти работу, а не устроиться в конкретную компанию, лучше хоть немного подготовиться уже перед первичным скринингом с эйчаром. Поищите, чем занимается компания, что выпускает, что происходило в последнее время. Эта тема обязательно всплывёт, и её можно использовать, чтобы показать свою мотивацию.
- Протестировать продукт  
Если собеседуетесь в продукт — протестируйте его несколько дней в разных сценариях и посмотрите на конкурентов. Это особенно важно для аналитика: если продукт не понравится, работать будет очень сложно.  
Интерес к продукту — тоже маркер мотивации. Например, у меня на собеседовании кандидат рассказывал, как сильно хочет работать именно в нашей мобильной игре, но так ни разу в неё и не поиграл. Звучало сомнительно.   
- Подготовить список вопросов   
Работодателю всегда приятно, когда у вас есть осмысленные вопросы о компании. Хорошо работают вопросы про команду: сколько человек, как всё устроено, будет ли у меня наставник и так далее. Такой интерес покажет, что человек приходит надолго и хочет влиться.
- Выбрать образ   
На собеседование в IT не стоит приходить в костюме, это не формат индустрии. Лучше надеть какой-то кэжуал-образ, в котором вам будет удобно, чтобы не отвлекаться на тесный пиджак. Если сомневаетесь, как принято одеваться в компании, — спросите у эйчара, это нормально.

Чего не стоит делать на собеседовании

- Уклоняться от вопросов  
Если кандидат на какой-то вопрос отвечает «не знаю» или «не хочу отвечать», то кажется, что он что-то скрывает, и это неприятный звоночек для собеседующего. Обычно работодатели хотят услышать не правильный ответ, а рассуждение. Если действительно не знаете ответа, хотя бы попытайтесь поразмышлять вслух.
- Врать о прошлой работе  
Скорее всего, после собеседования у вас попросят контакты человека с прошлой работы, у которого можно что-то прояснить, — и тогда всё тайное станет явным.
- Ругать прошлого работодателя  
Работодатель понимает, что на этом месте может оказаться он сам, и делает выводы. Кстати, это касается и плохих отзывов о собеседовании, если они необоснованны. Один кандидат не прошёл собеседование, обиделся и написал о компании гадости. Потом его не взяли на другую должность, потому что узнали об этом поступке.
- Перебивать и грубить  
На собеседовании работают все обычные правила вежливости — это касается и кандидатов, и работодателей. 
- Противоречить своим ответам
На собеседованиях задают похожие вопросы, чтобы выяснить, насколько человек определился со своими планами и ценностями. Если кандидат отвечает на них по-разному — это не очень хороший знак. Например, я спросила у кандидата, чем бы он хотел заниматься. Он ответил: продуктовой аналитикой. А на вопрос «Как вы видите трек своего развития?» ответил, что хочет развиваться в сфере machine learning. 
- Оценивать ответ за работодателя
Может показаться, что вы ужасно ответили на какой-то вопрос, и тогда запустится цепь: «сам ответил, сам оценил, завалил тут — завалю везде», и собеседование рискует окончиться провалом.   
Не стоит оценивать себя самому. Правильного ответа может вообще не быть, а если он есть — у вас мог быть правильный ход решения, и это тоже нормально. Самое важное: вас сравнивают не с эталоном, а с другими кандидатами, и они могут ответить хуже, даже если вы ошиблись.

Что сделать после собеседования  
- Запросить обратную связь, если о ней не предупредили  
Хороший тон для работодателя — предупредить, в какие сроки он передаст кандидату отзыв. В таком случае не стоит писать раньше этого времени: у работодателя много вакансий, и всё распланировано. А ещё это создаёт ощущение, что кандидат хватается за свой последний шанс, и это не играет в его пользу.  
Но если о сроках обратной связи вас не предупредили — приветствуется спросить о них самостоятельно.
- Задать вопросы
Советую делать заметки во время собеседования и задавать вопросы по ним. Например, какие будут задачи, какова ваша роль в команде, какие бенефиты и так далее. Если будет несколько компаний, то нужно будет иметь критерии, чтобы из них выбирать.

Есть большой пробел в опыте?  
Если вы не работали больше полугода, то интервьюер может спросить, почему. Мой базовый совет — не врать. Если нет «приличной» причины, то можно ответить, что были сложности какого-то рода — психологические, со здоровьем и так далее. Всегда важно сделать акцент на том, что вы сделали, чтобы такое больше не повторилось. Это поможет показать работодателю, что вы надёжный сотрудник и на вас можно положиться.

Спросили про мой главный провал?  
ак работодатель хочет понять, как вы себя ведёте в экстренных ситуациях и берёте ли на себя ответственность. Обычно на такой вопрос нет правильного ответа: у каждой компании разное понимание приемлемого.   
И лучше рассказать реальный пример из жизни, где во время провала ты не упал духом и не сдался.  
Заранее нужно вспомнить подходящую ситуацию и продумать ответ, но не врать.  

Собеседование проходит онлайн?  
Важно включить камеру и не отвлекаться на переписки и домашние дела. Позаботьтесь, чтобы у вас был опрятный фон и не было посторонних шумов — для интервьюеров это важно.

Если вы поняли, что опаздываете на собеседование, лучше не надеяться, что успеете. Напишите эйчару заранее и сообщите, на какое время вы можете опоздать, — лучше с запасом, чтобы точно в него уложиться. Хороший тон — спросить, могут ли интервьюеры подождать, или лучше перенести собеседование. 

кажется, что пришёл на допрос?  
Чтобы чувствовать себя на равных с интервьюером, важно помнить, что собеседование — это работа с двух сторон: кандидат хочет получить работу, а компания хочет нанять подходящего человека.

Самое главное это настрой.  
Настрой должен быть такой, что это ты им нужен.  
Любое собеседование воспринимаем как проверку своих знаний. Про возможность получить работу вообще не думаем.  
Как будто ты просто решил проверить свои знания.  
Также при первой же возможности, спрашивай что ты будешь делать и оценивай это.  
То есть если ты видишь, что аналитики на этом месте будет мало, то так и говори, что ты не будешь развиваться и прочее.  
Но это не должно превращаться во что-то подобное тому, как человек просто ходит по собеседованиям тренироваться.  
Должно быть ощущение, что ты готов работать у них, но прям огромного желания и волнения у тебя нет.  

Хороший лайфхак, когда тебе задают вопрос, и ты сразу не можешь ответить и нужно подумать,  
то нельзя молчать, скажи какие то вступительные слова, типа ну давайте подумаем, и дальше можно думать,  
или можно рассуждать вслух, что очень полезно, так как могут подсказать,  тут ещё плюс, что если  
ты в рассуждениях скажешь верный ответ или будешь близко, но ответишь неверно, то это будет большой плюс,  
нежели просто помолчишь и скажешь неверно.  
Вообще намного важнее как человек думает над вопросом, поэтому это будет большой плюс, если ты будешь рассуждать.

К чему нужно готовиться перед интервью
- Подготовить рассказ о своих проектах (которые указаны в опыте), цель, задачи, инструменты, результаты.
- Подготовьтесь к вопросам по ключевым словам в вакансии.
- Подготовьтесь к вопросам по мотивации (достижения, неудачи, куда хочеться дальше двигаться, что ждете от следующего места работы)
- Уточните детали вакансии, что нужно делать, как организована работа с данными, что контретно нужно будет делать
- Проявляейте проактивность, то есть задавайте уточняющие вопросы по будущей работе, по возможным проектам, задачам и показывайте максимальную заинтересенность в работе.  
- Уточните как проводится отбор и что ждут на собеседовании.
- Запросите материалы для подготовки, если они есть. Это важно, так как можно будет подготовиться лучше именно к тому, что нужно.
- Рекрутер в первую очередь проверяет ваши софт скилы и насколько вы замотивированы работать и развиваться в компании, поэтому именно на это нужно делать акцент. Проявляться интерес, показывать, что у вас горят глаза и вы хотите этим заниматься и важно развиваться дальше, если это вакансия джуна.  

Пример рассказа о себе (не нужно много говорить и углубляться в детали, нужно говорить то, что необходимо для рекрутера и нанимателя).
- Работал в компании Х в команде Х, был вовлечен в проект Х.
- Цель проекта Х.
- В мои задачи входило: 1, 2, 3.
- Удалось достичь результата Х.
- В работе чаще всего приходилось использовать инструменты Х, Х, Х, того-то и того.

Часто мотивация и горящие глаза важнее чем опыт и навыки.  
Так как нанимающей стороне важно, чтобы сотрудник развивался, а мотивация именно то что нужно.  
Поэтому если проявить искренний интерес и желание, то в итоге рекрутер посоветует нанимающему менеджеру попробовать вас.  

Бывает два основных вида интервью 
1. Когда дружелюбно начинается беседа, без проерок на стрессоустойчивость  
2. Когда при первом контакте, начинается проверка на стрессоустойчивость,  
задают неловкие вопросы, вопросы с подвохом и прочее.  
Также интервью может начинаться с тестов на ай кью или психологических тестов.  


Классическое интервью на джуна дата аналитика
- Задачи на базовую мат статистику, тер вер (мат ожидание, дисперсия, нормальное распределение, доверительный интервал, проверка гипотез, базовые критерии)
- Бизнес-кейсы (как посчитать качество, как оценить эффект)
- Метрики (бизнес, продуктовые, маркетинговые)
- Задачки с кодом на питоне, sql и тд

Есть три вида вопросов, которые скорее всего зададут  
1. Вопросы для проверки достоверности информации и вашей объективности
тут могут задать уточняющие вопросы по резюме, с попыткой проверить на ложь
относить нужно спокойно, отвечать максимально дружелюбно. 
чаще всего будет сверяться информация из резюме и ответы на похожие вопросы, поэтому не врем в резюме.  
Тут поможет совет, перед прохождением собеседования прочитать резюме.  
2. Вопросы о компетенции, навыках умениях и опыта.  
Тут также могут спросить как вы выходили из какой-нибудь ситуации в прошлом.  
Могут спросить как были получены навыки, чтобы понять сможете ли вы справиться с доложностью.  
Тут будет плюсом не просто сказать как получили навык, но и указать, что получался он практически,  
не просто изучалась теория, сделать акцент, что для вас важно именно практическое умение.  
3. Ситуационные вопросы
Будут обрисовываться ситуации и задаваться вопросы, чтобы понять как вы реагируете.

Нельзя упоминать любые причины, которые связаны с конфликтами с начальством, сотрудниками и другими.

Поэтому причины могут быть такие

    - захотел выйти на ноый уровень, а на предыдущей работе возможностей для роста не было, например, не было вызовов, все было просто.  
    - понял, что хочу заниматься другим, обучился и теперь хочу заниматься тем, что нравится  
3. Расскажите о ваших отрицательных чертах  
Тут хороший прием, взять вашу положительную сторону и возвести в абсолют, чтобы она стала недостатком.
    - часто лень составить план действий, когда задача явно не помещается в голове
    если это не сделать, то потом начинаешь осуществлять задание и понимаешь, что не знаешь что делать  
    в этот момент мозг уже устал и сложно думать о плане, поэтому очень важно составить план и прописать подзадачи, тогда все намного проще
    - переработка, потому что бывает нравится задача и не замечаешь как перегружаешься, важно уделять время отдыху, эмоциональная разгрузка спортом,  
    просто сменить деятельность, хотя би банально погулять по улице.  
    - бывает много времени трачу на перепроверку проектов, уже все сделал, но ощущение, что где-то недоделал, это не перфекционизм, это больше  
    желание не допустить ошибки.

    - люблю изучать новое, когда это можно сразу применить и использовать
    - это прозвучит странно, но нравится уставать от работы, потому что, тогда отдых намного приятнее  
    - настойчивость, не бросаю, если сразу не получается

    - люблю брать инициативу на себя

    - позитивно отношусь к критике и когда мне говорят. что что-то делаю не так, я точно задумаюсь, а если это явно ошибка, то попрошу научить делать правильно  
5. Почему вы бы хотели работать здесь?
В этом вопросе важно ответ связать с тем, что ты можешь дать работадателю
    - Уверен, что ваша компания позволит мне развиваться, смогу перенимать опыт у оытных коллег, что улучшить мою пользу для компании
    - Хочу, улучшать свои навыки, увеличивать опыт и чтобы это приносило пользу вашей компании.

    - Развить свои навыки до такого уровня, чтобы получать максимум удовольствия от работы, так как чем больше ты умеешь, тем эффективнее ты можешь решать задачи. 
    - Для меня самое главное, чтобы я приносил пользу и чтобы работа была в радость, тогда и развитие появляется само собой.  
7. Какие ваши ожидания по зарплате?
Тут важно не скзать выше, того что компания может дать, так как на этом собеседование закончиться, но при это не занизить, так как  
скорее всего ты не получишь больше чем скажешь.  Лушче ориентировать на аналогичные зарплаты в других компаниях.

8. Вы можете задать любые вопросы.
Нужно использовать эту возможнсоть. Во первых ты покажешь, что действительно заинтересован, во вторых получишь ответы на свои вопросы.  
Тут наверное самые главный вопрос, как у вас организована работа с данными? Откуда получаете данные, где храните, какие базы данных используете,  
какое хранилище, как я буду взаимодействовать с данными, как я буду получать, какие есть ограничения и прочее, что максимально опиешт ситуацию с процессами в компании.  
Могли бы описать типичный рабочий день на этой должности?
Используете ли вы такие то и такие то инструменты?
Задавай любые вопросы, которые покажут тебе, что ты будешь делать, так как может быть сюрпризом, что делать ты будешь не то. что ждешь. 
Поэтому спрашивай, какие будут твои задачи, какие дополнительные задачи, есть ли ревью кода и прочее, что позволит учиться у опытных колег.  


Также часто задают вопросы, ответы на которые ты не знаешь.  
Тут хотят проверить как ты будешь решать эту задачу.  
Тут важно показать ход сових мыслей. Умение мыслить нестандартно, логически думать и просто искать информацию.  
Самое простое это описать словами свои рассуждения, а потом можно взять телефон и загуглить. Сразу гуглить лучше не стоит.  
Также могут задаваться загадки, которые покажуться невозможными, тут главное успокоиться,  
так как часто ответ на поверхности, но немного скрыт. Нужно спокойно подумать и опять же рассуждай вслух.  
Тажке не бойся попросить листок и ручку, чтобы нарисовать или записать переменные задачи, если их много.  
Вообще эти задчи часто на то, как ты будешь искать решение и тут главное показать, что ты сразу начинашеь думать о решение, не опускаешь руки, даже если не знаешь.  
Даже если не сможешь решить, то постарайся показать максимум своих рассуждений, чтообы было понятно, что ты в любой ситуации ищещеь решение.  
Очень часто ответ на такие сложные задачи, это просто начать действовать, тут идет проверка впадешь ты в ступор или начнешь искать решение.  
Просим листок и ручку и начинаем рассуждать. И это уже будет больше половины ответа.
