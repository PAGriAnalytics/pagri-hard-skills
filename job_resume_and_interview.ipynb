# Составление резюме

Резюме формирует первое впечатление о человеке. Поэтому оно  очень важно.  
Рекрутеров учат читать резюме не больше 10 секунд.  
Поэтому нужно правильно составить резюме, чтобы не отлететь на этом этапе.  
Очень важно использовать элементы графического дизайна, чтобы привлечь внимание к важным моментам.  
И важно структурировать, чтобы можно было за 10 секунд понять, что человек может.  
Можно сказать, что резюме это пропуск на интервью.  
И тут важно оставить интригу в резюме, чтобы работадатель захотел познакомитсья поближе.  
Это значит, в резюме не нужно много писать, все должно быть кратко, но максимально интересно для работадателя.  
То есть нужно указать, что его заинтересует, и не врать конечно. Но при этом нужно понимать. что если резюме читают и  
оно длинное, или есть вода, то может сразу появиться негативное ощущение и это может сыграть против вас.  
Поэтому ценим время работадателя и при этом сообщаем все что он хочет узнать из резюме.  

Хороший совет, что добавить в резюме - это чтобы резюме максимально было похоже на вакансию.   
Нужно изучить вакансии, требования, технологии, основные ключевые слова.  
Можно собрать все чето просят в разных вакансиях и добавить к себе топ из всех.   
Тут дело в том, что рекруторы часто это скрипты и они просто мониторят то, что есть в вакансии и чем больше резюме похоже на вакансию, тем больше   
внимания к ней будет.  

Отличие сопроводительного письма от резюме  
Резюме представляет список квалификаций и достижений специалиста, а сопроводительное письмо подчеркивает их важность для новой работы. Готовое резюме должно выглядеть формально в виде тезисов, а сопроводительное письмо лично адресовано тому, кто его прочитает. 

Из каких пунктов состоит резюме  
Очень важное правило, что чем больше информации, то это не значит лучше, есть грань, за которой будет только хуже.  
Поэтому в каждый пункт помещаем только самое главное и релевантное. Но при этом количество и наполнение пунктов должно быть не малым.  
То есть нужно как бы самое главное, но и достаточно.  
1. Контактная информация  
ФИО, адрес, email для связи, номер телефона, город проживания, ссылку на гит хаб
2. Цель
В паре строчек описать почему именно вы подходите на данную должность,  
что вы можете дать компании. И эта строка может сказать больше чем все резюме.  
Эта строка может очень заинтересовать работадателя.  
3. Образование  
информация о высшем образовании, пройденных курсах (тут важно не указывать много, только самое релевантное)
4. Опыт  
Очень важно, чтобы опыт был релевантен именно той вакансии, которую мы подаем резюме (то есть для маркетиноговой аналитике будут цнеиться свои навыки аналитики, чем для продуктовой и поэтому проекты должны быть в идеале со спецификой вакансии)  
И инструменты в проектах долны использоваться те, что указанаы в вакансии.  
Лучше всего указать релевантный опыт, если его нет, то пишем проекты, которые делал.  
В идеале ссылку на сайт с проектами.
Краткое описание проекта, какие навыки использовал. И ссылки на гитхаб и подобное.  
Лучше немного проектов, но обширных и важно, чтобы они отличались от других людей.  
Полезно вместе с проектом написать какие инструменты и для чего испольовались в этом проекте)
То есть не нужно туда помещать проекты из обучения, так как многие это делали и рекрутеры это замечают.  
информация о всех компаниях в обратном хронологическом порядке  
лучше указать не больше 3-х, чтобы не растягивать резюме, тем более мало интересно, что было до этих трех компаний,  
тем более если это было давно, лучше указать 3 последнии и указать хорошие достижения на них  
название компании, занимаемая должность, дату начала и окончания работы  
будет большим плюсом, если указать 3 наиболее значимых достижения на этой должности  
5. Навыки  
информация о всех ваших навыках, умениях, программах, которыми владеете, владение иностранными языками  
Приложить ссылку на репозиторий со своим кодом, если есть.  
6. Пара слов о себе и своей мотивации.  
Хобби, интересы, увлечения   
Почему хочешь попасть в дата аналитику. Пишем, что это драйвит нас, что есть большое желание развиваться, так как нравится анализ данных.  
обязательно включить то, что связано с будущей должностью, и также можно указать те, которые будут положительно  
влиять на будущую работу, занятия спортом, но спокойным, отстутствие вредных привычек и прочее, то есть то, что укажет,  
что не будете постоянно на больничном  


Очень важно выделить ключевые требования из вакансии и чтобы в указаных проекта были использованы эти инструменты и навыки.  


Пример опыта, когда нет опыта работы, берем 2 крупных проекта и они должны быть на разные направления, чтобы в каждом были немного свои подходы, чтобы показать больше навыков (тут важно дать ссылку на проект, где максимально релевантный опыт и в самом резюме вместе со ссылкой написать что делалось и инструментарий и обязательно итоговый результат, то есть самое главное, что удалось улучшить)  
Тут нужно понимать, что собеседования ещё не было и нужно донести до рекрутера, что ты умеешь что-то делать.  
- Пишем название проекта (тут проект должен быть большим)  
- Далее кракто пишем, что использовался SQL для загрузки данных, работа выполенена на питоне, использовался пандас для изучения, предобработки данных и обогащения данных.  
С помощью beautifl soup обогащались данных. С помощью библиотеки plotly проводилась визуализация данных. Для проверки гипотез использовался аб тест. Использовалас математическая статистика для определения статистической значимости выводов и гипотез.  
Для построения дашборда использовалось Tableau.  
Были рассчитаны метрики: LTV, ARPU, CAC, ROI, APC. Проведен RFM анализ и когортный анализ.  
Были определены сегменты, с которыми нужно работать отдельно.  
И в таком же ключе пишем другие библиотеки и инструментарий, который использовался, но важно не писать лишнее. Самое главное, что находится в вакансии.  
- И в конце пишем самый главный результат, кратко, лучше циферки насколько удалось увеличить что-то или обнаружить.  
В том же аб тесте можно написать, что гипотеза подтвердилась и удалось улучшить конверсию на столько-то процентов.  

Частые ошибки, при составлении резюме
1. грамматические и пунктуационные ошибки
важно внимательно и несколько раз проверить в разные дни и проверить в онлайн сервисах правописания  
2. небрежное форматирование и использование разнообразных шрифтов  
максимально простой шрифт, контрастные цвета, лучше черный и белый, достаточно крупный шрифт
также важно использовать абзацы, отступы, чтобы легко читалось все
Вообще любое напряжение, которое создасть резюме, может привести к тому, что просто пойдут дальше,  
поэтому самое главное, чтобы ничего не раздражало в резюме.  
3. Слишком короткое или слишком длинное  
В идеале 1 страница, максимум 2, но лучше уложиться в 1 страницу.  
4. Ложь и недоговорки. 
тут важно точно указывать особенно даты начала и окончания работы,  
чтобы работадатель четко понимал сколько вы работали, если потом вскроется, что привравли, то  
будет большой минус, так как о опыте будут доп вопросы и можно спалиться  
5. устаревшие контактные данные или ошибки в этих данных  
email должен быть максимально официальным, в идеале содержать фио, без дат и других слов.  

Когда разместил резюме на сайтах типа hh.ru и других, то нужно убедиться, что ты есть на радарах,  
для этого можно поделать запросы с ключевыми словами и посмотреть есть ли ты в выдаче  


У многих компаний оценку резюме делают роботы.  
Поэтому резюме нужно писать не только для человека, но и для робота,  
так как очень много резюме сначала ищуться роботами, а человек уже из этой выборки смотрит,  
поэтому очень важно сделать так, чтобы робот тебя взял, самая частая ошибка  
это не писать ничего в опыт, если его нет, обязательно нужно что-то написать, чтобы робот тебя не отфильтровал.  


Автоподсказки на hh.ru нужно заполнять, так как это работает в поиске, так быстрее будет замечено твое резюме.  
Это также влияет на резюме, которые будут показываться hr, когда он будет смотреть резюме.  
Чтобы среди них было и твое, нужно не игнорировать подсказки.  

Для составления резюме хорошо просмотреть резюме на карьерных сайтах и выбрать из них самые удачные на ваш взгляд.  
Далее выберите из них наиболее удачные части и кастомизируется по себя.  
Но не нужно копировать, обязательно переделываем под свой стиль и опыт.

# Интервью 

ОЧЕНЬ ВАЖНО  
На собеседование нужно идит со своим мини блокнотом и ручкой.  
Чтобы можно было записать задачку и решить ее.  
Так как иначе придется просить, а это уже не очень.

ОЧЕНЬ ВАЖНО  
Перед собеседованием (не с рекрутером, а уже для проверки навыков и теории) нужно изучить специфику компании.  
То есть если это e-comerce то почитать что важно и как они работают. Прочитать про конкрерную компанию.  
Изучить какие метрики важны именно для этой компании и области в целом, как устроен бизнес, на основе чего принимается решение, что влияет на успех и прчоее.  
Это очень важно, так как ситуационные и бизнес вопросы скорее всего будут связаны с самой компанией и будет отлично заранее понимать контекст.

Самое главное это настрой.  
Настрой должен быть такой, что это ты им нужен.  
Любое собеседование воспринимаем как проверку своих знаний. Про возможность получить работу вообще не думаем.  
Как будто ты просто решил проверить свои знания.  
Также при первой же возможности, спрашивай что ты будешь делать и оценивай это.  
То есть если ты видишь, что аналитики на этом месте будет мало, то так и говори, что ты не будешь развиваться и прочее.  
Но это не должно превращаться во что-то подобное тому, как человек просто ходит по собеседованиям тренироваться.  
Должно быть ощущение, что ты готов работать у них, но прям огромного желания и волнения у тебя нет.  

Хороший лайфхак, когда тебе задают вопрос, и ты сразу не можешь ответить и нужно подумать,  
то нельзя молчать, скажи какие то вступительные слова, типа ну давайте подумаем, и дальше можно думать,  
или можно рассуждать вслух, что очень полезно, так как могут подсказать,  тут ещё плюс, что если  
ты в рассуждениях скажешь верный ответ или будешь близко, но ответишь неверно, то это будет большой плюс,  
нежели просто помолчишь и скажешь неверно.  
Вообще намного важнее как человек думает над вопросом, поэтому это будет большой плюс, если ты будешь рассуждать.  
Можно попробовать вовлечь в процесс тех, кто задает вопросы.  

К чему нужно готовиться перед интервью
- Подготовить рассказ о своих проектах (которые указаны в опыте), цель, задачи, инструменты, результаты.
- Подготовьтесь к вопросам по ключевым словам в вакансии.
- Подготовьтесь к вопросам по мотивации (достижения, неудачи, куда хочеться дальше двигаться, что ждете от следующего места работы)
- Уточните детали вакансии, что нужно делать, как организована работа с данными, что контретно нужно будет делать
- Проявляейте проактивность, то есть задавайте уточняющие вопросы по будущей работе, по возможным проектам, задачам и показывайте максимальную заинтересенность в работе.  
- Уточните как проводится отбор и что ждут на собеседовании.
- Запросите материалы для подготовки, если они есть. Это важно, так как можно будет подготовиться лучше именно к тому, что нужно.
- Рекрутер в первую очередь проверяет ваши софт скилы и насколько вы замотивированы работать и развиваться в компании, поэтому именно на это нужно делать акцент. Проявляться интерес, показывать, что у вас горят глаза и вы хотите этим заниматься и важно развиваться дальше, если это вакансия джуна.  

Пример рассказа о себе (не нужно много говорить и углубляться в детали, нужно говорить то, что необходимо для рекрутера и нанимателя).
- Работал в компании Х в команде Х, был вовлечен в проект Х.
- Цель проекта Х.
- В мои задачи входило: 1, 2, 3.
- Удалось достичь результата Х.
- В работе чаще всего приходилось использовать инструменты Х, Х, Х, того-то и того.

Часто мотивация и горящие глаза важнее чем опыт и навыки.  
Так как нанимающей стороне важно, чтобы сотрудник развивался, а мотивация именно то что нужно.  
Поэтому если проявить искренний интерес и желание, то в итоге рекрутер посоветует нанимающему менеджеру попробовать вас.  

Бывает два основных вида интервью 
1. Когда дружелюбно начинается беседа, без проерок на стрессоустойчивость  
2. Когда при первом контакте, начинается проверка на стрессоустойчивость,  
задают неловкие вопросы, вопросы с подвохом и прочее.  
Также интервью может начинаться с тестов на ай кью или психологических тестов.  


Классическое интервью на джуна дата аналитика
- Задачи на базовую мат статистику, тер вер (мат ожидание, дисперсия, нормальное распределение, доверительный интервал, проверка гипотез, базовые критерии)
- Бизнес-кейсы (как посчитать качество, как оценить эффект)
- Метрики (бизнес, продуктовые, маркетинговые)
- Задачки с кодом на питоне, sql и тд

Есть три вида вопросов, которые скорее всего зададут  
1. Вопросы для проверки достоверности информации и вашей объективности
тут могут задать уточняющие вопросы по резюме, с попыткой проверить на ложь
относить нужно спокойно, отвечать максимально дружелюбно. 
чаще всего будет сверяться информация из резюме и ответы на похожие вопросы, поэтому не врем в резюме.  
Тут поможет совет, перед прохождением собеседования прочитать резюме.  
2. Вопросы о компетенции, навыках умениях и опыта.  
Тут также могут спросить как вы выходили из какой-нибудь ситуации в прошлом.  
Могут спросить как были получены навыки, чтобы понять сможете ли вы справиться с доложностью.  
Тут будет плюсом не просто сказать как получили навык, но и указать, что получался он практически,  
не просто изучалась теория, сделать акцент, что для вас важно именно практическое умение.  
3. Ситуационные вопросы
Будут обрисовываться ситуации и задаваться вопросы, чтобы понять как вы реагируете.  


Самые распространенные вопросы и правильные ответы на них
1. Как бы вы себя описали?
Тут важно, чтобы это было правда. Заученные фразы сразу заметны и не интересны.  
Поэтому с ответом приводим примеры, когда это было с вами, немного деталей
Способный быстро приспосабливаться к обстоятельствам  
Трудолюбивый  
Комуникабельный  
Надежный  
Честный  
Организованный  
Целеустремленный  
умею работать в команде  
быстро обучающийся  
2. Расскажите, почему вы ушли с предыдущего места работы?
Нельзя упоминать любые причины, которые связаны с конфликтами с начальством, сотрудниками и другими.  
Любые ответы, которые ты скажешь, будут трасполироваться на новую работу и сразу же возникнут мысли, что тут может быть также.  
Поэтому причины могут быть такие
    - сократили, так как уменьшилось финансирование  
    - захотел выйти на ноый уровень, а на предыдущей работе возможностей для роста не было, например, не было вызовов, все было просто.  
    - понял, что хочу заниматься другим, обучился и теперь хочу заниматься тем, что нравится  
3. Расскажите о ваших отрицательных чертах  
Тут хороший прием, взять вашу положительную сторону и возвести в абсолют, чтобы она стала недостатком.
    - часто лень составить план действий, когда задача явно не помещается в голове
    если это не сделать, то потом начинаешь осуществлять задание и понимаешь, что не знаешь что делать  
    в этот момент мозг уже устал и сложно думать о плане, поэтому очень важно составить план и прописать подзадачи, тогда все намного проще
    - переработка, потому что бывает нравится задача и не замечаешь как перегружаешься, важно уделять время отдыху, эмоциональная разгрузка спортом,  
    просто сменить деятельность, хотя би банально погулять по улице.  
    - бывает много времени трачу на перепроверку проектов, уже все сделал, но ощущение, что где-то недоделал, это не перфекционизм, это больше  
    желание не допустить ошибки.  
4. Расскажите о ваших положительных чертах
    - люблю изучать новое, когда это можно сразу применить и использовать
    - это прозвучит странно, но нравится уставать от работы, потому что, тогда отдых намного приятнее
