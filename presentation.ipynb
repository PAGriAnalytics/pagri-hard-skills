Презентация - это любая ситуация, где нужно объяснить, убедить и вдохновить на действие. 


Самое главное отличать презентацию от презентационного буклета.  
Одно дело, когда человек будет без вас знакомиться с вашей презентацией, это будет буклет и там полный текст.  
Другое дело ваша презентация словами. Тут минимум тектса на слайдах и больше речи.  
Важно, количество слайдов должно соответствовать времени выступления, нужно четко понимать сколько есть времени,  
и с запасом готовить текст и слайды.  
Основной принцип презентации - на слайде краткий заголовок, возможно формула или картинка, но минимум, а словами  
раскрывается смысл заголовка. Заголовок должен быть сочным, понятным, в идеале захватывающим и интересным.  
На слайде должны быть заголовки, главные выводы, результаты и то что словами будет плохо понятно.  Остальное говориться словами.  
В конце подводим итоги. Делаем выводы. И не только выводы, а дополнительно расскажите что вам дал этот проект, что нового узнали.  
Также можно расскзать интересные факты, которые были в процессе создания проекта. Это интерсено и оставляет хорошее впечетление.  Но это в конце после выводов.  

Частые обишки в презентациях
- слишком много информации на слайдах (не нужно писать весь текст на слайд и потом читать с экрана)
информация на слайде должна быть краткой и самое главное интересной. Лушче чтобы создавалась интрига того,  
что вы будете говорить. И шрифт должен быть крупный, чтобы его легко можно было прочитать издалека. 
Самый простой способ проверить размер шрифта и картинок, включить презентацию, отойти на два метра и посмотреть,  
все ли можно прочитать.  
- отсутствует структура повествования и содержание
должен быть четкий план презентации, он должен быть последовательный, должна быть структура, одно вытекать из другого и приводить к результату.  
начинать лучше с расказа о том что будет, например, 
Всем привет, сегодня я хочу рассказать про дипломный проект, структура моего выступления будет строиться на следующих разделах, первый блок немного расскажу про себя,  
второй блок я расскажу о целях и задачах проекта, и далее уже основная часть презентации, которая будет разбита на 2-3 части такие то и такие.  
Это важно, сначала представился, чтобы люди поняли кто ты, кратко, имя, можно откуда, где работаешь, не просто кто-то говорит, а уже знакомый типа.  
Далее структура или содеражание нужно, чтобы слушатели поняли, что будет далее, чтобы заинтересовались. В идеале можно создать интригу, что будет то то и то то интересное, по ходу презентации вы поймете о чем речь.  Также это важно, чтобы люди понимали в каком месте презентации мы находимся и что ждать далее.  
- много булетов и списков
Минимум тезисов на слайде, только главные идеи, остальное говориться словами.  

Дополнительные советы  
- Проверить, чтобы шрифты были системные, распространенные. Так как может быть проблема. что ваш крутой красивый шрифт просто не установлен в системе.  
можно ещё экспортировать в pdf, тогда вид будет одинаков в любой системе.  
- Используйте контрастные цвета. Не используйте серую и бледную палитру.  
презентация часто это проектор, поэтому цвета могут тускнеть и прочее. Поэтому изначально доложно быть максимум контраста.  
- В идеале все данные на слайдах и доп материалы должны иметь доступ не из интернета.  Интернет может быть меделенный, может отвалиться и прочее.  
лучше все заранее загрузить и организовать презентацию и рассказ так, чтобы не зависеть от интернета.  
- Если до вас уже обозначили проблему, с которой вы вышли, то лучше сократить рассказ о проблеме, это уже слышали и будет даже большой плюс,  
слушающие это заметят, скажи, что проблему уже озвучили, кратко в паре слов повтори, но кратко и к сути. 
- картинка на слайде должна быть понятно и человек, который на нее посмотрел, должен сразу понять что на ней и не думать о ней, пока вы далеьше говорите.  
- графики максимально простые, 1 - 2 лини максимум. Если бары, то легко различимые и понятные. Если нужно показать разницу, то она должна легко читаться.  
Один столбик явно больше другого, иначе плохо. Все контрастно. Все выводы однозначно читаются с картинки и графика. Сразу понятно и доступно.  
- взаимодействуйте с аудиторией. задавай вопросы, причем ответ может быть не важен (риторические вопросы), важен сам факт контакта, тут отлично подойду вопросы шутки и прочее.
- узнать какое будет помещение, кто будет выступать до вас,  кто будет слушать, это поможет понять что интересно этим людям, понять какая будет обстановка. 
- вообще задача спикера, донести до аудитории, то что он знает, а аудитория нет, а для эффективного донесения информации нужно знать кто тебя будет слушать.  
- мозг лучше усваивает картинку, поэтому, если можно визуализировать мысль, то нужно это делать.  


План презентации  
- Сначала постановка цели, текст презентации, потом выбрасываение лишнего если есть. и только потом делаем слайды.  
это частая ошибка, сразу делать слайды. Слайды это уже следствие текста и будущей речи.  
- Постановка цели — важнейший этап в подготовке любой презентации. Надо ответить на вопрос: зачем я делаю презентацию? У меня цель продать свои услуги? Убедить людей в собственной квалификации? Или я просто хочу вызвать у публики интерес к какой-то теме? Правильная цель всегда лежит на стыке интересов выступающего и его аудитории. Какой результат получу я после презентации, что получат от моего выступления слушатели? Не всегда конечный результат — это продажа. Например, стартапы часто делают презентации, чтобы заинтересовать инвесторов и перейти на следующий раунд переговоров. Есть, например, отчетные презентации, где главная задача — рассказать о результате деятельности проекта или компании и вдохновить стейкхолдеров и клиентов на продолжение работы.
У презентации есть три цели: впечатлить, убедить и вдохновить.
И важно задать себе вопросы - зачем эта презентация мне и зачем это слушателям. Ответы на эти вопросы помогут сформировать более интересный рассказ.  
- Прежде чем думать как сделать что-то, ответь на вопрос зачем, это не только в презентации.  
Если мы понимаем зачем это нам и другим, то нам будет намного проще думать как это сделать.  Как вытекает из зачем.  
А если мы сразу начинаем думать как, то мы будем путаться и не знать что делать, конечно цели то нет и плана тоже.  
- Наверное самое главное, это рассказывать историю. Постройте рассказ изначально как историю, сделайте ее интересной. 
- представиться, рассказать чем занимаешься и какая будет структура выступления 1-2 минуты. Не пытайся быть идеальным и безошибочным. Наоборот, расскажи какие были  
проблемы, что не получалось, кратко, но это действительно добавляет эмоциональности в презентацию, люди начинают сочувствовать и станут более расположены.  
И правило 30 секнд. Самые первые 30 секунд созадают впечателние о человеке.  
нужно представиться и сказать что-то интересное о себе или происходящем. В идеале быть нестандартным, выделиться от других.  
вот представь выходит человек, ты его не знаешь и он просто здоровается, говорит фио и погнал,  
а с другой стороны он пару минут говорит о себе кратко, но какие-то личные вещи, это распологает, сразу создается контакт.  
- Также важно рассказать о важности темы после расскза о структуре и о себе.  Постарайтесь максимально объяснить важность темы,  
в идеале показать, что это будет полезно слушающим.  
- весь рассказ не более 10 минут, все что более становится очень скучно. Лучше коротко до 10 минут, но интересно.  
когда 10 и более минут, то охота в какой-то момент сказать, ну что у тебя интересного, хватит столько речи.  
В идеале выкинуть лишние предложения, которые не важны. Оставить самое главное, что интересно.  
Самое плохо, когда текста на 5 минут и начинают разбавлять водой. Вот это плохо. Пусть будет 5 минут, но интересно.  
Принчип простой, меньше, да интереснее, мало кому интересно слушать детали того, как ты это получил и прочее, люди хотят кратко  
погрузиться в процесс получения результата и увидеть результат.  
В идеале 7-8 минут, интересные картинки, шутки, коротко о себе и структура выступления 1- 2 минуты, шаги как получался результат, и результаты минуты 3-4.  
- речь должна быть живой. вспомни сам, какое ощцщение, когда человек читает по бумажке или со слайда. это нудно. лучше что-то забыть и прсото  
сказать о чем-то другом или просто перепрыгнуть этот кусок, чем читать по бумажке. Вообще скорее всего даже не заметят, что ты что-то забыл сказать.  
- пара шуточек в процессе рассказа это круто. и в конце можно расскзать что-то интересное о результатах.  
- не нужно пытаться говорить как-то сложно, стараться быть диктором и прочее. Лучше свободно, на обычном простом языке.  
представь, что расскзавывешь другу или знакомому, не нужно официоза.  
- очень важно преследовать одну идею на протяжении всего выступления, это должна быть красная линия презентации.  
должно быть четко понятно зачем ты выступаешь, что хочешь сказать. В начале ставим проблему задачу, далее расскзаз кратко, как эта проблема решалась,  
и в конце выводы. Каждый слайд не отходит от задачи, немного в начале при рассказе о себе и в конце можно немного про интересные выводы отвлечься от задачи,  
а в остальном красной линией цель проходит через весь рассказ.  
- также каждый слайд преследует одну свою мысль. Одна презентация - одна цель. Один слайд - одна цель, которая ведет к общей цели.  
плохо, когда одна мысль разноситься между слайдами, может потеряться суть и слушатели запутаются.  
- KYSS-принцип — Keep Your Slide Simple. Не тратьте время на мишуру вроде оформления или сложного дизайна презентации. Это не конкурс художников. 

Структура слайда с графиком или картинками. 
Сверху слева заголовок
Далее слайде делится на две части, 66 на 34 примерно, слева 60 прцоентов для контент, схемы, картинки,  
а справа 34 (треть примерно) это мысль краткое пояснение  
Например, слева график и справа пишем словами выводы или какие-то показатели, например, график выручки, а справа от крафика итогове LTV и другие показатели.  
Это правильная последовательнсоть. Сначала читается заголовок (он важен и нужен, так как он несет в себе основную мысль слайда), далее человек спускается вниз глазами,  
смотрит на график или схему, и потом уже смотрит направо на показатели.  Также и в даш борде, лучше слева графики и правее него индикаторы и счетчики.  



Как держаться на выступлении
- Вначале нужно помнить, что изначально аудитория за вас. Вспомни, когда ты кого-то слушаешь, ты же не хочешь. чтобы он провалился. Вообще люди тратят свое время,  
и они думают, что это будет хорошо. Если человек начал слушать, то он уже за тебя. Это очень важно понимать.  
- будь максимально естественен. Если волнуешься сильно, то самое лучшее это сказать, ребят я так сейчас волнуюсь, это даже вызов для меня справится с этим волнением. Посмотрим как у меня это получится. И это реально даст буст.  
Хуже будет, если начинать пытаться скрыть волнение или вести себя неестественно. Если тебе важно донести какую-то идею до людей, то можно прям так и сказать, чем  
пытаться донести ее неестественно. Так и скажи, что мне важно донести до вас вот такую то идею, я волнусь по этому поводу. Но мы справимся.  
- Жестикуляция помогает донести мысль. Она должна быть спокойно и согласовывться с речью. В идеале жести показывают слова (образы).  
- Посомтри кто тебя слушает, постарайся понять кто они, эти люди прожили большую жизнь и сейчас тебя слушают. Прояви искренний интерес к людям, и они это заметят.  
- Не бойся говорить, что не получалось, говорить о провалах и неудачах. Люди охотнее симпатизируют тем, кто не идеальный.  И вообще когда человек говорит, что у него не   получалось, но он справился. то это вызывает доверие.
