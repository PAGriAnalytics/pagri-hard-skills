Как выбрать формат отчета
- Кто ЦА отчета? (для кого отчет?)
- Какая перед нами стоит задача?
- Сколько у нас времени?
- Как нужно будет выполнить аналогичную задачу?

Чтобы определиться, какую информацию добавить, а какую — отсечь, руководствуйтесь правилом: чем пользователь презентации дальше от анализа данных, тем проще она должна быть. Не погружайте руководителя компании в детали вашей борьбы с пропусками. Хотя он и обрадуется, что вы такой профессионал, не отнимайте его время. Пусть концентрируется на важном.

Прежде чем начинать делать презентацию, нужно узнать для каких целей она будет использоваться.  
Если менеджер просит предоставить ему данные по результатам какой-то акции, то нужно нужно спросить для каких целей он эти данные будет использовать.  
И далее определить какие метрики ему лучше подойдут и в каком формате ему предоставить результат.  
Таким образом мы избавим себя от возможных проблем с переделыванием задачи.

Структура презентации
- На вводном слайде укажите название и период исследования.
- Далее на следующем слайде указываем общие выводы по исследованию (не более 7 выводов на лист, максимум 10, если больше, то разбиваем на 2 слайда и если на 2 не входит, то убираем неважные).
- Далее Один слайд — одна мысль.
- Выводы идут по группам, и внутри групп они упорядочены по важности. Не нужно прыгать сначала от одной метрике, потом к другой и потом к разрезу первой. Идем последовательно, взяли сущность рассказали о ней от главного к деталям и далее пошли к следующей.  
- В заголовке слайда — ключевой вывод (но кратко без деталей, детали ниже булитами возле графика справа).  
Укажите в заголовке вывод, который укажет, на что нужно обратить внимание и ответ на какой вопрос искать на слайде  
- На слайде с графиком пишем выводы, если это презентация не для выступления. Важно кратко и понятно. Так как это слайд, а не отчет.   
Тут можно использовать такую схему: если вывод один, то пишем прям в заголовке и тогда выводов на слайде не нужно, а если выводов несколько, то слайд называем кратко и понятно (например, Цены на квартиры в Москве), и сбоку графика уже пишем выводы булитами
- Поддерживайте единую историю  
Решайте задачи исследования постепенно. Завершайте каждый блок слайдов так, чтобы пользователи презентации делали нужные выводы.
- По возможности под графиком указывайте источники данных. Например - Внутренние данные Яндекс Недвижимость, или прям ссылку на источник.
- В конце презентации — рекомендации.  
Необязательно подробные, но отражающие суть.
- На слайдах должен быть индикатор движения, это может быть просто справа внизу набор точек, которые по меере движения окрашиваюся, либо индикатор в любом месте, который показывает сколько прошли, а сколько осталсь.  
- Единый стиль
- Меньше цветов
- Больше визуализаций (стараться максимум числовых данных представлять в виде графиков или минивизуализаций)
- Тексты короче (чем больше текста, тем меньше будут слушать вас, а больше читать и когда прочитают, то будут ждать следующий слайд)
- Гиперссылки
- Давайте только нужную информацию
- Делайте выводы под каждым инсайтом
- Заканчивайте рекомендациями или призывом к действию
- Просите фидбек (просим колег и тимлида проверить и сказать что улучшить)

Дерево гипотез и принцип MECE  
- mutualy exclusive (взаимоисключающие)
- collectively exhaustive (совместно исчерпывающие)

Суть  в том, что каждый следующий этап дерева гипотез является взаимно исключающим и совмсетно исчерпывающим.  
Благодаря этому нам не нужно будет возвращаться по дереву в обратную сторону.

Используем принцип пирамиды Минто  
Это способ подачи информациии в строгой иерархии по принципу пирамиды.  
Где на вершине главная идея, а в основании - уточняющие детали.  
Ключевая идея - сосредоточится на основной идее, чтобы аудитории было легче воспринимать информацию.  
Информация, представленная впоследствии, разветвляется на конкретные детали в четкой и содержательной форме.  
Золотые правила Минто
- Начните с ответа, тезиса или ключевого момента
- Группируйте аргументы и суммируйте основные моменты (на каждом уровне пирамиды обобщаются выводы и идеи уровня ниже)  
(идеи в каждой группе должны быть взаимосвязаны и одного вида)
- Представляейте вспомогательную информацию и идеи логически 

Когда мы сделали презентацию или отчет, то нужно себя проверить по золотым правилам Минто.

Название слайда содержит основной вывод без детализации, если она изначально есть.  
А детализация пишеться булитами рядом с графиком.  
Напримре, если вывод типа в таком то регионе выручка больше на 30%, чем в таком то.  
То в заголовке пишем - В таком-то регионе выручка больше.  
А уже рядом с графиком детализируем - Типа в этом регионе выручка настолько больше и так далее.  

### Презентация для продукт менеджера 

Это презентация, где может быть много текста. Не путать с презентацией для выступления.  
Тут задача презентовать результат своей работы. Причем это не отчет, где есть промежуточные шаги.  

Если мы делаем отчет с результатами или презентацию для продукт менеджера, то тут будут только результаты,    
но важно, чтобы там было следующее

Если мы пишем названия компаний, которые не измвестны всем, или термины, которые не очевидны, то обязательно пишем про них кратко в русле, выдержек из википедии.  
Так как человеку нужно понимать что это, кто это. И чтобы ему не нужно было гуглить, пиешм инфу сами.  

Лушче использовать шаблон, чтобы презентация заиграла красками,  
но не сильно броские шаблоны, просто чтобы не было белого листа и текста.  

В презентации обязательно должно быть 
- имя
- дата
- источники с указанием точных координат
- хотя бы краткое содержание в начале
- рекомендации
- спасибо за внимание

1. Методология сбора, глоссарий (метрики, формулы), вводные данные: период,
количество собранных сущностей, количество/названия сайтов.
Рассказываем что собирали, как собирали, что считали, как считали, даже если мы думаем, что он должен знать расчеты метрик, лучше написать как считали.  
Откуда собирали данные. Если мы собирали на многих форумах и их все не указать, то хотя бы пишем темы форумов (форумы мамочек, туристов, автомобилистов и прочее)  
Обязательно не просто написать словами откуда, а лучше приложить точные координаты (ссылки и прочее), чтобы можно было самим скачать и убедиться.   
И тут важно, чтобы источник был более менее авторитетный. Стараемся брать данные из таких источников.  
Также обязательно указать дату. Иначе не понятна актуальность исследования.
2. Выводы и рекомендации.
3. Инфополе: поводы, тренды, сезонность, ключевые темы, популярные бренды.  
Задача ввести его в ситуацию. Чтобы он понимал контекст. Погрузился в инфополе, связанное с работой.  
То есть мы не просто выводим график сколько у нас таких, например, упоминаний, а столько таких.  
А пишем что это упоминание вот про это, например, у нас упомнинание светлая кожа, будет полезно рящдом с графиком на слайде вывести  
почему упоменается светлая кожа, взять из той же википедии какие проблемы связаны с этим и нашей темой.  
То есть тут суть в том, что нужно предоставить больше информации по неочевидным моментам. Ты же не будешь ему рассказывать.  
Он будет читать и чтобы он точно понимал связь светлой кожи и выводами и графиком, важно добавлять информацию об этом.  
Также у нас сравнение брэндов, можно в кратце написать про них, если они не всем известные.  
4. Аудитория: написать кто пользуется тем о чем работа, демография пользователей.  
То есть указать какой процент мужчин например, женщин, другие сегментирования пользователей. То есть детализировать кто именно пользуется и сколько их в каждой группе  
5. Пользовательский опыт  
Как именно пользуются, в контексте чего пользуются, с чем используют, когда используют и прочее.  


ВАЖНО  
Хорошее решение делать заголовок как вывод.  
То есть мы проводили исследования. Построили графики. Теперь нам нужно сделать презентацию.  
Мы берем основные выводы и график (или несколько) к ним и помещаем на одни слайд  
Вверху крупно один вывод и ниже графики, таблицы. Сразу понятно о чем слайд и вывод на слайде.    
И на предпоследнем слайде регомендации.

### Презентация для выступления

Презентация - это любая ситуация, где нужно объяснить, убедить и вдохновить на действие. 


Самое главное отличать презентацию от презентационного буклета.  
Одно дело, когда человек будет без вас знакомиться с вашей презентацией, это будет буклет и там полный текст.  
Другое дело ваша презентация словами. Тут минимум тектса на слайдах и больше речи.  
Важно, количество слайдов должно соответствовать времени выступления, нужно четко понимать сколько есть времени,  
и с запасом готовить текст и слайды.  
Основной принцип презентации - на слайде краткий заголовок, возможно формула или картинка, но минимум, а словами  
раскрывается смысл заголовка. Заголовок должен быть сочным, понятным, в идеале захватывающим и интересным.  
На слайде должны быть заголовки, главные выводы, результаты и то что словами будет плохо понятно.  Остальное говориться словами.  
В конце подводим итоги. Делаем выводы. И не только выводы, а дополнительно расскажите что вам дал этот проект, что нового узнали.  
Также можно расскзать интересные факты, которые были в процессе создания проекта. Это интерсено и оставляет хорошее впечетление.  Но это в конце после выводов.  
Очень удобно делать слайды, на которых показано в каком месте презентации мы находимся.  
То есть просто слайд, на котором название выступления и индикатор  
продвижения. Например можжно сделть несколько кружков и в одном из них  
цифра и эта цифар будет двигаться по кружкам. То есть есть 5 гружков  
в начале в певом кружке 1, остальные пустые, потом во втором кружке 2, остальные пустые и так далее. И 5 раз за отчет будут поялвяться такие слайды. Это даст зрителю понимать где мы находимся.

Частые обишки в презентациях
- слишком много информации на слайдах (не нужно писать весь текст на слайд и потом читать с экрана)
Информация на слайде это опорные слова и предложения, чтобы не забыть сказать важное  
информация на слайде должна быть краткой и самое главное интересной. Лушче чтобы создавалась интрига того,  
что вы будете говорить. И шрифт должен быть крупный, чтобы его легко можно было прочитать издалека. 
Самый простой способ проверить размер шрифта и картинок, включить презентацию, отойти на два метра и посмотреть,  
все ли можно прочитать.  
- отсутствует структура повествования и содержание
должен быть четкий план презентации, он должен быть последовательный, должна быть структура, одно вытекать из другого и приводить к результату.  
начинать лучше с расказа о том что будет, например, 
Всем привет, сегодня я хочу рассказать про дипломный проект, структура моего выступления будет строиться на следующих разделах, первый блок немного расскажу про себя,  
второй блок я расскажу о целях и задачах проекта, и далее уже основная часть презентации, которая будет разбита на 2-3 части такие то и такие.  
Это важно, сначала представился, чтобы люди поняли кто ты, кратко, имя, можно откуда, где работаешь, не просто кто-то говорит, а уже знакомый типа.  
Далее структура или содеражание нужно, чтобы слушатели поняли, что будет далее, чтобы заинтересовались. В идеале можно создать интригу, что будет то то и то то интересное, по ходу презентации вы поймете о чем речь.  Также это важно, чтобы люди понимали в каком месте презентации мы находимся и что ждать далее.  
- много булетов и списков
Минимум тезисов на слайде, только главные идеи, остальное говориться словами.  
- размытые графики  
Используем формат `svg`

Дополнительные советы  
- указать кто делал и когда делал на первом слайде с темой презентации  
Это очень важно, так как понятно кто делал и когда, то есть мы можем прикинуть актуальность работы, и знакомимся.  
- одна мысль - один слайд
Тут суть  в том, что читатель не должен обгонять выступающего. А если на слайде написано больше одной мысли, то слушатели  
прочитают вторую мысль и уже будут впереди и вообще могут не слушать первую мысль даже.  
- не больше 7-10 слов в строке на слайде
- не более 7-10 строк на слайде
- Если для одного заголовка слайда получилось больше строк, то разбиваем слайд на несколько и всем даем один заголовок,  
то есть будет несколько слайдов с одним заголовком.  
- Хорошее решение делать заголовок как вывод.  
То есть мы проводили исследования. Построили графики. Теперь нам нужно сделать презентацию.  
Мы берем основные выводы и график (или несколько) к ним и помещаем на одни слайд  
Вверху крупно один вывод и ниже графики, таблицы. Сразу понятно о чем слайд и вывод на слайде.  
- На слайде не должно быть 80 процентов пустого места  
Широкие отступы могут быть, но пустых мест быть не должно.  
- Если графики или таблицы из разных программны пакетов, то нужно сделать их хотя бы все в одной цветовой гамме, чтобы не бросалось в глаза.  
- Межстрочный отступ не менее 1.5 иногда даже больше и иногда даже стоит увеличить кернинг (межбуквенное растояние), чтобы на растоянии было видно
- Правило расстояинй  
Пробел по ширине больше, чем расстояние межбуквенное  
Межстрочный интервал должен быть шиче, чем пробел
Между абзацами или блоками - шире, чем межстрочный и так далее  
Это относится и к подписям и к графикам
- Проверить, чтобы шрифты были системные, распространенные. Так как может быть проблема. что ваш крутой красивый шрифт просто не установлен в системе.  
можно ещё экспортировать в pdf, тогда вид будет одинаков в любой системе.  
Шрифт должен быть без засечек, так как он лучше читается с далекого растояния.  
Хорошие шрифты - гельветика (Helvetica), san francisco display (для mac)
times new roman и arial не нужно использовать, так как они уже затасканы и надоели всем
- Используйте контрастные цвета. Не используйте серую и бледную палитру.  
презентация часто это проектор, поэтому цвета могут тускнеть и прочее. Поэтому изначально доложно быть максимум контраста.  
- В идеале все данные на слайдах и доп материалы должны иметь доступ не из интернета.  Интернет может быть меделенный, может отвалиться и прочее.  
лучше все заранее загрузить и организовать презентацию и рассказ так, чтобы не зависеть от интернета.  
- Если до вас уже обозначили проблему, с которой вы вышли, то лучше сократить рассказ о проблеме, это уже слышали и будет даже большой плюс,  
слушающие это заметят, скажи, что проблему уже озвучили, кратко в паре слов повтори, но кратко и к сути. 
- картинка на слайде должна быть понятно и человек, который на нее посмотрел, должен сразу понять что на ней и не думать о ней, пока вы далеьше говорите.  
- графики максимально простые, 1 - 2 лини максимум. Если бары, то легко различимые и понятные. Если нужно показать разницу, то она должна легко читаться.  
Один столбик явно больше другого, иначе плохо. Все контрастно. Все выводы однозначно читаются с картинки и графика. Сразу понятно и доступно.  
- взаимодействуйте с аудиторией. задавай вопросы, причем ответ может быть не важен (риторические вопросы), важен сам факт контакта, тут отлично подойду вопросы шутки и прочее.
- узнать какое будет помещение, кто будет выступать до вас,  кто будет слушать, это поможет понять что интересно этим людям, понять какая будет обстановка. 
- вообще задача спикера, донести до аудитории, то что он знает, а аудитория нет, а для эффективного донесения информации нужно знать кто тебя будет слушать.  
- мозг лучше усваивает картинку, поэтому, если можно визуализировать мысль, то нужно это делать.  
- никаких анимационных переходов от слайда к слайду
- никаких анимационных текстов в виде всплывающих кусков, появляющихся плавно блоков и прочего
- разрешение картинок должно быть большим (dpi высокий или формат svg)
Так как на экране ноута это одно, а когда слайд на всю стену, то будет другой вид
- никаких красивых фонов
- никаких стоковых картинок  
Это просто информационный мусор, который вообще не нужен, хуже того они ещё и отвлекают зрителя  
Вообще все. что есть на слайде должно подкреплять мысль, которую мы говорим
- Лушче использовать шаблон, чтобы презентация заиграла красками,  
но не сильно броские шаблоны, просто чтобы не было белого листа и текста.  
- Последний слайд 'Спасибо за внимание'
- На каждом слайде сделать какую-то визуализацию оглавления, чтобы было понятно где мы находимся  
Мы в начале показывем краткое оглавление работы и потом на каждом слайде в виде прозрачных булитов где-нибудь сбоку, чтобы не мешало обзору или снизу,  
делаем как бы каркас оглавления. Без слов, можно просто типа булиты и текущий булит выделен ярко и понятно  где мы сейчас. Это крайне удобно.  
- На графиках и в названиях слайдов, да и вообще везде стараемся использовать короткие названия. Не сокращаем. Это важно.  
Нужно придумать например одно словао вместо двух.  
Можно использовать интернет. Пишем синоним того-го и пишем фразу и выбираем короткое понятное название.  
Это особенно важно, когда у нас графики и названия слайдов. Большие названия перегружают слайд.  
- не нужно в презентации (даже в презентации буклете) писать типа из графиков видно, из графиков можно сделать вывод.  
Это все лишнее. Сразу пишем что видим в виде вывода. И стараемся сформулировать кракто, но понятно.  


План презентации  
- Первый лист стандартно - тема большими буквами и ниже дата. Также по ситуации автор.  
- Сначала постановка цели, текст презентации, потом выбрасываение лишнего если есть. и только потом делаем слайды.  
это частая ошибка, сразу делать слайды. Слайды это уже следствие текста и будущей речи.  
- Постановка цели — важнейший этап в подготовке любой презентации. Надо ответить на вопрос: зачем я делаю презентацию? У меня цель продать свои услуги? Убедить людей в собственной квалификации? Или я просто хочу вызвать у публики интерес к какой-то теме? Правильная цель всегда лежит на стыке интересов выступающего и его аудитории. Какой результат получу я после презентации, что получат от моего выступления слушатели? Не всегда конечный результат — это продажа. Например, стартапы часто делают презентации, чтобы заинтересовать инвесторов и перейти на следующий раунд переговоров. Есть, например, отчетные презентации, где главная задача — рассказать о результате деятельности проекта или компании и вдохновить стейкхолдеров и клиентов на продолжение работы.
У презентации есть три цели: впечатлить, убедить и вдохновить.
И важно задать себе вопросы - зачем эта презентация мне и зачем это слушателям. Ответы на эти вопросы помогут сформировать более интересный рассказ.  
- Прежде чем думать как сделать что-то, ответь на вопрос зачем, это не только в презентации.  
Если мы понимаем зачем это нам и другим, то нам будет намного проще думать как это сделать.  Как вытекает из зачем.  
А если мы сразу начинаем думать как, то мы будем путаться и не знать что делать, конечно цели то нет и плана тоже.  
- Наверное самое главное, это рассказывать историю. Постройте рассказ изначально как историю, сделайте ее интересной. 
- представиться, рассказать чем занимаешься и какая будет структура выступления 1-2 минуты. Не пытайся быть идеальным и безошибочным. Наоборот, расскажи какие были  
проблемы, что не получалось, кратко, но это действительно добавляет эмоциональности в презентацию, люди начинают сочувствовать и станут более расположены.  
И правило 30 секнд. Самые первые 30 секунд созадают впечателние о человеке.  
нужно представиться и сказать что-то интересное о себе или происходящем. В идеале быть нестандартным, выделиться от других.  
вот представь выходит человек, ты его не знаешь и он просто здоровается, говорит фио и погнал,  
а с другой стороны он пару минут говорит о себе кратко, но какие-то личные вещи, это распологает, сразу создается контакт.  
- Также важно рассказать о важности темы после расскза о структуре и о себе.  Постарайтесь максимально объяснить важность темы,  
в идеале показать, что это будет полезно слушающим.  
- весь рассказ не более 10 минут, все что более становится очень скучно. Лучше коротко до 10 минут, но интересно.  
когда 10 и более минут, то охота в какой-то момент сказать, ну что у тебя интересного, хватит столько речи.  
В идеале выкинуть лишние предложения, которые не важны. Оставить самое главное, что интересно.  
Самое плохо, когда текста на 5 минут и начинают разбавлять водой. Вот это плохо. Пусть будет 5 минут, но интересно.  
Принчип простой, меньше, да интереснее, мало кому интересно слушать детали того, как ты это получил и прочее, люди хотят кратко  
погрузиться в процесс получения результата и увидеть результат.  
В идеале 7-8 минут, интересные картинки, шутки, коротко о себе и структура выступления 1- 2 минуты, шаги как получался результат, и результаты минуты 3-4.  
- речь должна быть живой. вспомни сам, какое ощцщение, когда человек читает по бумажке или со слайда. это нудно. лучше что-то забыть и прсото  
сказать о чем-то другом или просто перепрыгнуть этот кусок, чем читать по бумажке. Вообще скорее всего даже не заметят, что ты что-то забыл сказать.  
- пара шуточек в процессе рассказа это круто. и в конце можно расскзать что-то интересное о результатах.  
- не нужно пытаться говорить как-то сложно, стараться быть диктором и прочее. Лучше свободно, на обычном простом языке.  
представь, что расскзавывешь другу или знакомому, не нужно официоза.  
- очень важно преследовать одну идею на протяжении всего выступления, это должна быть красная линия презентации.  
должно быть четко понятно зачем ты выступаешь, что хочешь сказать. В начале ставим проблему задачу, далее расскзаз кратко, как эта проблема решалась,  
и в конце выводы. Каждый слайд не отходит от задачи, немного в начале при рассказе о себе и в конце можно немного про интересные выводы отвлечься от задачи,  
а в остальном красной линией цель проходит через весь рассказ.  
- также каждый слайд преследует одну свою мысль. Одна презентация - одна цель. Один слайд - одна цель, которая ведет к общей цели.  
плохо, когда одна мысль разноситься между слайдами, может потеряться суть и слушатели запутаются.  
- KYSS-принцип — Keep Your Slide Simple. Не тратьте время на мишуру вроде оформления или сложного дизайна презентации. Это не конкурс художников. 

Структура слайда с графиком или картинками. 
Сверху слева заголовок
Далее слайде делится на две части, 66 на 34 примерно, слева 60 прцоентов для контент, схемы, картинки,  
а справа 34 (треть примерно) это мысль краткое пояснение  
Например, слева график и справа пишем словами выводы или какие-то показатели, например, график выручки, а справа от крафика итогове LTV и другие показатели.  
Это правильная последовательнсоть. Сначала читается заголовок (он важен и нужен, так как он несет в себе основную мысль слайда), далее человек спускается вниз глазами,  
смотрит на график или схему, и потом уже смотрит направо на показатели.  Также и в даш борде, лучше слева графики и правее него индикаторы и счетчики.  



Как держаться на выступлении
- Вначале нужно помнить, что изначально аудитория за вас. Вспомни, когда ты кого-то слушаешь, ты же не хочешь. чтобы он провалился. Вообще люди тратят свое время,  
и они думают, что это будет хорошо. Если человек начал слушать, то он уже за тебя. Это очень важно понимать.  
- будь максимально естественен. Если волнуешься сильно, то самое лучшее это сказать, ребят я так сейчас волнуюсь, это даже вызов для меня справится с этим волнением. Посмотрим как у меня это получится. И это реально даст буст.  
Хуже будет, если начинать пытаться скрыть волнение или вести себя неестественно. Если тебе важно донести какую-то идею до людей, то можно прям так и сказать, чем  
пытаться донести ее неестественно. Так и скажи, что мне важно донести до вас вот такую то идею, я волнусь по этому поводу. Но мы справимся.  
- Жестикуляция помогает донести мысль. Она должна быть спокойно и согласовывться с речью. В идеале жести показывают слова (образы).  
- Посомтри кто тебя слушает, постарайся понять кто они, эти люди прожили большую жизнь и сейчас тебя слушают. Прояви искренний интерес к людям, и они это заметят.  
- Не бойся говорить, что не получалось, говорить о провалах и неудачах. Люди охотнее симпатизируют тем, кто не идеальный.  И вообще когда человек говорит, что у него не   получалось, но он справился. то это вызывает доверие.
