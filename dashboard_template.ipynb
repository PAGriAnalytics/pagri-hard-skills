ВАЖНО  
Когда мы создаем дашборд, которым будут пользоваться разные люди, то полезно сделать выбор палеты цветов.  
И добавить палету для людей, которые плохо различают цвета.  

Основные моменты графического дизайна  
- Самое главное, чтобы мозг не терялся, когда первый раз видит дашборд, для этого нужно выстраивать иерархию используя размер шрифтов, расстояния между объектами, последоваательность расположения объектов слева направо, сверху вниз. И справа сбоку менее важное, чем слева в области 70-80% от экрана.
- Люди лучше воспринимают информацию от общего к частному
- Если на дашборде нет определенной иерархии, то пользователю будет сложно ориентироваться
- Праввило близости: если объекты расположить на одном растоянии друг от друга, то ни будут считываться как одна группа. А если часть из них отделить и расположить немного отдельно, то срауз мозг разделяет на разные группы. И меняя растояние между группами, мы можем задавать то, насколько эти группы похожи друг на друга.  
Все что мы хотим, чтобы пользователь воспринимал как целоое (связанное по какому-то признаку) должно быть близко друг к другу.
- Рамки для группировки плохо считываются и создают шум. Лучше испльзовтаь подложку (делать фон под группой другим, чем у дргуих объектов)
- Важно использовать иерархоично деление на группы  
    - Сначала делим при помощи растояния
    - Затем делим используя подложку внутри подгрупп, которые мы разделили расстоянием
    - И уже только если не хватает расстояния и подложки, используем рамку
- Если мы хотим сгруппировать группы по двум разрезам, то мы сначала разделяем по первой категории расстоянием, а потом уже подложкой или рамкой объединяем нужные элементы по второй категории (тут важно что деление идет на 2 группы в каждой категории, иначе будет месиво)
- Также можно внутри групп выделять цветом определенные элементы (можно просто более и менее контрастно выбирать цвета внутри одной группы, чтобы разделить визуально или привлечь внимание к определенным элементам)  
Либо также можно формой разделять, то есть часть элементов в группе сделать с квадратной рамкой, а другие с круглой.  
И также можно использовать размер, то есть одни элементы более крпные, а другие более мелкие, тогда внутри одной группы они будут визуально разделены.  
- Выделение размером не только разделяет на группы визуально, а дополнительно создает как бы уровень восприятия. То есть крупные объекты сразу воспринимаются как более значимые.    
- Правило размера относится и к заголовкам. Название дашборда самый крупный шрифт, разделы меньше, названия графиков ещё меньше и вспомогательный тест ещё меньше и подписи графиков меньше по размеру, чем название графика.  
- Мозг так устроен, что если на странице есть текст разного размера, то сначала он прочитает самый курпный текст, потом более мелкий и так далее.  
- Когда мозг смотрит на дашборд, то информация считывается слева направо и сверху вниз. И как бы по строчно. Как читается книга.  
То есть самое главное слева вверху, далее правее вверху и далее уже полосками вниз слева направо.  
Но тут все таки движение вниз имеет преимущество перед сильным движением вправо.  
То есть мы действительно сначала читаем слева направо, но примерно до половины страницы или до 60%, а далее глаза уже начинают переходить на следующую строку.  
Поэтому с самого правого края лучше распологать дополнительную информацию, а более главную слева сверху вниз.  
- фильтры, которые применяются ко всему датафрейму ставятся отдельно от всех объектов, например, справа вверху. А фильтры, которые примменяюся к конкретному объекту, ставятся рядом с этим объектом.  

Дашборды нужны для
- отслеживать оперативные показатели (приборная панель) Этот вид дашборда ближе к презентации, то есть за 3 минуты человек долежн понять что на нем представленно, как считаются показатели и так далее, то есть он должен быть простым и понятным. Это как спидометр в авто. Хорошоий показатель качественного операционного дашборда, когда чтобы принимать решения, не нужно постоянно менять фильтры, что-то выбирать и прочее.   
    - мониторинг проблем  
        по сути это просто таблички, где перед проблемными полями стоят цветные точки (или подсветки ячеек) (несколько градаций - сильная проблема, средняя проблема, нейтрально, хорошо, лучше плана) Это как лампочка в автомобиле, которая показывает состояние батареи, топлива или наличие неисправности. 
    - текущий статус  
        отображает нужные метрики целиком (вверху фактоиды) и в разных разрезах, которые отображают текущую ситуацию.   
        В этом дашборде минимум описания тектом, они скрываются за всплывающими подсказками если нужно, а для новых пользователей делается курс ознакомления.  Это классический спидометр в авто.  
    - сравнение с планом  
        - Тут наша задача просто показать, сравнение плана с фактом. Отлично работают подсветки как в дашборде с проблемами, с разной градацией.  
        По сути это разные метрики для которых вывденеы графики с линией плана и динамикой факта (небольшиие графики)
        - Либо может быть одна метрика, но в разрезе разных категорий и отображается сравнение плана и факта.  
        - В принципе такой дашборд может состоять из набора фактоидов и sparkline графиков вместе с фактоидами. 
- узнать что-то конкретное про отдельную сущнсоть (страница сущности) Это страница отображения информации о какой-то еденице в базе данных. Еденицей может быть какой-то срез (категория). Например, у како-то регион и все метрки по нему. И также это может быть конкретная метрика, которую на странице сущности раскладывается по разным срезам. Тут сущностью будет метрика.  
    - в один клик. В этом варианте, чтобы выбрать сущность, нужно просто применить один фильтр и получить срез по нужной метрике (выбрать город в выпадающем списке). Аналогия с путеводителем, когда мы хотим поехать в какую-то страну, то берем путеводитель и находим страницу про это страну.   
        - может выглядить как лонгрид дашборд (графики уходят вниз). В идеале должна быть менюшка, чтобы быстро перейти к нужно информации на странице. Это подобие сайта, где можно прокручивать вниз и получать более детальную информацию о сущнсоти. Вверху общая информация о сущности (сколько всего продаж и прочее), а ниже уже идет информация в разрезе разных категорий. Чтобы получить более детальную информацию. Вкладки более предпочтительно, чем лонгрид без меню перехода в разные участки страницы.  
        Если это лонгрид, то мы слева делаем графики, которые уходят вниз друг за другом, а справа вертикально друг над другом делаем фактоиды для разных метрик или разрезов, так как на лонгриде их будет много и они не войдут вверху в линию. По сути там обычно значение метрики и сравнение с планом или с предыдущим годом (любым периодом). Например, продажи за 2019 год и рядом изменение к предыдущему периоду. Ниже закупки и также сравнение с предыдущим периодом.  
        Пример
            - Закупки 2019 год  
            - Всего продажи правее Всего по точкам
            - Изменение в абсолютных значениях и правее также по точкам
            - Процентное изменение, например год к году и правее также по точкам.  
            То есть тут Вниз будут идти фактоиды для разных метрик, в каждом ряду слева общая информация, а правее уже если нужно в разрезе  
        Если у нас в виде вкладок, тогда фактоиды размещаем сверху в линию, так как тут их будет меньше и они хорошо входят в линию.  
        - либо это несколько страниц, на каждой уже информация о разных срезаж. На первой вкладке общая информация и далее на каждой вкладке, например, продажи по отделам, по поставщиками и прочее.  
    - сложносостовная. Это когда нужно применить несколько фильтров, чтобы получить нужную сущность. Аналогия, когда мы уже были в стране, но хотим узнать   
    температуру в каком-то городе. И тут нам нужно выбрать страну, город и узнать температуру.  
        - Слева делаем все фильтры которые нужны сверху вниз
        - Информацию лушче вообще в начале не выводить на дашборд (чтобы при запуске грузился быстрее), так как все равно искать будут не то, что отображается на старте
        - и правее уже идут графики в виде лонгрида или в виде мозайки для нужного среза.
        - В идеале сделать фильтры скрываемые после выбора, чтобы не отнимали место для изучения сущности.  
        - И тут часто справа не куча графиков. а один или несколько. Так как тут специфика в том, что все срезы выбираются и поэтому нужно  отобразить только конкретную информацию по выбранному срезу.  
        - В идеале метрики, которые будут отображены (графики справа) также отображаются фильтрами слева. Например checklist в вдие списка вниз с возможностью поставить галочки для нужных метрик и они отобразяться справа.  
        - Если фактоидов много, то можно главные разместить вверху, а остальные справа в полоске сверух вниз.
- получить какой-то инсайт, не зная что конкретно ищу и сделать выводы, имея в голове определенный алгоритм (аналитический инструмент)  
Это интерактивные дашборды, с большим количеством фильтров и должны позволять смотреть на данные с разных сторон.  
Должно быть много интерактивности (нажал на одном графике столбик - на другом отфильтровались данные и так далее)

Алгоритм составления отчета  
https://miro.com/app/board/o9J_kpOMVFA=/  
- 0. Составлять план отчета, проводить интервью с закачиком нужно, если дашборд будет использоваться не одни раз.  
Иначе это просто лишняя работа.  
Если нужно сделать дашборд одноразово, то подумать, возможно достаточно обычной презентации, где будет график и выводи общие рекомендации.  
Если действительно для ad-hoc задачи нужен отчет, то его нужно просто составить для ответа на поставленную задачу, тут важно даже не удобноство,  
а просто чтобы были нужные графики и нужные фильтры и интерактивность.  
Задача просто ответить на поставленные вопросы.  
Если действительно нужен отчет для многократного использования, то используем следующий алгоритм
- 1. Команда, то есть кто будет участвовать в команде по разработке дашборда. Возможные роли: заказчик, менеджер проекта/продукта, инженер даннх, аналитик, дизайнер, BI-аналитик, тестировщик, специалист по поддержке. Скорее всего все роли выполняют 1-2 человека, главное понять кто и какие будет выполнять. Создается общий чат, доска, канбан-доска или просто гугл-документ.
- 2. Проводим интервью с заказчиком и понимаем задачи дашборда. Тут универсальное правило - исполнитель понимает задачу. Профессионал не начинает работу, пока не убедится, что понял задачу, и что клиент согласен с этим пониманием. Тут очень важно, чтобы исполнитель проговорил, как он понял задачу. И заказчик подтвердил, что исполнитель правильно все понял. В идеале сразу накидать макет даже на бумажке примерный, просто основные блоки, так будет проще визуализировать задачу и примерное решение. Также тут нужно написать чем занимается отдел (заказчик) для которого будет создаваться дашборд. Это лучше делать до интервью, чтобы на интервью лучше все понимать. Если мы приходим со стороны, то важно понять специфику деятельности. Чтобы лучше понять задачу и цель. И важно также выячнить как задача решается сейчас, какие моменты нравятся, а какие нет и обговорить сроки реализации и лучше сразу их увелличить. Также очень важно выяснить не просто  нужные метрики и разрезы, но понять как именно нужно будет считвать эти значени. Это нужно будет анализировать в динамике или нужны буудт точечные значения. И в зависимости от этого для одних метрик нужны будут графики, для дургих таблицы или и то и то. После интервью нужно тезисно сотаить результат интервью (по сути вопросы, на которые хотят стэкхолдеры получить ответы и задачи которые нужно решить) и отправить заказчику. Это важно, чтобы убедиться, что мы все правильно поняли и записали.  
- 5. (это делается на интервью) Определяем на какие вопросы пользователь должен увидеть ответы и что происходит после того, как пользователь посмотрел на дашборд (можно добавить возможность по клику на ячейку, которая подсветилась красным, отправить сообщение кому-то или чтобы появлися контакт чей-то с кем нужно связаться, если эта метрика упадет, то есть тут мы как бы помогаем пользователю с действиями после того, как он увидел что-то важное на дашборде). То есть на этом этапе цели и портреты пользователей превращаются в конкретные действия. Исходя из набора этих действий и вопросов, проектируются конкретные визуализации и интерфейс дашборда. Примеры вопросов, на которые пользователь хочет увидеть ответы (эти вопросы мы по сути получаем от заказчика, когда проводитм интервью):
    - Как изменилась метрика Х относительно прошлой недели?
    - Какова доля продукта Х в общем портфеле?
    - Какая динамика метрики Х за последние 30 дней или за все время?
    - Что случилось с метрикой Х после запуска эксперимента?
    - В каких регионах наибольшее значение меттрики Х?
    - Если метрика Х отличается от плаана на 10%, то я звоню менеджеру и выясняю, что случилось.
- 6. (это делается на интервью) Реальность, источники и структура данных. Тут нужно выяснить какие данные есть, какие данные можно дополнительно получить, определить достаточно ли их для решения задачи.
- 3. Портрет пользователя. Определяем кто будет пользоваться дашбордом. Если пользователей мало, то обязательно включаем их в тестовую группу. Если их много, то выбираем по 3-5 представителей от каждой позиции. Тут важно, если группы разные (например, пользоваться будет и начальник, и отдел кадров и менеджены и аналитики), то в фокус группу включить представителей из кажой группы по несколько человек. Это очень важно, так как именно они смогут в процессе создания отчета корректировать процесс, чтобы результат был максимально удобен и понятен именно им.
- 4. Определяем контекст и формат взаимодействия с дашбордом. Это условия, при которых используется данный дашборд. Нужен для совещаний (то есть на совещании будет команда смотреть на дашборд на большом экране и обсуждать ситуацию, тут уже не нужны фильтры, нужны более крупные шрифты и определенное располоение графиков под большие мониторы) или для самостоятельного использования (тут уже будут фильтры и скорее всего использоваться будет на ноутбуке и для этого другие размеры и компановка). Нужен ли доступ с мобильного. Как часто будут использовать дашборд. Возможно дашборд будет использоваться не раз в како-то промежуток времени, а когда случиться какое-то событие. Нужно ли будет показыввать  отчет на больших экранах или на проекторе (это важно для определения цветовых решений и шрифтов).
- 7. Что происходит в реальном мире и создает данные. Реальность данных, то есть что в реальном мире генерит те данные, которые помогут ответить на наш вопрос. Понять весь процесс, по которому проходит одна еденица, например, от создания заявки до доставки товара. Так как это поможет определить данные, которых нет в полученных таблицах, но они оказывают существенное влияение. Выячнить как получить данные. К каким базам нужно подключиться. У кого взять логин и пароль.  
- 8. Разрабатываем основные блоки дашборда, типы графиков. Тут важно для каждого вопроса подобрать визуализацию и затем собрать их в блоки по вопросам.  Тут все строиться на ответах, которые мы получили на интервью. Есть метрики, категории и вопросы, на которые нужно ответить. Просто берем и подбираем самый подходящий график для каждого вопроса. Создаем набросок, возможно просто чертех с изображением графиков на дашборде. Но тут важно текст лучше напечатать, чтобы лучше читался, а сами графические элеементы накидать от руки.  
    - Как изменилась метрика Х относительно прошлой недели? - Фактоид с индикатором прироста.
    - Как менялась доля продукта Х в общем портфеле? - Area-chart во времени и можно все другие категории сделать бледными, а нужную ярче
    - Какая динамика метрики? - Фактоид + спарклайн.
    - В каких регионах наибольшее значение меттрики Х? - Бар-chart
    - Если метрика Х отличается от плаана на 10%, то я звоню менеджеру и выясняю, что случилось. - Таблица с отклонениями, фактом и планом. С возможнсотью сортировки по отклонению. С контактами менеджера.
- 9. Верстка и создание прототипа. Реализация дашборда. Блоки, которые придумали на предыдущем этапе, нужно расставить в согласии с направлением чтения, правил композиции, важности контекста. Схема составления дашборда
- Сначала определяем тип дашборда (операционный, страница сущности и так далее)
- Далее выибираем метрики и категории , которые будут отображаться на дашборде
- Затем определяем что будет более главное, что менее главное, какие фильтры
- Далее для каждой метрики выбираем подходящий график и для разбивки по категориям выбираем подходящий график или таблицу и думаем какие улучшайки сделать (добавить спарлайн, добавить подписи макс мин на графике и на фактоиде и так далее)
- Далее продумываем иерархию, что важнее, что менее важно и как их расположить на дашборде.
- Продумываем цветовые решения, то есть цвета фактоидов должны быть такими же как цвета этих метрик на графика, также категории должны быть одного цвета в разных блоках. Это убирает однообразие и дает связь элементов. И аткже это удобно не нужно на каждом графике думать какая категория где.  
В идеале в описании дашборда под заголовком написать что разбивака по таким то категориям и их написать цветом как на графике ниже.  
Тогда не нужно будет для каждого графика делать легенды. Но это только главные легенды, которые есть в нескольих метсах.  
Остальные легенды можно укзаывтаь прям в названии объекат, например зарплаты для отдела А, Б и С и эти отделы написать цветом как линии на графике.  
Либо можно не в описании, а у каждого элемнта в описании так сделать и далее в других элементах для этих легенд вообще ничего не писать.   
Вообще цвет это элемент кодирования, за счет которого мы можем облегчить восприятие и сократить количество текствой информации на дашборде.  
В одном месте написал цветом нуные категории и далее уже на других графиках просто используется цвет без подписей легенд.
- Хорошая практика все графики делать серым, если нужно в одном цвете и потом уже нужные элементы подствечивать другим цветом (например отклонение от плана). На серый цвет лучше всего ложаться другие цвета. То есть серый цвет используем как базовый и потом уже использем где нужно кодирование данных другими цветами. 
- Продумываем единый стиль. То есть если у нас несколько бар графиков, то они долны быть одного стиля (например, на обоих нет оси и есть тектовые подписи). Если есть линии, и на одной подписаны макс и мин или вместо легенд подписи возле линии, то на всех линейных графиках также.  
- И уже распологаем блоки учитывая последовательность считывания информации, используя разделение блоков расстоянием и расположением справа слева и внизу.  
- Сверху идут фактоиды
- Ниже идут метрики, которые указаны в фактоидах в виде графиков в времени
- Ниже уже идут разные разрезы этих метрик, либо справа или снизу добавляется таблица, в которой разбивка по нужным категориям и при нажатии на поле таблицы гарфики метрик фильтруются и показывают срезы.
- Можно  делать и отдельные графики для важных срезов, а остальные через таблицу и фильтрацию 
- Вырвниваем объекты и важно подписи для оси Y ровняем по левому краю, так смотрится намного аккуратнее.  
- Для адаптивной верстки (когда мы используем рэнж режим для размер) важно чтобы часть графиков была фиксированная, а часть растягивалась.  
И вот тут нужно для каждого ряда определить какие будут растягиваться и какие будут фиксироваться.

Основные моменты верстки
- табличку можно распологать либо ниже графиков, либо справа от них. Зависит от размера графиков, хватает ли им места слева и хватате ли места правее них для таблицы. Такой варинат хорошо считывается. Но если получается узко, тогда сначала идут графики сверху вниз и ниже таблица. А справа можно расположить доп информации, фактоиды, фильтры или графики и таблицы, которые занимают меньше пространства.  
- фактоиды лучше размещать вверху или в правой полоске шириной 20-30% от ширины экрана.
- если фильтров мало (операционный дашборд или страница сущности), то они распологаются справа вверху или спрва сверху вниз от самомго верха. А если фильтров много, то это скорее всего многосоставной дашборд, когда нужно сначала выбрать, что хотим увидеть, тогда фильтры идут в левой полоске (в идеале она скрываться должна)
- размерами графиков управляем важностью графика
- расстоянием между графиков и использованием подложки управляем группировкой графиков
- испоьзуем цвета для метрик одни и те же на всех графиках дашборда
- цвет названия метрики по возможности делаем такой же как и цвет на графике для лучших ассоциаций
- используем деленеие на страницы, когда информации сильно много и лонгрид дашборд становится сильно длинный и навигация затрудняется.
- сверху вниз идем от общего к частному и спрва, если есть блок, то там менее важная информация, чем слева сверху вниз идет в основной части экрана. Если снизу вне поля зрения экрана, то она уже менее значимая чем в блоке слева

Пример тех задания для создания дашборда

- Бизнес-задача: анализ взаимодействия пользователей с карточками Яндекс.Дзен;
- Насколько часто предполагается пользоваться дашбордом: не реже, чем раз в неделю;
- Кто будет основным пользователем дашборда: менеджеры по анализу контента;
- Состав данных для дашборда:
    - История событий по темам карточек (два графика - абсолютные числа и процентное соотношение);
    - Разбивка событий по темам источников;
    - Таблица соответствия тем источников темам карточек;
- По каким параметрам данные должны группироваться:
    - Дата и время;
    - Тема карточки;
    - Тема источника;
    - Возрастная группа;
- Характер данных:
    - История событий по темам карточек — абсолютные величины с разбивкой по минутам;
    - Разбивка событий по темам источников — относительные величины (% событий);
    - Соответствия тем источников темам карточек - абсолютные величины;
- Важность: все графики имеют равную важность;
- Источники данных для дашборда: cырые данные о событиях взаимодействия пользователей с карточками (таблица log_raw);
- База данных, в которой будут храниться агрегированные данные: дополнительные агрегированные таблицы в БД zen;
- Частота обновления данных: один раз в сутки, в полночь по UTC;
- Какие графики должны отображаться и в каком порядке, какие элементы управления должны быть на дашборде (макет дашборда):


Базово конфигурация любого дашборда состоит из 3-х блоков:

меню и фильтры;

переключение между дашбордами (кнопки навигации);

основное рабочее пространство с визуализациями, т.е. «тело» дашборда.

Основные задачи дашборда:

Визуализация ключевых метрик и показателей

Мониторинг текущего состояния бизнеса и операций

Отслеживание в динамике прогресса достижения целей и задач

Выявление трендов и паттернов в аналитике

Анализ целей и задач
На этом этапе необходимо определить:

кто будет целевой аудиторией дашборда? Сколько человек, какие роли им будут даны в системе?

какие цели и задачи пользователь или компания будут решать с помощью дашборда

какие ключевые метрики и показатели понадобятся? Сколько их будет?

Информация на дэшборде должна быть в логике от общего к частному. 

Распределение данных по важности сверху вниз. Сначала бизнес метрика (в идеале north star), затем экономические метрики.  
Причем распределение в строке по важности слева направо. Так как глаз движется сверху вниз и слева направло.  
Продуктовые метрики выводятся через дропдаун, если есть необходимость, на основной экран их выводить не нужно.  
Здесь пригодится пирамида метрик.   
Самая важная информация на дашборде должна находиться сверху и левее,   
чем она более детализированная и узкоспециальная – тем ниже и правее она располагается.   
Т.е. самая важная информация – в левом верхнем углу, самая узкоспециальная – в правом нижнем.  
Таблицы выводим только если это явно нужно, и тогда их выводим внизу.  

То есть слева вверху у нас выводиться значение главной бизнес метрики. Левее от нее в той же строке выводятся экономические метрики,  
из которых состоит бизнес метрика. Можно выводить числовое значение, можно индикаторы типа спидометров и прочее.  
Можно первый ряд числовые, а второй ряд спидометры и прочее, главное, чтобы это дополняло друг друго и доносило информацию.  
Далее ниже графики. Также по важности графика сверху вниз и слева направо.  
Можно даже разбить экран на 3 части.  
Вверху бизнес метрика и экономические метрики в числовом формате.  
Далее детализация этих показателей, но не сильная детализация, тут задача представить те же метрики.  
И третий блок это уже детализация информации и важные зависимости.  

Среднестатистический человек одновременно может воспринимать от 4-х до 7 объектов – и на дашборд лучше добавлять именно такое количество визуализаций.

При этом, старайтесь, чтобы дашборд нужно было минимально скроллить, и все данные умещались в один-два экрана.  
Если визуализаций много, и они не помещаются на экран, то лучше создать еще один логически связанный дашборд,   
на который пользователь сможет переключиться через панель закладок из основного дашборда.

Цвета должны сочетаться. Для подбора цветов можно использовать  
https://colorscheme.ru/  
https://gradients.app/ru/palette  

Общие рекомендации - дашборд не должен быть слишком пестрым или слишком однотонным.   
Лучше выбирать сочетающиеся оттенки или цвета из одной тепловой группы – теплые или холодные.   
Иногда используется такой прием: для большинства визуализаций применяются только неинтенсивные пастельные оттенки, а для ключевых деталей используется яркий цвет.

Элементы и Графики должны быть на одной линии, если их несколько в ряду. 

Цвета метрики на графике должен совпадать с цветом шрифта в цифровом блоке.  
Так намного удобнее воспринимать информацию.  
То есть мы выбираем цвет для какйо-то метрики и во всех местах дашборда все что с ней связано изображено этим цветом.  

Не нужно выводить все значения категорий, если их много.  
Все графики, где у нас идет сравнение категорий, делаем так, чтобы был топ 5-7 значений, а остальные можно посмотреть,  
нажав на кнопку посмотреть все. Делаем эту кнопку рядом с графиком.  
Если мы выводим топ чего-то и у нас не входят все записи, то есть мы выводим топ 5 значений категории,  
чтобы не перегружать дэшборд, и под этим топом распологаем кнопку 'Показать все' и если на нее нажать,  
то мы провалимся на другой уровень, где будет полностью все значения этой категории и можно дополнительно по ней ещё что-то выветси.  
Хорошая практика делать такие кнопки со словами показать все

Наглядность важнее точности.  
То есть если есть возомжность изобразить данные в виде, при котором будет считываться лучше, но не будет точности, то нужно это делать.  
Например, у нас есть информация какие вещи чаще покупают. Тут можно вывести бары со значениями,  
а можно вывести картинку целовека и на нем одежда и эта одежда подсвечена по шкале количества покупок.  
Пример синтетический, но суть ясна. Дашборд это больше визуальное считвание информации, поэтому нужно стараться сделать информацию более виуальной.  


Хорошая практика использование пикторамм, особенно из корпоративного брэндбука.  

Стараться доносить чиловую информацию картинками, то есть выводим число и инфографику, которая помогает воспринимать это число.  
Например у нас значение, что 7 человек из 100 покупают продукт.  
Мы можем вывести число 70 процентов и дополнительно картинку, на которой будет 93 кружка черные и 7 красные, а 3 нет.  
Или другое в этом же духе. Смысл в том, чтобы стараться перевести число в образ.  

Когда строим график линиями, то удобно в конце линии показывать значения, то есть не показывать все значения точек, а только последнее, тогда мы не загрезняем график и показываем самое важное, то есть последнее значение.  

Pie chart лучше вообще не испоьзовать, если уж у нас категории хорошо отличаются и их разницу видно, то можно использовать график бублик.  
Он лучше считывается, так как на нем мы оцениваем как бы длинну окружности, а не угол как у пай чарта.  
И также в бублике можно отобразить общую сумму в центре бублика.  
И важно упорядочить размеры категорий по часовой стрелке от большого к маленькому. И начинать от 12 часов самый большой сегмент.  
И значения нужно подписывать возле секторов. Не нужно делать сбоку легенду, это очень неудобно и приходиться прыгать туда сюда.   

Когда мы хотим отобразить план факт значения в виде процентов, то не нужно использовать всякие бары и подобное.  
Нужно просто отобразить числов и знак процента. Это отлично считывается, а визуализация даже лишняя.  
Вот если нам нужно отобразить план факт числами не процентами, то тут можно использовать бары с наполнением, но тоже

Если у нас бар чарт или другой чарт, где можно отобразить значения метрики у точек, то лучше убирать осеь метрики, чтобы место не занимала.

Для area chart и для подобных графиков лучше ставить прозрчность 0.5, когда у нас несколько цветов рядом,

Это хорошо экономит место на дашборде.

И для фактоидов важно добавлять знаки + - и % для процентных изменений. Значек валюты. Графические элементы с цветом, чтобы показать динамика вверх или вниз. Разбивку по категориям, если их много и ещё и есть несколько категорий, то лучше изображать в виде табличке.  
То есть у нас строки это категории (может быть составной индекс при разбивке по 2 категориальным переменным), а столбцы это нужные метрики  и также миниграфики (могут быть sparkline, мини бар чарты, мини боксплоты) для каждой строчки.  
Это удобно и считывается хорошо, когда нам нужно отобразить все категории или их много.

На дашборде вертикальный скролл допустим, а вот горизонтальный нет. Так как он очень неудобен.  

На дашборде очень важны padding, за счет них мы используем правило внутреннее - внешнее.  
То есть мы как бы отделяем визуально объекты внутри от снаружи.  
Для этого нужно делать padding у всех элементов.  
Например если мы сделаем фактоид и не сделаем padding, то он будет прижат и к названию и к графику ниже.  
Это очень частая ошибка. Все логические блоки должны иметь padding.  

Отступы между элементами на дашборде (padding) также называют паузами визуализации. Чтобы человек мог прочитать объект, далее была бы пауза и потом новый объект.  

Если мы делаем фактоиды и рядом с одним из фактоидов хотим сделать sparkline, то нужно расстояние между фактоидами сделать больше, чем расстояние между фактоидами и sparkline для одной метрики.  

Часто бывает удобно убирать один из графиков метрики в область фактоидов и делать sparkline, но важно не для линий.  
Если у нас какие-нибудь ration или просто бары и нам не важна динамика (то есть маленький график отобразить все хрошо), то его можно поместить рядом с его фактоидом.  
Так мы освобождаем место ниже для других элементов.  
Если у нас вверху не входит, то можно либо в фактоидах отобразить только числа, а графике уже ниже, либо если у нас есть место справа, то использовать это место, то есть часть отображзить вверху (самые главные метрики) с миниграфиками (в основном это бары или боксполты, то есть график не должен зависеть от высоты), а справа уже остальные фактоиды и графики.  
Важно. что не нужно для всех фактоидов делать графики рядом. Все таки основные должны быть в областе графиово, чтобы их можно было удобно изучать, а уже остальные для экономии места отправляем к фактоидам и тоже чтобы они доносили информацию, просто так они там не нужны, лучше вообще убрать и без графика.

Очень удобно названия графиков и таблиц на дашборде делать в виде вопроса, на который этот график или таблица должны ответить.

Фактоиды конечно называем просто по названию метрики.  

Нужно стараться не дублировать информацию, например, если у нас несколько фактоидов и у них у всех внизу стоит разница в процентах от среднего,

Достаточно оси на нижнем графике и выровнять их все, чтобы бары или точки были друг над другом.
