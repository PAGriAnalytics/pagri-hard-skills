Когда мы создаем дашборд, которым будут пользоваться разные люди, то полезно сделать выбор палеты цветов.  
И добавить палету для людей, которые плохо различают цвета.  

Пример тех задания для создания дашборда

- Бизнес-задача: анализ взаимодействия пользователей с карточками Яндекс.Дзен;
- Насколько часто предполагается пользоваться дашбордом: не реже, чем раз в неделю;
- Кто будет основным пользователем дашборда: менеджеры по анализу контента;
- Состав данных для дашборда:
    - История событий по темам карточек (два графика - абсолютные числа и процентное соотношение);
    - Разбивка событий по темам источников;
    - Таблица соответствия тем источников темам карточек;
- По каким параметрам данные должны группироваться:
    - Дата и время;
    - Тема карточки;
    - Тема источника;
    - Возрастная группа;
- Характер данных:
    - История событий по темам карточек — абсолютные величины с разбивкой по минутам;
    - Разбивка событий по темам источников — относительные величины (% событий);
    - Соответствия тем источников темам карточек - абсолютные величины;
- Важность: все графики имеют равную важность;
- Источники данных для дашборда: cырые данные о событиях взаимодействия пользователей с карточками (таблица log_raw);
- База данных, в которой будут храниться агрегированные данные: дополнительные агрегированные таблицы в БД zen;
- Частота обновления данных: один раз в сутки, в полночь по UTC;
- Какие графики должны отображаться и в каком порядке, какие элементы управления должны быть на дашборде (макет дашборда):


Базово конфигурация любого дашборда состоит из 3-х блоков:

меню и фильтры;

переключение между дашбордами (кнопки навигации);

основное рабочее пространство с визуализациями, т.е. «тело» дашборда.

Основные задачи дашборда:

Визуализация ключевых метрик и показателей

Мониторинг текущего состояния бизнеса и операций

Отслеживание в динамике прогресса достижения целей и задач

Выявление трендов и паттернов в аналитике

Анализ целей и задач
На этом этапе необходимо определить:

кто будет целевой аудиторией дашборда? Сколько человек, какие роли им будут даны в системе?

какие цели и задачи пользователь или компания будут решать с помощью дашборда

какие ключевые метрики и показатели понадобятся? Сколько их будет?

Информация на дэшборде должна быть в логике от общего к частному. 

Распределение данных по важности сверху вниз. Сначала бизнес метрика (в идеале north star), затем экономические метрики.  
Причем распределение в строке по важности слева направо. Так как глаз движется сверху вниз и слева направло.  
Продуктовые метрики выводятся через дропдаун, если есть необходимость, на основной экран их выводить не нужно.  
Здесь пригодится пирамида метрик.   
Самая важная информация на дашборде должна находиться сверху и левее,   
чем она более детализированная и узкоспециальная – тем ниже и правее она располагается.   
Т.е. самая важная информация – в левом верхнем углу, самая узкоспециальная – в правом нижнем.  
Таблицы выводим только если это явно нужно, и тогда их выводим внизу.  

То есть слева вверху у нас выводиться значение главной бизнес метрики. Левее от нее в той же строке выводятся экономические метрики,  
из которых состоит бизнес метрика. Можно выводить числовое значение, можно индикаторы типа спидометров и прочее.  
Можно первый ряд числовые, а второй ряд спидометры и прочее, главное, чтобы это дополняло друг друго и доносило информацию.  
Далее ниже графики. Также по важности графика сверху вниз и слева направо.  
Можно даже разбить экран на 3 части.  
Вверху бизнес метрика и экономические метрики в числовом формате.  
Далее детализация этих показателей, но не сильная детализация, тут задача представить те же метрики.  
И третий блок это уже детализация информации и важные зависимости.  

Среднестатистический человек одновременно может воспринимать от 4-х до 7 объектов – и на дашборд лучше добавлять именно такое количество визуализаций.

При этом, старайтесь, чтобы дашборд нужно было минимально скроллить, и все данные умещались в один-два экрана.  
Если визуализаций много, и они не помещаются на экран, то лучше создать еще один логически связанный дашборд,   
на который пользователь сможет переключиться через панель закладок из основного дашборда.

Цвета должны сочетаться. Для подбора цветов можно использовать  
https://colorscheme.ru/  
https://gradients.app/ru/palette  

Общие рекомендации - дашборд не должен быть слишком пестрым или слишком однотонным.   
Лучше выбирать сочетающиеся оттенки или цвета из одной тепловой группы – теплые или холодные.   
Иногда используется такой прием: для большинства визуализаций применяются только неинтенсивные пастельные оттенки, а для ключевых деталей используется яркий цвет.

Элементы и Графики должны быть на одной линии, если их несколько в ряду. 

Цвета метрики на графике должен совпадать с цветом шрифта в цифровом блоке.  
Так намного удобнее воспринимать информацию.  
То есть мы выбираем цвет для какйо-то метрики и во всех местах дашборда все что с ней связано изображено этим цветом.  

Не нужно выводить все значения категорий, если их много.  
Все графики, где у нас идет сравнение категорий, делаем так, чтобы был топ 5-7 значений, а остальные можно посмотреть,  
нажав на кнопку посмотреть все. Делаем эту кнопку рядом с графиком.  
Если мы выводим топ чего-то и у нас не входят все записи, то есть мы выводим топ 5 значений категории,  
чтобы не перегружать дэшборд, и под этим топом распологаем кнопку 'Показать все' и если на нее нажать,  
то мы провалимся на другой уровень, где будет полностью все значения этой категории и можно дополнительно по ней ещё что-то выветси.  
Хорошая практика делать такие кнопки со словами показать все

Наглядность важнее точности.  
То есть если есть возомжность изобразить данные в виде, при котором будет считываться лучше, но не будет точности, то нужно это делать.  
Например, у нас есть информация какие вещи чаще покупают. Тут можно вывести бары со значениями,  
а можно вывести картинку целовека и на нем одежда и эта одежда подсвечена по шкале количества покупок.  
Пример синтетический, но суть ясна. Дашборд это больше визуальное считвание информации, поэтому нужно стараться сделать информацию более виуальной.  


Хорошая практика использование пикторамм, особенно из корпоративного брэндбука.  

Стараться доносить чиловую информацию картинками, то есть выводим число и инфографику, которая помогает воспринимать это число.  
Например у нас значение, что 7 человек из 100 покупают продукт.  
Мы можем вывести число 70 процентов и дополнительно картинку, на которой будет 93 кружка черные и 7 красные, а 3 нет.  
Или другое в этом же духе. Смысл в том, чтобы стараться перевести число в образ.  

Когда строим график линиями, то удобно в конце линии показывать значения, то есть не показывать все значения точек, а только последнее, тогда мы не загрезняем график и показываем самое важное, то есть последнее значение.  

Pie chart лучше вообще не испоьзовать, если уж у нас категории хорошо отличаются и их разницу видно, то можно использовать график бублик.  
Он лучше считывается, так как на нем мы оцениваем как бы длинну окружности, а не угол как у пай чарта.  
И также в бублике можно отобразить общую сумму в центре бублика.  
И важно упорядочить размеры категорий по часовой стрелке от большого к маленькому. И начинать от 12 часов самый большой сегмент.  
И значения нужно подписывать возле секторов. Не нужно делать сбоку легенду, это очень неудобно и приходиться прыгать туда сюда.   

Когда мы хотим отобразить план факт значения в виде процентов, то не нужно использовать всякие бары и подобное.  
Нужно просто отобразить числов и знак процента. Это отлично считывается, а визуализация даже лишняя.  
Вот если нам нужно отобразить план факт числами не процентами, то тут можно использовать бары с наполнением, но тоже  
лучше перевести фактическое число в проценты и отобразить только его.  
Это важно, так как так быстрее считать информацию, чем сравнивать сколько осталось до плана.  

Area chart хорошо показывает общую динамику (самая верняя линия) и отображает как примерно разбиты значения по категориям.  
Но анализировтаь динамику каждой категории очень сложно и сравнивать значения между категоирями тоже.  
Но все таки, когда цель показать общую динамику и разбивку,то он полезен.  
Если есть место, то можно построить рядом line и area.  
Тогда мы будем видеть и динамику каждой категории и общую суммарную динамику и разбивку по категоирям. 

Если у нас есть 2 линии и нам нужно сравнить их между собой или одна является как бы целевой.  
То тут можно одну отобразить в виде line а другую в виде area.  
Тогда удобно считывается выход за границы area.  

Числовое представление KPI и фактоидов (это просто отображение фактического значения) удобно делать так
- сверху пишем за какой период отображется значение (бледным цветом)
- далее ниже не бледным цветом название метрики словами
- далее ниже не бледным значение метрики 
- далее ниже отображем значок либо треугольник вверх маленький, либо вниз и правее от него процентное изменение от другого числа (определяется что нужно)  
- и далее ниже можно бледным цветом отобразить непосредственно занчение, от которого считаем процентное изменение метрики и рядом можно отобразить насколько изменилось
