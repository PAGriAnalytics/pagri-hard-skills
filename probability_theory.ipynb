Вероятность — это количественная мера возможности наступления определённого события. Она определяется как отношение числа благоприятных исходов к общему числу возможных исходов.  
Вероятнсоть - это число от 0 до 1, которое показывает, насколько вероятно, что какое-то событие произойдет.  
Вероятность помогает оценить, насколько вероятно, что произойдет какое-то событие.

Шанс — это всегда вероятность успеха, чего-то хорошего, желательного, ожидаемого человеком.   
Шанс - это состоявшаяся вероятность.  
А вот вероятность — слово нейтральное. Степень возможности, с которой то или иное событие произойдет или не произойдет, варьируется вне зависимости о того, хорошее это событие или плохое.  
Если событие А наступило 4 раза, а событие Б 7 раз, то  
- вероятность наступления события А равна 4 / 11
- шанс наступления события А равен 4 к 7

чтобы ответить на вопрос какая вероятнсоть наступления события, нужно знать историю наступления этого события в прошлом. То есть чтобы ответить на вопрос какова вероятнсоть встретить динозавра будет близка к нулю, так как если взять всех людей, которые выходили на улицу за последние 100 лет, то почти все не встречали динозавра.

С точки зерния теории вероятнсоти, можно ещё сказать так: вероятность события, которое невозможно по определению (например, встреча с динозавром в наше время), всегда равна нулю. Это не требует статистического анализа.

В теории вероятностей допущение о вероятности элементарных событий является ключевым для решения задач.  
Поэтому прежде чем решать задачу, нужно иметь вероятность элементарных событий.  
Без этих допущений мы не сможем рассчитать вероятность более сложных событий (например, выпадения четного числа на кубике или двух орлов подряд на монетке).

Теория вероятностей позволяет вычислять вероятности сложных событий, основываясь на вероятностях более простых (элементарных) событий, из которых эти сложные события состоят.



Эксперимент — это выполнение определённых действий или наблюдение за процессом, в результате которого может произойти или не произойти какое-то событие. В теории вероятностей эксперимент часто называют случайным экспериментом, потому что его исход заранее неизвестен и зависит от случая.

Примеры экспериментов:

- Подбрасывание монеты (исход: орёл или решка).
- Бросание игральной кости (исход: число от 1 до 6).
- Измерение температуры воздуха в определённый день (исход: конкретное значение температуры).

случайная величина — это величина, которая принимает определённые значения в результате эксперимента. Эти значения зависят от исхода эксперимента и могут быть как числовыми, так и категориальными (например, "орёл" или "решка").

случайная величина в ходе эксперимента может принимать разные значения - это исходы
А вот событие это множество исходов, оно может совпадать с исходом, а может состоять из нескольких исходов

Событие — это подмножество всех возможных исходов эксперимента. Оно может состоять из одного или нескольких исходов. Событие может произойти или не произойти в результате эксперимента.  
- При подбрасывании монеты событие "выпал орёл" состоит из одного исхода: {орёл}.
- При бросании игральной кости событие "выпало чётное число" состоит из трёх исходов: {2, 4, 6}.

Зная вероятности исходов, мы можем определить вероятность события. Это одна из ключевых идей теории вероятностей

Большинство задач теории вероятностей строятся на том, что нам известны вероятности элементарных исходов, и нужно вычислить вероятность наступления определённого события. Это основа основ теории вероятностей. 

Чтобы определить вероятность чего-то нам нужно либо знать вероятность наступления элементарного исхода для этого, либо провести эксперимент и примерно определить вероятсноть наступления события? иначе мы не можем говорить о вероятнсоти события.

Если у нас нет ни теоретической модели, ни экспериментальных данных, то говорить о вероятности события действительно затруднительно. 

в задаче либо должно быть указана верояность элементарных исходов, либо указано, что элементарные исходы равновероятны и мы сами можем посчитать эту вероятнсоть и далее уже решать задачу о вероятности наступления какого-то события.

Комбинаторика помогает определить m и n для формулы вероятности.

Комбинаторика лишь предоставляет инструменты для подсчёта количества возможных исходов, которые затем используются в формулах теории вероятностей. 

Комбинаторика хорошо работает, когда все исходы равновероятны (например, подбрасывание монетки или игральной кости). Однако, если вероятности исходов различны, комбинаторика становится неприменимой.

Комбинаторика предполагает, что события независимы. Если события зависят друг от друга, комбинаторика не подходит.

Нужно различать:
- Количество пермутаций Permutation (n!)

- Количество сочетаний Combination $C^k_n = \frac{n!}{k!(n-k)!}$

- Количество размещений Acomodation $A^k_n = \frac{n!}{(n-k)!}$


Количество размещений вычисляется как n! делим на (n-k)! так как в данном случае нам важен порядок размещения k элементов и на них мы не должны делить, а делим только на (n-k)! так как эти значения мы не трогали и нужно сократить на их комбинации, а количество сочетаний мы должны делить и на k! дополнительно, так как на не важен порядок первых k элементов, поэтому нам нужно делить и на их количество возможных комбинация

Биномиальный эксперимент

Если у нас n исходов и мы проводим k независимых экспериментов, то у нас будет $n^k$ исходов.

### Задачки на вероятность последовательности событий. Формула Бернулли.

Формула Бернулли в теории вероятностей позволяет вычислить вероятность того, что событие A произойдет ровно k раз из n независимых испытаний. 

$P(k) = C^k_n * p^k * (1-p)^{n-k}$

Тут важно понять, что вероятность одного события из k успехов и n-k неудач будет равна $p^k * (1-p)^{n-k}$  
Но у нас может быть много комбинаций таких событий. Наступит или первое или второе и так далее, то есть нужно сложить все эти вероятности.  
И так как все эти вероятности одинаковые, то мы умнажаем на количество комбинаций.

### Совместные и независимые события.

### Зависимые и независимые события

Событие A называется независимым от события B, если условная вероятность события A равна его безусловной вероятности:

$P(A|B) = P(A)$

### Условная вероятность

Условной вероятностью P(B∣A) события B при условии, что произошло событие A, называется отношение вероятности произведения этих событий к вероятности события A:

Проще запомнить так  
Чтобы получить вероятнсоть события B при условии, что A произошло нам нужно произведение разделить на вероятность события A, чтобы с обоих сторон в итоге было вероятнсоть B

$P(B|A) = \frac{P(AB)}{P(A)}$

$P(A|B) = \frac{P(AB)}{P(B)}$

### Вероятность произведения событий

$P(AB) = P(A)*P(B|A) = P(B)*P(A|B)$  

$P(ABC) = P(A)*P(B|A)*P(C|AB)$

### Вероятность суммы событий

$P(A+B) = P(A) + P(B) - P(AB)$

### Вероятность появления хотя бы одного события

Если события A1​,A2​,... ,An​ имеют одинаковую вероятность, равную p, то вероятность появления хотя бы одного из этих событий равна 

$P(A) = 1 - q^n$  

где $q = P(\overline{A_i}), i = 1, 2, ..., n$

### Формула полной вероятности

Гипотезы ( H_i ) не должны происходить одновременно (то есть они должны быть попарно несовместными), но при этом они могут быть зависимыми.
И гипотезы должны образовывать полную группу событий.  
Гипотезы - это дополнительные условия, которые могут повлиять на исход опыта.

$P(A) = P(H_1)P(A|H_1) + P(H_2)P(A|H_2) + ... +  P(H_n)P(A|H_n)$

По сути у нас есть вероятность наступления события А и есть гипотезы.  
И вероятность наступления события А равна сумме вероятностей наступления события А в этих гипотезах.  
тут важно не путать условную вероятность наступления события А при условие наступления гипотезы, с вероятностью события А в этой гипотезе

В формуле полной вероятности каждое слагаемое действительно можно интерпретировать как вероятность наступления события ( A ) при условии, что произошло одно из событий ( B_i ), образующих полную группу несовместных событий. 

### Формула Байеса

Формула Байеса позволяет разворачивать условную вероятность.  
То есть если мы не знаем вероятность события А при условии Б, но знаем вероятность события Б при условии А,  
то используя формулу Байеса мы можем вычислить вероятность события А при условии Б.

Можно сказать, что H тут гипотеза, а A - событие.  
То есть есть событие, которое наступило и есть гипотеза, что до этого случилось что-то ещё.  
По сути в знаменателе полная верояность, а в числителе часть полной вероятности для гипотезы H (то есть вероятность события А в гипотезе H)

$P(H|A) = \frac{P(H)*P(A|H)}{P(A)}$

Чтобы запомнить числитель, нужно просто помнить, что в числителе вероятность для гипотезы H.  
Поэтому там будет вероятность гипотезы H умножить на вероятность события A при условии, что гипотеза H верна.
