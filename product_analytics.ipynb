Задача любого бизнеса -  увеличить прибыль.  
Но чтобы бизнес начал приносить прибыль, нужно найти пользователей, которые будут его покупать.  
Чтобы найти пользователей, нужно понять их боль, что им нужно. И далее , нужно создать продукт, который будет решать эту проблему.  
Нужно определить где могут находиться пользователи, которым нужен наш продукт, чтобы решить их проблему.  
Далее нужно провести опрос и убедиться, что наш продукт действительно решает проблему.  
Далее нужно создать mvp (минимально жизнеспособный продукт), который будет решать проблему.  
Это этап проверки соответсвия нашей идеи проблемам пользователей. Может ли она его решить.
Если наши предположения о продукте верно (что он позволит решить проблемы пользователей), то пользователи будут покупать его.  
Далее нужно развивать продукт и масшатбировать его.  
Это этап достижения соответствия продукта рынку. Когда мы уже не для пары пользователей сделали продукт и убедились, что он решает их проблемы, а уже для большого количества пользователей. И тут возникает множество проблем, даже если продукт решает проблемы пользователей. Это и конкруенция, привлечение клиентов, масштабирование и другое.  
На данном этапе компания может полностью изменить свою идею, это называется pivot.  


Основные вопросы создателей продукта и продукт менеджеров
- Нужно ли людям то, что я хочу сделать?
- Кто эти люди и где их найти?
- Каким должен быть продукт, чтобы его захотели купить?

Customer development - это подход к созданию стартапов и новых продуктов, когда через взаимодействие с потенциальными пользователями валидируются гипотезы о проблемах потребителей, ее решении, рынке, каналах привлечения. 

Задача продукт менеджера, чтобы продукт развивался не на предположениях команды, а на реальных данных, которые получены при взаимодействии с пользователями.  
Иначе высока вероятность сделать то, что не нужно пользователями.  

Очень важно сегментирование пользователей.  
Сегментирование позволяет выявить именно тот сегмент, которому будет интересен наш продукт.  
Далее сегментация позволяет уже внутри имеющегося сегмента (потенциальных пользователей) выделить группы, с которыми нужно взаимодействовать по разному.  
У пользователей разные запросы, потребности, мотивации. Разным пользователям нужен разный функционал продукта. Эти пользователи несут для компании разную ценность.  

Задача маркетинга - донести ценность продукта до потенциальных клиентов, чтобы они купили его.

### Фреймворки для взаимодействия с пользователяим

#### Метод персон  

Метод персон  
Собирается образный персонаж потенциального пользователя, он максимаьно реалистичен. Он описывает поведение потенциальной группы пользователй.  
Суть в том, что мы уходим от многих людей к одному, так намного проще продумывать что нужно конкретному пользователю. А затем масштабируем на всех.
- имя
- фотография
- демографические данные
- предыстория
- цели
- мотивация
- барьеры

После того как мы сотавили портрет потенциально пользователя, то нужно проникнуться его проблемами.  
Нужна эмпатия.  
Эмпатия - осознанное сопереживание эмоциональному состоянию другого человека.
Виды 
- эмоциональная (когда мы на уровне эмоций чувствуем человека)
- когнитивная (когда мы глазами другого человека смотрим на мир и за счет этого понимаем его мысли, чувства, но без своих эмоция, рационально)

Для развития продукта важна когнитивная эмпатия, так как привлечение эмоций может мешать и искажать действия.  

Карта эмпатии

Представляем какого-нибудь воображаемого конкретного нашего покупателя

Чтобы лучше запомнить, можно представить говолоку и сверху идти.  
Вверху мозг (что думает, чувствует), далее глаза (что видит), далее рот (что говорит), далее уши (что слышить) и внизу тело (что делает).  
Тут важно, чтобы предолжения, которые мы пишем в каждый раздел перекликались с другими.  
То есть мы пишем, что он что-то видит или слышит, то скорее всего он и говорить будет на вещи, связанные с этим.  

И отвечаем себе на следующие вопросы
- Что он думает и чувствует?  
что на самом деле важно, основные занятия, проблемы и радости
- Что он видит?  
среда, друзья, что предлагает рынок, где живет и работает, 
- Что он говорит и делает?  
как держит себя в обществе, как выглядит, отношение к окружающим
- Что он слышит?  
что говорят друзья, что говорит начальник, что говорят влиятельные лица

Также представляем какая у него боль, страхи, разочаровния и препятствия

И думаем какие у него достижения, потребности, желания, мерило успеха, препятствия

Теперь когда мы много  знаем о нашем покупателе, мы можем пойти от частно к общему.  
Мы можем понять
- что он хочет
- в какой момент он этого хочет
- в каком канале он этого хочет
- как это должно выглядеть
- как мы его можем завлечь

В итоге мы получили возможные точки роста.  
Мы знаем как сделать так, чтобы она принесла нам больше денег.  

Дополнительно можно придумать целую историю о придуманном пользователе.  
Например, у нас аналитик Миша, который живет в новом жк, увлекается спортом, но времени мало.  
Переехал Миша в новый жк, а там велодорожка или спорт плошадка классная. И захотел Миша купить себе то-то и то-то и так далее приудмываем.  
В итоге можно так набрести на отличные точки роста.  

Далее представляем других наших покупателй, отличающихся друг от дурга.  
В итоге мы получаем сегментацию наших покупателей.
