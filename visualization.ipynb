быть разные зависимости, главное. что мы пока не детализируем, а хотим понять динамику в целом, как будто у нас нет категорий.   
А потом уже начинаем детализировать и смотреть продажи в разрезе разных категорий.

Какие графики используем
- Изменения во времени  
Не забываем range date plotly
    - line plot with markers, bar plot    
    Обязательно сравнить бары и линию, и выбрать то, что лучше доносит информацию.  
    Часто линия с маркерами (есть маркеров не много, то ещё цифры возле маркеров, кстати тогда ось можно убрать) лучше показывает динамику,  
    или сравнение двух категорий во времени. Бары они ботлично работают, когда мало столбиков, когда их много, то тут линия лучше.  
    Так как если много баров, то уже появляется динамика, а тут линия лучше.  
    Важно помнить, что линейный график лучше покажет динамику, когда у нас например все значения  
    больше 4000 и изменяются от 4000 до 4500, а бар плот в этом случае покажет высокие стобики и будет плохо видно разницу.  
    Наша задача подать информацию, чтобы она считывалась так, как нам нужно. Поэтому строим линейный и бар и сравниваем какой нам  
    больше подходит .  
    - area plot  
    Когда хотим показать не тренд, а объем  
    - calmap  
    `pip install calmap`  
    Во временных графиках могут быть тренды, цикличность, сезонность. Это все нужно показать.    
    calmap (календарный heat map) помогает определить цикличность в данных  
    - priestley timeline
    Отлично подходит, когда у нас есть начало события и его продолжительность  
    На графике будут закрашенные участки, когда событие происходит и будет пусто в обратном случае  
- Отклонения 
    - Bar plot  
    Берем вычитаем среднее из всех значений и у нас будет отклонение от среднего.  
    Также можем взять любое значение, от которого хотим посмотреть отклонения и вычитаем его из всех наших значений и строим бар плот.  
    Тогда у нас будет бары вверх и вниз от этого значения, которое мы вычли. Очень информативно.  
- Корреляция
    - Scatter plot    
    - Connected scatter plot  
    Если мы хотим проследить не только просто как выглядит зависимость двух переменных, но и хотим увидеть какая точка за какой появлялась,  
    то можно сделать scatter и наложить на него line plot (можно просто линейный и вклчюить маркеры), тогда у нас точки будут связаны линией.  
    Если данные не хаотично появлялись, то можно проследить некоторые зависимости.  
    То есть таким образом мы вносим временную шкалу на график scatter plot.  
    Например, у нас две совокупности точек, просто точечный график покажет две кучки,  
    но как понять как эти кучки накапливались, может сначала одна накопилась, а потом другая, а может нет, вот это очень хорошо показывают  
    точечный график плюс линейный
    - heat map
    Возможно не нужно выводить все значения, так как симметрично одинаковые значения, лучше выбрать верхний или нижний треугольник и   
    показать только его, а остальную часть оставить пустой 
    Этот тип графика отлично работает после pivot table. Мы берем группируем по индексу и по колнкам, то есть у нас категориальные переменные,  
    и также у нас занчения, которые дает аггрегирующая функция. Вот мы и получили тепловую карту. Строки будут по строкам, столбцы по столбцам,  
    уникальные значения этих категорий как раз будут подписями ячеек, а значения подкрасятся от минимума до максимума по шкале.  
    vmin, vmax: Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
    - scatter plot с третим измерением (bubble plot)
    Хорошо подходит, когда у нас есть 2 переменные и есть тертья, если третья числовая, то можно добавить размер каждой точки,  
    в зависимости от значения этой третьей переменной, а если третья категориальная, то можно добавить цвет.  
- Отношение часть - целое  
    - pie chart  
    не увлекаемся доляи, используем, только если явно видна разница между долями, иначе график бессмысленный
    - stacked bar plot  
    по сути это детализированный по како-то переменной Pie chart  
    То есть у нас есть какие-то значения и есть катеориальная переменная, это будет обычный пай чарт,  
    но если мы добавим время или другую переменную, то у нас будет stacked bar plot  
    информативность графика будет, только если есть разница между облостями, иначе лучше строить бары рядом
    - normolized stacked bar plot
    аналогично предыдущему, но мы нормализуем данные по каждому столбику от 0 до 100 например,  
    и наносим на график, тогда у нас будут все слобики от 0 до 100 и внутри каждого будет деление по категории  
    В этом графике важно понимать, что не нужно сравнивать размеры столбиков как абсолютные значения,  
    это нормализованные значения, тут нужно смотреть у кого больше доля той или иной категории.

- Распределения
    - Гистограмма  
    - kde plot
    - qqplot
    - distplot
    это гистограмма и линия kde вместе  
    - boxplot  
    хорошо показывает размах значений внутри квартилей  
    но он не покзаывает распределение, поэтому если мы его хотим добавить в отчет, то  
    лучше делать joint plot, чтобы был бокс плот и сверху линия kde (или гистограма), это будет информативнее.  
    - violin plot 
    лучше использовать его вместо box plot, так как он содержит в себе и бокс плот и kde  
    И все это в одном графике, а не joint, так как joint все таки содержит 2 графика и нужно сопоставлять.  
- Сравнение категориальных переменных и одной числовой  
    - treemap  
    отлично подходит, когда у нас есть иерархия (по сути есть 2 категориальной переменной и числовая)
    - barplot, lineplot with markers
    Обязательно посмотреть оба варианта, если баров много, то линия часто лушче преподности информацию   
    Важно отсортировать данные перед построением графика, чтобы визально удобно было сравнивать  
    - px.density_heatmap  
    Удобно использовать, когда у нас ong-form (то есть есть две колонки категориальные и одна числовая)  
    Ему не нужно преобразовывать данные в матрицу.  
    - heatmap  
    Можно представить, что у нас для каждого значения по оси, есть много категорий, то мы можем поставить их друг на друга (сделать стэк),  
    можем расположить рядом столбики, а можем значения разместить друг над другом квадратиками и будет heatmap.  
    Отличная вещь, если у нас больше 3х значений в категориях и на бар плоте будет уже плохочитаемо.  
    Мы просто строим сводную таблицу по категориальным признакам, выбираем агрегацию и строим тепловую карту.  
    В итоге мы можем сравнить значения в для пар категорий.  
    Если у нас есть время, то это не  значит что в тепловой карте его только по оси икс делать.  
    Самый лучший способ - это по оси y откладывать категорию, у которой длинные названия значений, так как они будут горизонтально и  
    лучше читаться.   
    vmin, vmax: Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
    - sankey diagram  
    Можно одну категориальную переменную сделать источником, а вторую целью.  
    А числовую сделать значением в sankey диграмме.  
    Хорошо покажет зависимости между категориями  
    Дополнительно вывести процент связи от всего источника
    - seaborn.stripplot (plotly.express.strip)
    В seaborn Не забываем про параметр dodge, который может разделить категории внутри другой категории
    Это можно представить как бар чар, но вместо бара распределение точек по категории  
    По стуи это бар плот для анологичной ситуации, когда у нас высота столбика это числовая пременная,   
    а другие перменные категориальные и мы разбиваем на столбики по этим категориальным переменным.   
    Но stripplot нам дает не агрегированную информацию по какой-то категори как это делает бар плот,   
    по сути он берет сумму, среднее, количество и просто агрегирует.    
    stripplot покажет распределение точке без агрегации и разобьет по категориям   
    - seaborn.swarmplot (как seaborn.stripplot, но точки разделяет, чтобы не слипались)  
    Можно использовать, когда нам не охота делить точки с помощью dodge в stripplot, но  
    хотим, чтобы точки не слипались
    - с помощью slider можно добавить на график ещё одну переменную, например мы сравниваем продажи по отделам и в разрезе пола
