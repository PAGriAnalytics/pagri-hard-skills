Сквозная аналитика - это когда мы идентифицируем пользователя и отслеживаем его поведение в разных системах.  
Системы могут быть самые разные, и билинговые системы, макретинговые каналы, всякие скидочные карты, которые привязываются к клиенту,  
crm система и прочее. И во всех системах мы одного пользователя идентифицируем одинаково.  
У нас как бы пользователь насквозь проходит через все системы и мы получаем сквозную аналитику по пользователю.    
Сквозная аналитика очень важна, так как она нам делает пользоваетелй действительно уникальными.  
Если бы не было сквозной аналитики, то пользователь зашел с пк, мы его считаем одним пользователем, зашел с мобайла, считаем другим и так далее,  
в итоге у нас пользователь реально только 1, но мы в системе считаем много пользователей.

Следующий подход к анализу называется расчётом экономики одной продажи, или юнит-экономикой (от англ. unit economics, «единица»; «экономика»). Цель: рассчитать количество единичных продаж, которого хватит, чтобы выйти в плюс.
- расходы нужно делить на постоянные и переменные.
Постоянные расходы не зависят от объема продаж (аренда, зар.плата, расходы на электичество и другие постоянные расходы), а переменные зависят (себестоимость, маркетинкогвые расходы, если они не постоянные, доставка, упаковка и прочее).
- посчитать сколько выручки поступает с одной еденицы товара/услуги. Постоянные расходы не учитываем.   
Тут важно взять все переменные расходы на одну еденицу, вычесть из выручки за одну еденицу.
- если с одного товара мы не получаем прибыль, то нужно либо сокрашать себестоимость, либо увеличивать цену.
- если прибыль с одного товара есть, далее определяем сколько нам нужно продавать товаров / услуг, чтобы выйти в ноль (x * выручку за еденицу - постоянноые расходы)
- увеличение расходов будет вести к прибыли, только расходы должны быть переменные, которые зависят от объема продаж.

Юнит-экономика — это ответ на вопрос, зарабатываете вы на вашем пользователе (юните) или нет. Чтобы понять это, вам нужно выяснить:

- Сколько денег вы потратили на привлечение этого пользователя;
- Сколько денег вы заработали с этого пользователя.

Margin – это разница между ценой продукта и его себестоимостью, маржа.   
Простыми словами – то, сколько компания накидывает, чтобы заработать. Может выражаться как в деньгах, так и в процентах.  
Есть Margin (Маржа) и Markup (Наценка).  
В абсолютных цифрах Маржа и Наценка одинаковы.  
Маржа (М) = Цена (Ц) - Себестоимость (С).  
Наценка (Н) = Цена (Ц) - Себестоимость (С)  
Но чаще всего они выражаются в процентах, и тут они разные, поскольку   
Маржа выражается в % по отношению к цене,   
а Наценка - к себестоимости.  
Не делайте ошибки, которая может Вам стоить серьезных проблем!  
Маржа в процентах называется маржинальностью  
Маржинальность - показатель прибыли компании в общей стоимости товара или услуги, выраженный в процентах   
М% = (Ц-С) / Ц * 100%  
Н% = (Ц-С) / С * 100%   


Turnover – это оборот компании, объем всех продаж (net sales) за определенный период **поступивших в кассу или на расчетный счет**.   
Этот показатель никак не затрагивает расходы организации. Выражается в деньгах.    
Оборот — денежные средства, поступающие на расчетный счет и в кассу предприятия  

Revenue (Выручка) — средства, полученные компанией от продажи работ, услуг, товаров, имущества, инвестиционной деятельности и т.п

Термин «оборот компании» часто путают с выручкой, что некритично. Объединяет их одно – поступление денег.   
Оборот отражается на расчетном счете, а выручка – это получение дохода от всего в целом: продажи товара, оказания услуг, инвестиционной деятельности.

Profit – это чистая прибыль компании; то, что остается от оборота (доходов) после вычета всех расходов. Выражается в деньгах.



Рентабельность - это доля чистой прибыли в выручке.

Большинство количественных наблюдений фиксируют в абсолютных величинах. Абсолютная величина — это объём, величина или размер наблюдаемого явления или события. Например, итоговая стоимость в рублях или количество покупок в штуках.

Если взять стоимость покупок и разделить на их количество, получим среднюю стоимость купленного товара. Это будет относительная величина — показатель, который даёт числовую меру соотношения двух сопоставимых величин.

При расчёте относительных показателей важно, чтобы их компоненты были связаны. Например, стоимость всех покупок можно разделить на их количество — эти величины сопоставимы. А вот отношение суммы из чека к числу пройденных до магазина шагов имеет мало отношения к продуктовой аналитике.

Метрики делятся на 
- метрики действия. Дают понятную мотивацию влиять на продукт. Мы понимаем как мы через них можем повлиять на продукт. (Retention rate, CR, CAC)
- метрики тщеславия. Они кажутся приятными, их можно репортить в отчетах, рассказывать о них, но не говорят какие действия нам нужно предпринять (DAU, показы рекламы, количество скачиваний, просмотры страниц, количество зарегистрированных пользователей)

Хорошая метрика для конкретной задачи
- релевантна этой задаче
- понятна бизнесу
- чувствительна к изменениям (иначе мы не сможем по ней определить дают ли дополнительную пользу наши изменния для пользоватлея)  
И для каждой задачи нужно выбирать наиболее чувствительную метрику для этой задачи.  
Например, если мы хотим улучшить воронку, то наиболее чувствительной метрикой будет конверсия.
- управляема (если мы не можем управлять метрикой, то мы не сможем ее улучшить)

Даже DAU не является метрикой действия, так как он может расти за счет усилий маркетинга, но при этом у нас может быть сильный отток клиентов и прочее, то есть само по себе количество пользователей не говорит, что мы можем за счет его увеличения увеличить прибыль.

Это позволяет выделить ключевые метрики, которые оказывают наибольшее влияние на доход или эффективность продукта.

$$ \text{K(фактор)}=\text{X}*\text{Y}*\text{Z} $$  
X - % пользователей, которые пригласили других пользователей  
Y - среднее количество людей, которых пригласил один пользователь  
Z - число пользователей, принявших приглашение  
Виральность (virality) — это стремительное и широкое распространение контента, идей или продуктов через цифровые платформы, часто благодаря сарафанному радио и социальным сетям.

далее те, кто дал 9 и 10, считаются лояльными, а те, кто дал 0 - 6 - не лояльны

NPS = % of promoters - % of detractors

$$ \text{CSAT} = \frac{\text{Количество оценок 7, 8, 9, 10}}{\text{Количество ответов на вопрос}} * 100 $$

ARPU
$$\text{ARPU} = \frac{\text{Revenue}}{\text{users}} $$

$$\text{ARPDAU} = \frac{\text{Revenue}}{\text{DAU}}$$

В зависимости, что мы берем в `users` мы можем получить    
ARPU - все пользователи за все время

ARPUx - доход с привлеченного пользователя за x дней с первой сессии (или момента регистрации)

ARPPU
$$\text{ARPPU} = \frac{\text{Revenue}}{\text{paying users}} $$

то
$$\text{ARPU} = \text{ARPPU} * \frac{\text{paying users}}{\text{users}} = \text{ARPPU} * \text{first conversion} $$

ARPPU также можно записать через среднее количество покупок APC 
$$\text{ARPPU} = {\text{APC}}*{\text{Average price}} $$

APC (Average Purchase Count) — это среднее число покупок.  
Отношение количества заказов к количеству покупателей за период  
Этот показатель позволяет оценить, сколько покупок клиенты в среднем совершают за определённый период.
$$APC = \frac{Orders}{Buyers}$$

ARPPU также можно записать в следующем виде  
av.price - средняя цена  
cogs (cost of good sold) - стоимость оказания услуги или товара (себестоймость 1 продажи или услуги)  
apc - среднее число продаж на клиента  
1cogs - особенные затраты на 1 сделку (например, покупки по промокоду, тестовые периоды пользования услугой, мы при этом бесплатно оказываем всю услугу или часть)      
cogs и 1cogs не имеют отношения к маркетингу  
затраты на маркетниг учитываются в CPA и CAC.

$$\text{ARPPU} = (\text{av.price} - \text{cogs}) * \text{apc} - \text{1cogs} 

Общая формула для выручки будет выглядеть так  
UA - количество активных пользователей   
CPA - стоимость привлечения 1 активного клиента, тут целевое действие будет покупка  
$$\text{gross profit} = \text{UA} * (\text{ARPU} - \text{CPA})

Тогда выручку можно записать  
С1 (first conversion) - это conversion rate (конверсия) первой покупки.  
$$\text{gross profit} = \text{UA} * (((\text{av.price} - \text{cogs}) * \text{apc} - \text{1cogs}) * С1 - \text{CPA})

Margin (в смысле маржинальности, то есть отношение маржи к выручке)  
Маржинальность - показатель прибыли компании в общей стоимости товара или услуги, выраженный в процентах
$$Margin(\%) = \frac{Average Price - COGS}{Average Price}$$
$$Margin(\%) = \frac{Revenue - Costs}{Revenue} = \frac{Margin}{Revenue}$$

Costs
$$\text{Costs} = {\text{Marketing Costs}} + {\text{COGS}}*{\text{Orders}} + {\text{1st COGS}}* {\text{Buyers}} +  {\text{fix COGS}}$$  
fix CCGS - постоянные издержки (например, аренда офиса, зарплаты персонала, управленчиские расходы, то есть постоянные расходы существования бизнеса)

среднее арифметическое всех длин сессий за некоторый временной интервал. Вычисляется по формуле: $$ASL = \frac{T}{N}$$

Формула для расчёта LTV (customer lifetime value):

LTV = Доход за весь период «жизни» покупателя − Затраты на его привлечение и удержание

Для расчёта LTV необходимо знать доход от одного человека (ARPU), суммы, которые тратятся на его привлечение (CAC) и удержание (CRR).

Существует несколько формул расчёта LTV, которые различаются данными, сложностью и точностью результата. Вот некоторые из них:

Простая формула, которая поможет вычислить средний доход от одного человека за конкретный промежуток времени: LTV = Доход за период ÷ Количество купивших в этот период.

Формула, которая показывает, сколько прибыли принесёт один заказчик за всё время: LTV = Время сотрудничества человека с брендом × Средний доход от него.

Формула, которая помогает прогнозировать развитие бизнеса в ближайшей перспективе: LTV = Средний доход от человека × Период сотрудничества × Количество повторных заказов.

Чтобы выбрать наиболее подходящую формулу, нужно ориентироваться на специфику бизнеса, метрики, которые знает предприниматель, и цели, для которых делается расчёт.

LTV = средняя стоимость покупки * количество покупок в год * средняя продолжительность взаимодействия с клиентом.
Рассмотрим эту формулу на примере магазина бытовой техники.   
Если клиент покупает 2 товара в год со средней стоимостью 9 000 рубле на протяжении пяти лет,   
значение его LTV составит: 9 000 * 2 * 5 = 90 000 рублей.

Lifetime — это метрика, которая показывает, сколько дней  
среднестатистический пользователь пользуется вашим приложением от

Однако, ждать последнего входа пользователя часто приходится долго,  
поэтому этот показатель обычно определяет период неактивности, после  
которого пользователь считается «отвалившимся».    
Посчитать Lifetime можно так:  
    - Мы определяем некоторый период неактивности, то есть время, после  
    которого пользователь скорее всего уже не вернётся в приложение.  
    Определяют это либо на основании значений Retention, либо, что  
    чаще, «экспертно». Обычно «экспертно» это значение задают  
    равным одной или двум неделям.    
    - Каждый день мы смотрим на пользователей, у которых в этот  
    конкретный день истек период неактивности.  
    - Для каждого пользователя вычисляем количество дней от его  
    первого визита до текущего дня.  
    - Рассчитываем среднее значение по всем пользователям.   
    В итоге получаем среднее вермя жизни пользователей за выбранный период, которые не появлялись больше чем у нас отсечка, это и есть Lifetime. 

Ну а ARPU (в данном случае ARPU = ARPDAU) рассчитывается как дневной  
Revenue, делённый на DAU. Умножаем Lifetime на ARPU и получаем LTV

Чтобы рассчитать lifetime, нужно проинтегрировать функцию Retention Rate за определенный период  
Дело в том, что Lifetime — это площадь фигуры под графиком Retention (зависимость Retention в процентах от времени),  
иначе говоря — интеграл от Retention по времени.  
Нужно найти фиункцию для графика retention и посчитать интеграл
Функция имеет вид  
$$\text{F}(\text{t}) = \text{A} + \frac{\text{B}}{\text{t} + \text{C}}$$  
где A, B, C — константы, которые надо найти.  
Но функция может быть разная, нужно подбирать коэффициенты и смотреть ложаться ли точки на кривую, если нет, то нужно использовать другое уравнение для функции, например  
$$\text{F}(\text{t}) = \text{A} + \frac{\text{B}}{\text{t}^{\text{D}} + \text{C}}$$    
$$\text{F}(\text{t}) = \text{A} + \frac{\text{B}}{\ln{\text{(t)}^{\text{D}}} + \text{C}}$$    
$$\text{F}(\text{t}) = \text{A} + \frac{\text{B}}{\ln{\text{(t + E)}^{\text{D}}} + \text{C}}$$    
оценивать качество подгонки можно с помощью коэффициента детерминации R^2, который показывает, насколько хорошо данные соответствуют модели.  
Или просто по сумме квадратов отклонений.  
Нужно взять разные уравнения и посмотерть у каких меньше сумма квадратов отклонений и выше коэффициент детерминации.  
И тут ещё важно, что чем на большем промежутке мы будет считать, тем результаты могут быть разные, то есть на коротком промежутке одна функция может быть лучше, а на большом промежутке другая.  
Поэтому нужно понимать за какой период мы будем в итоге считать.

ВАЖНО  
Если у вас мало исходных точек, то лучше использовать ту
кривую, в которой меньше коэффициентов.  

Далее ищем эти константы, интегрируем и считаем lifetime

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import quad

# Определяем модель функции retention
def model(X, A, B, C):
    return A + B / (X + C)

# Пример данных о retention (замените на ваши данные)
X_data = np.array([1, 2, 3, 4, 5])  # Дни
Y_data = np.array([0.8, 0.6, 0.4, 0.3, 0.2])  # Retention

# Указываем начальные параметры
initial_params = [1, 1, 1]  # Начальные значения для A, B, C

# Находим коэффициенты с увеличением maxfev
try:
    popt, pcov = curve_fit(model, X_data, Y_data, p0=initial_params, maxfev=10000)
    A, B, C = popt
    print(f"Найденные коэффициенты: A = {A}, B = {B}, C = {C}")

    # Предсказанные значения
    Y_pred = model(X_data, A, B, C)

    # Сумма квадратов ошибок (SSE)
    sse = np.sum((Y_data - Y_pred) ** 2)
    print(f"Сумма квадратов ошибок (SSE): {sse}")

    # Коэффициент детерминации (R²)
    ss_total = np.sum((Y_data - np.mean(Y_data)) ** 2)  # Общая сумма квадратов
    r_squared = 1 - (sse / ss_total)
    print(f"Коэффициент детерминации (R²): {r_squared}")

    # Определяем функцию для интеграции
    def retention_integrated(X):
        return model(X, A, B, C)

    # Интегрируем функцию от 0 до бесконечности
    lifetime_value, error = quad(retention_integrated, 0, X_data[-1])

    print(f"Lifetime Value (LTV): {lifetime_value}")

    # Для визуализации
    X_fit = np.linspace(0, 10, 100)
    Y_fit = retention_integrated(X_fit)

    plt.scatter(X_data, Y_data, label='Данные о retention')
    plt.plot(X_fit, Y_fit, color='red', label='Подходящая модель')
    plt.xlabel('Дни')
    plt.ylabel('Retention')
    plt.legend()
    plt.title('Retention Function')
    plt.show()

except RuntimeError as e:
    print(f"Ошибка оптимизации: {e}")
```


- LTV = Средний доход от человека × Период сотрудничества × Количество повторных заказов

Когда мы хотим из метрик каждой когорты получить общую метрику на каждый период жизни когорыт, то мы можем взять
- среднее
- медиану
- квантиль
- средневзвешенное

Среднее брать не правильно, так как тут мы не учитываем размер когорт  
И если у нас в какой-то когорте придет мало людей и у них будет большой ретеншн (на малой выборке может быть такое), то они исказят картинку  
Поэтому нужно брать средневзвешенное значение.  
Для этого мы берем значение метрики умножаем на количество людей в когорет складываем так все когорты и затем делим на общее количество людей во всех когортах  
Например у нас в первый день пришло 100 человек и на второй день осталось 60
во второй день пришло 10 человек и на второй день осталось 80  
тогда средневзвешенное будет = (60*100 + 80*10)/(100+10) = 68.18

Очень полезно, когда считается не просто среднее значение по когортам, а средневзвешенное.  
То есть если мы хотим посчитать среднее значение по всем когортам какого-то показателя, то мы не просто суммируем все значения в когортах и делим на  
число когорт, мы взвешиваем какждое значение когорты.  
Есить 2 способа
- умножать значение каждой когорты на ее размер и потом всю эту сумму разделить на общий размер всех когорт
- сразу дать веса каждой когорты, то есть определить долю каждой когорты в общем.  
Берем размер когорты и делим на размер всех когорт. И потом умнажаем значения в когортах на коэффиценты.  

LTV расчитывается по той же схеме как и RR(retention rate). Только вместо количества вернувшихся пользователей, у нас размер выручки.  
которую они принесли, размер этой выручки мы также делим на размер когорты. чтобы оценить LTV накопительным итогом. 

Метриками лояльности являются churn rate и retention rate.

retention может быть
- классическим, когда мы просто смотрим сколько у нас осталось пользователей на n день после регистрации
- rolling, когда мы для каждого дня считаем сколько пользователей появлялись в продукет после n дня до конца окна расчета rolling  
То есть если у нас окно месяц, то на 10 день мы будет брать всех пользователей, которые зашли в приложение с 10 по 31 день  
Можно сказать, что это кумулятивная сумма пользователей, которые зашли в приложение с 10 по 31 день  
Rolling retention позволяет быстро определить какая доля пользователей больше не появлялась в продукте.  
То есть если у нас на 10 день rolling retention равен 60% значит с этого дня до конца у нас появлялись 60% пользователей, значит 40% пользователей больше не появлялись в продукте.

Пирамида метрик  
каждый нижележащий уровень вносит вклад в более верхний уровень  
- Бизнес метрики
Тут находится сама цель существования бизнеса, то за чем следят топ менеджеры    
Целевая метрика задает вектор, на который опираются нижележащие метрики  
Выручка  
- Экономические метрики
Экономические метрики напряму вытекают из бизнес цели (выручки) через формулы  
MAU, DAU, ARPU, LTV, LT    
Например Выручка = MAU * ARPU * LT    
вот прямая зависимость бизнес метрики от экономических метрик      
Поэтому экономические метрики анализируются бизнесом для принятия решения о развитии бизнеса   
- Продуктовые
Продуктовые метрики формируют экономические метрики, но формул связи между ними может не быть  
То есть для продуктовых метрик целевыми будут экономические метрики  
Продуктовые метрики очень специфичные и различаются у разных бизнесов   
визиты, время в сервисе, частотность, установки, удаления, retention rate, churn rate, показатель отказов сервиса,  
разные конверсии, доступность сервиса, CSI (Customer Satisfaction Index), Индекс потребительской лояльности (готовность к повторным покупка) (NPS)

Бизнес и экномические метрики практически всегда одни и те же и они не меняются   
А вот продуктовые метрики выбираются исходя из корреляции с экномическими метриками  
Для этого нужно взять экномические метрики (MAU/DAU, ARPU, LTV, LT) и все имеющиеся продуктовые метрики  
И построить матрицу корреляции  
И оставить только те продуктовые метрики, у которых хотя бы с одной экономической метрикой корреляция больше 0.7 (число выбирается индивидуально)

Матрица вовлеченности - показывает зависимость процента mau (dau, wau и др) от количества совершенных действий.  
То есть, например, у нас приложение заказа еды. Мы мониторим вход в приложение, открытие страницы с товарами, покупку и прочее.  
Мы берем и считаем число пользователей от всех активных пользователей за определенный период (% dau например), которые совершили это действие  
хотя бы раз за период и также считаем сколько раз в среднем они это действие сделали (берем сколько всего совершено этих действий и делим на число 
уникальных пользователей совершивших это действие)
И можем построить график, по оси x будет процент от общего количества пользователей, которые совершили это действие, а по осси y будет среднее  
число соверешных действий на одного пользователя.  
Мощный инструмент показывет  
- Какая доля пользователей пользуется определенной фичой?
- Насколько эта фича популярна среди тех, кто хоть раз ее попробовал (то есть сколько в среднем раз используют)

Когда мы расчитываем какие-то метрики, то важно отобразить изменение метрики во времени, также можно построить violine и  
обязательно вывести среднее значение метрики или моду (если распределение метрики ненормальное) за все время, так как числовое значение нужно для расчета других экономических метрик  

Когорта в аналитике — это группа пользователей, объединенных общим признаком.   
Для создания когорты нужно, чтобы пользователи совершили одинаковое действие в один период времени.  
ВАЖНО - когорта всегда создается по времени, без времени, это просто категории.

Когортой будут люди, которые прожили 1, 2, 3, 4 и так далее периодов. И мы смотрим динамику по месяцам первого визита.  
То есть когорта это именно люди, которые прожили определенный период.  

Очень важно понимать, что мы значение метрики для групп проживших 1, 2, 3 и так далее делим именно на количество пользователей в нулевой период  
Так как нам нужен общих знаменатель для всех. чтобы сравнивать разыне когорты, иначе если бы мы каждую когорту делили на количество пользователей в ней,  
то у нас бы было все запутно и непонятнО, так как в разные периоды может быть разное коилчество пользователей и нет общего знаменателя.  

Когортный анализ — это один из методов исследования поведения потребителей.   
Он показывает, как ведут себя узкие группы людей (когорты),   
которые совершили определенное действие в какой-то конкретный промежуток времени.
Это позволяет маркетологам выявить закономерности, которые могут быть неясны при рассмотрении общих данных,   
а также понять поведение конкретных групп клиентов.  

Когортный анализ — это маркетинговый инструмент, с помощью которого исследуют поведение реальных и потенциальных клиентов.

Пользователей объединяют в группы (когорты) на основании действий, которые они совершили в определённый период.  
Важны именно эти две составляющие: временной промежуток и конкретные цели, которые были выполнены.

Когорты могут быть разными, в зависимости от того, какие показатели хочет проанализировать маркетолог:

люди, которые зашли на сайт после рекламы у крупного блогера и совершили покупку спустя несколько месяцев;

пользователи, установившие приложение и сделавшие заказ сразу после этого события;

получатели рассылки, которые воспользовались присланным кодом на скидку в офлайн-магазине.

Сколько сессий в день можно нести два смысла
- сколько всего сессий в среднем проходит в день  
покажет среднюю нагрузку в день на сервис
- сколько в среднем совершает сессий 1 пользователь в день  
покажет поведение посетителей на платформе (как часто  они в днеь появляются)   

Важны оба ответа.  

Сколько длится одна сессия?
- рисуем гистограму или violine
- вычисляем все средние меры и выбираем наиболее подходящую из них и приводим число  
Если среднее, мода и медина примерно равны, то можно брать среднее.  
Но если они разные, то нужно подумать какое лучше взять и обосновать выбор.  
Как всегда пишем причины, почему выбрали эту метрику.  
Тут важно какой вопрос, если вопрос сколько обычно длиться одна сессия, и у нас ненормальное распределение длительностей сессий,  
то среднее значение будет некоректно, так как обычно в данном случае это будет мода, так как среднее будет  
смещенно вправо, елси у нас экспоненциальное распределение времени.  
Вот этот момент очень важен.  
Так как среднее введет в заблуждение.  
Обычно сессия длить как раз модальное значение.  

Как часто люди возвращаются?
- расчитываем retention rate, для этого
- необходимо сделать когортный анализ по времени (по месяцам или неделям или дням и тп)  
Чтобы сделать когортный анализ, мы по одной оси откладываем даты, а по второй оси откладываем номер месяца,  
таким образом мы получим для одной даты разные значения retention rate черзе 1, 2, 3, 4 и так далее месяцев  
Будет динамика по месяцам от стартового месяца.  
Также дополнительно можно сделать тепловую карту времени и какой-нибудь каетгории (например возраст, пол польователей),  
retention rate можно считать просто относительно предыдущего месяца (тут уже не будет динамики по месяцам)

Ответ среднее значение, тепловая карта выводы  
Для описания тенденции, берем среднее значение за первый период (после нулевого) и смотрим как дальше относительно этого значения  
идут изменения

Когда люди начинают покупать?
Нужно ответить на 2 вопроса
- Через сколько времени после первого посещения пользователь совершил первую покупку?  
- В какую по счету сессию люди начинают покупать?
Ответом будет среднее значение метрики (обоих вариантов) и выводы  
Дополнительно можно попробовать построить бар чар и посмотреть эти значения в разрезе какой-то категории

Сколько покупок в среднем совершает пользователь?   
среднее количество покупок = общее количество покупок / количество уникальных пользователей

Сколько раз покупают за период?  
то есть сколько в среднем один покупатель делает покупок за период  
Для выбора периода нужно понять какой период нам лучше подходит, например если у нас данные за год,  
то период 6 месяцев будет нормально. 
- Суммируем все покупки за все периоды и делим на количество покупателей, которое было в нулевой период  
Берем суммирем все покупки за первые 6 месяцев всех пользователей и   
делим на общее число пользователей в начале периода (в первый месяца)   
Важно уточнение за какой период  
Для расчета показателей за период X месяцев можно использовать только данные  
- по когортам, которые взаимодействуют с компанией в течении X месяцев или более  
- за первые X месяцев жизни когорт  

Ответ среднее значение метрики и выводы

Какой средний чек?
- суммируем все покупки (получаем полную выручку) и делим на количество покупок  
Ответ среднее значение (тут не нужно медиан и прочего, так как это устоявшееся понятие и берут среднее),  
строим график изменения во времени, можно сделать когортный анализ и делаем выводы

Сколько денег приносят пользователи (LTV)  
В финансовом менеджменте ответ на вопрос когда, также важен как ответ на вопрос сколько.  
Тут важно указывать за какой промежуток, так как LTV считается за период и еденица периода должна идти в ответе.    
Например, 3 y.e. это неверно    
правильно будет 3 y.e./год   
Тут у нас еденица расчета ltv это год     
Для выбора периода нужно понять какой период нам лучше подходит, например если у нас данные за год,  
то период 6 месяцев будет нормально. 
И нельзя сравнивать ltv за разные периоды.  
Если мы выбрали год, то мы берем отсечку год от первого сессии и считаем сколько всего люди все потратили за это время  
и делим на количество людей в первый период.  
Очень частая ошибка считать сначала среднее по разным периодам (например среднее людей, которые прожили месяц, потом 2 и так далее),  
а потом усреднять эти значения.  
Тут нужно суммировать все доходы от всех и разделить на количество пользователей на старте.  
Когорты это хорошо, но их нужно считать отдельно и не выводить из них общее среднее для всех.    
Чтобы проверить правильно ли мы рассчитали LTV можно сравнить его с   
LTV = среднее число покупок за выбранный период на одного пользователя * средний чек   
До этого мы как раз посчитали среднее число покупок за период на одного пользователя и средний чек.  

Маркетинг
- Сколько денег потратили всего / на каждый источник / по времени (значит нужно изучить как данные по расходам в всего изменяются во времени)  
Ответ среднее занчение, график, выводы (можно сравнить с общей прибылью и можно cac сравнить с ltv)  
график затрат на маркетинг не только по периодам, но ещё график затрат по источникам (бар плот и heatmap)  
и сделать вывод какие источники самые затратные, какие менее затратные  
- сколько стоило привлечение одного **покупателя** в целом и из каждого источника CAC (customer acquisition cost)  
Важно, что метрика рассчитывается по покупателям, поэтому нам нужно использовать только тех, кто совершил хотя бы одну покупку  
Сложность в том, что пользователи заходят с разных источников.  
Возможные решения этой проблемы    
    - вычислить в какую именно сессию была совершена покупка и привязать ее именно к этому источнику (с которого эта сессия произошла)
    - привязать пользователя к тому источнику, с которого он чаще заходил  
    - привязать пользователя к источнику, с которого он первый раз зашел на сервис     
    этот метод предпочтительнее    
Ответ - среднее значение метрики в целом и по разным источникам  
таблица средних значений метрики по источника или/и график (бар плот)
выводы
- На сколько окупились расходы (ROI)    
Окупаемость считаем не от выручки, а от LTV (в целом и по источникам)  
Не забывать про сопостовляемый период (как с LTV)
Ответ - среднее значение метрики, таблица средних значений метрики по источникам и графики (бар плот и heatmap)

Все метрики нужно посмотреть в разрезе разных параметров (которые есть, например, по платформам, полу пользователй и другим категориям) и сделать выводы,  
о том как параметр влияет на каждую метрику. Посмотреть как они меняются во времени.

Выводы должны не просто описывать графики, а содержать рекомендации.  
И общий вывод должен включить все метрики, которые мы получили (числами).  
И сделать вывод основываясь на всех полученных метриках, а не просто сказать, что вот этот канал лушче других.  
То есть в какой канал рекламы лучше направить больше выручки, какая категория больше приносит выручки (можно сделать предоложения почему). 
И итоговый вывод это всегда рекомендация что можно сделать, чтобы улучшить сервис, основываясь на полученных данных.  

Сначал смотрим общую картину. Общую выручку, общие расходы, рентабельность и другие бизнес метрики, а потом уже спускаемся к детальным метрикам.  
Хорошая проверка, это когда мы сначала просто смотрим выручку и расходы и получаем вывод о окупаемости сервиса.  
А потом когда разобрали по метрикам, то результат должен сходиться, не должно быть так, что выручка - расходы говорит, что все плохо,  
а выводы по метрикам, что все хорошо.
